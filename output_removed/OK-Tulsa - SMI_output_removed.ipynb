```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT


def main():
    # fill out the following three district-specific fields
    db_name = 'ESLREPS-OK-TULSA'
    district = 'ok-tulsa'
    source_file_name = 'SMI Table.csv'

    file_name = '{}/{}/spring tests/{}'.format(WORKSPACE_ROOT, district, source_file_name)
    df = pd.read_csv(file_name, dtype=str)
    
    df.rename(columns={
            'Student ID' : 'StudentLocalID',
            'Grade' : 'StudentGradeLevel',
            'Test Date' : 'TestAdministrationDate',
            'Score' : 'Score1Value',
            'Percentile' : 'Score2Value'
        },
             inplace=True)
    
    df.loc[:, 'DistrictDBName'] = db_name
    df.loc[:, 'StudentTestID'] = np.NaN
    df.loc[:, 'TestGradeLevel'] = df.loc[:, 'StudentGradeLevel']
    df.loc[:, 'TestID'] = '0'
    df.loc[:, 'TestName'] = 'SMI'
    df.loc[:, 'TestTypeID'] = '0'
    df.loc[:, 'TestTypeName'] = 'SMI'
    df.loc[:, 'TestSubjectID'] = '0'
    df.loc[:, 'TestSubjectName'] = 'Math'
    df.loc[:, 'Score1Label'] = 'Quantile'
    df.loc[:, 'Score1Type'] = 'Scale'
    df.loc[:, 'Score2Label'] = 'Percentile'
    df.loc[:, 'Score2Type'] = 'Scale'
    df.loc[:, 'SourceFileName'] = source_file_name
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    df.loc[:, headers].to_csv(r"C:\Workspace\ok-tulsa\spring tests\ELLSTDTST_OK-Tulsa_SMI_Spring_2016_2017.csv", index=False)

main()
```
