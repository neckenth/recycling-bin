```python
import pandas as pd
import numpy as np

from pyep import qualify_file_name
from pyep.settings import WORKSPACE_ROOT

def main():
    
    district =  'pa-reading'
    db_name = 'ESLREPS-PA-READING'
    source_file_name = 'PSSA 2017 Preliminary Results.csv'
    
    df = pd.read_csv('{}/{}/pssa/{}'.format(WORKSPACE_ROOT, district, source_file_name))
    
    test_subject_mapping = {
        'E' : 'ELA',
        'M' : 'Math',
        'S' : 'Science'
    }
    
    test_date_mapping = {
        'E' : pd.to_datetime('04/04/2017'),
        'M' : pd.to_datetime('04/25/2017'),
        'S' : pd.to_datetime('05/03/2017')
    }
    
    test_score_mapping = {
        'Bel' : '1 - Below Basic',
        'Bas' : '2 - Basic',
        'Pro' : '3 - Proficient',
        'Adv' : '4 - Advanced'
    }
    
    df.rename(columns={
            'PASecureID' : 'StudentTestID',
            'Grade': 'StudentGradeLevel',
            'Scaled Score' : 'Score1Value',
            'Total Raw Score' : 'Score3Value'
        }, inplace=True)
    
    df.loc[:, 'DistrictDBName'] = db_name
    df.loc[:, 'StudentLocalID'] = np.NaN
    df.loc[:, 'TestAdministrationDate'] = df.loc[:, 'Subject'].map(test_date_mapping)
    df.loc[:, 'TestName'] = 'PSSA'
    df.loc[:, 'TestID'] = '0'
    df.loc[:, 'TestTypeName'] = 'PSSA ' + df.loc[:, 'Subject'].map(test_subject_mapping)
    df.loc[:, 'TestTypeID'] = '0'
    df.loc[:, 'TestSubjectName'] = df.loc[:, 'Subject'].map(test_subject_mapping)
    df.loc[:, 'TestSubjectID'] = '0'
    df.loc[:, 'TestGradeLevel'] = df.loc[:, 'StudentGradeLevel']
    df.loc[:, 'Score1Label'] = 'Scaled Score'
    df.loc[:, 'Score1Type'] = 'Scale'
    df.loc[:, 'Score2Label'] = 'Performance Level'
    df.loc[:, 'Score2Type'] = 'Level'
    df.loc[:, 'Score2Value'] = df.loc[:, 'Performance Level Name'].map(test_score_mapping)
    df.loc[:, 'Score3Label'] = 'Raw Score'
    df.loc[:, 'Score3Type'] = 'Scale'
    df.loc[:, 'SourceFileName'] = source_file_name
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    print len(df)
    
    df = df[df.loc[:, 'Score1Value'].notnull()]
    
    print len(df)
    
    df.loc[:, headers].to_csv('{}/{}/pssa/'.format(WORKSPACE_ROOT, district) + 'ELLSTDTST_PA-Reading_PSSA_2017_2.csv', index=False)
    
main()
    
```
