```python
import pandas as pd
import numpy as np
import glob
import os

from pyep.settings import WORKSPACE_ROOT

def main():
    district='ok-tulsa'
    db_name = 'ESLREPS-OK-TULSA'

    subset = []
    
    necessary_columns = [
        'Student ID (State)',
        'Student ID (Local)',
        'ELA - OPI',
        'ELA - Performance Level',
        'Mathematics - OPI',
        'Mathematics - Performance Level',
        'Science - OPI',
        'Science - Performance Level'
#         'U.S. History - OPI',
#         'U.S. History - Performance Level'
    ]
    
    for f in glob.glob('{}/{}/ostp/2018 OSTP prelim/*.csv'.format(WORKSPACE_ROOT, district)):
        df = pd.read_csv(f, skiprows=3, dtype=str)
        df.loc[:, 'SourceFileName'] = os.path.split(f)[1]
        necessary_columns = [
            'Student ID (State)',
            'Student ID (Local)',
            'Enrolled Grade',
            'SourceFileName']
        for col in df.columns:
            if '- OPI' in col or '- Performance Level' in col:
                necessary_columns.append(col)
        df = df[necessary_columns]
        subset.append(df)

    source = pd.concat(subset)

    perf_level_mapping = {
        'Unsatisfactory': '1 - Unsatisfactory',
        'Limited Knowledge': '2 - Limited Knowledge',
        'Proficient': '3 - Proficient',
        'Advanced': '4 - Advanced'
    }

    test_type_mapping = {
        'ELA':'ELA',
        'Mathematics': 'Math',
        'Science': 'Science',
        'U.S. History' : 'US History'
    }

    test_subj_mapping = {
        'ELA' : 'ELA',
        'Mathematics' : 'Math',
        'Science': 'Science',
        'U.S. History': 'Social Studies'
    }
    
    subject_subset = []

    subjects = ['ELA', 'Mathematics', 'Science'] #, 'U.S. History']
    
    for s in subjects:
        df = source[[
                'Student ID (State)',
                'Student ID (Local)',
                'Enrolled Grade',
                '{} - OPI'.format(s),
                '{} - Performance Level'.format(s),
                'SourceFileName'
            ]]

        df.loc[:, 'StudentTestID'] = source.loc[:, 'Student ID (State)']
        df.loc[:, 'StudentLocalID'] = source.loc[:, 'Student ID (Local)']
        df.loc[:, 'StudentGradeLevel'] = source.loc[:, 'Enrolled Grade'].map(
            lambda x: x.split(' ')[1].lstrip('0'))
        df.loc[:, 'TestGradeLevel'] = df.loc[:, 'StudentGradeLevel']
        df.loc[:, 'TestTypeName'] = 'OSTP ' + (test_type_mapping[s])
        df.loc[:, 'TestSubjectName'] = (test_subj_mapping[s])
        df.loc[:, 'Score1Value'] = source.loc[:, '{} - OPI'.format(s)]
        df.loc[:, 'Score2Value'] = source.loc[:, '{} - Performance Level'.format(s)].map(perf_level_mapping)
        df.drop(labels=['{} - OPI'.format(s),
                        '{} - Performance Level'.format(s)], axis=1, inplace=True)
        
        subject_subset.append(df)
    
    df_final = pd.concat(subject_subset, axis=0)

    df_final.loc[:, 'DistrictDBName'] = db_name
    df_final.loc[:, 'TestAdministrationDate'] = '04/16/2018'
    df_final.loc[:, 'TestName'] = 'OSTP'
    df_final.loc[:, 'TestID'] = '0'
    df_final.loc[:, 'TestTypeID'] = '0'
    df_final.loc[:, 'TestSubjectID'] = '0'
    df_final.loc[:, 'Score1Label'] = 'Scale Score'
    df_final.loc[:, 'Score1Type'] = 'Scale'
    df_final.loc[:, 'Score2Label'] = 'Performance Level'
    df_final.loc[:, 'Score2Type'] = 'Level'
    
    condition_1 = df_final.loc[:, 'StudentGradeLevel'].astype(int) < 5
    condition_2 = df_final.loc[:, 'StudentGradeLevel'].astype(int) < 10
    condition_3 = df_final.loc[:, 'TestSubjectName'] == 'Science'
    condition_4 = df_final.loc[:, 'TestSubjectName'] == 'Social Studies'
    
    df_final = df_final[~(condition_1 & condition_3) & ~(condition_2 & condition_4)]

    headers = ['DistrictDBName',
                    'StudentTestID',
                    'StudentLocalID',
                    'StudentGradeLevel',
                    'TestAdministrationDate',
                    'TestID',
                    'TestName',
                    'TestTypeID',
                    'TestTypeName',
                    'TestSubjectID',
                    'TestSubjectName',
                    'TestGradeLevel',
                    'Score1Label',
                    'Score1Type',
                    'Score1Value',
                    'Score2Label',
                    'Score2Type',
                    'Score2Value',
                    'Score3Label',
                    'Score3Type',
                    'Score3Value',
                    'Score4Label',
                    'Score4Type',
                    'Score4Value',
                    'Score5Label',
                    'Score5Type',
                    'Score5Value',
                    'Score6Label',
                    'Score6Type',
                    'Score6Value',
                    'Score7Label',
                    'Score7Type',
                    'Score7Value',
                    'Score8Label',
                    'Score8Type',
                    'Score8Value',
                    'SourceFileName']
    
    print len(df_final)
    
    df_final = df_final[df_final.loc[:, 'Score1Value'].notnull()] 
   
    print len(df_final)
    df_final.loc[:, headers].to_csv(r"C:\Workspace\ok-tulsa\ostp\temp\ostp_test.csv", index=False)

    
def fix_scores(x):
    try:
        if float(x).is_integer() == True:
            return x
    except ValueError:
        return np.NaN

    
main()
```
