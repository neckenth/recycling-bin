```python
from pyep import read_delimited, qualify_file_name
from pycron import fetch_from_sftp, copy_to_processor_directory
from pyep.transform import derive_from_date_exited_4y

import numpy as np

def main():
    # Edit these variables
    district = 'ma-westborough'
    filename = 'LEP_Students.txt'
    remote_directory = '/data/'
    data_type = 'student'
    file_watcher_type = 'SIS'

    file = fetch_from_sftp(district_name=district,
                           file_name=filename,
                           remote_directory=remote_directory,
                           data_type=data_type,
                           add_datestamp=True)



    df = read_delimited(file, castNull=True, isInteractive=False)

    df = df[df.loc[:, 'U_DEF_EXT_STUDENTS.ellentrydate'].notnull()]

    condition_1 = df['S_MA_STU_SPED_X.sped3to5placement'] == '00'
    condition_2 = df['S_MA_STU_SPED_X.sped6to21placement'] == '00'
    condition_3 = df['S_MA_STU_DEMOGRAPHIC_X.lepprogramstatus'] == '04'

    df.loc[(condition_1 | condition_2), 'Elv_IEP'] = 'No'
    df.loc[(~condition_1 | ~condition_2), 'Elv_IEP'] = 'Yes'

    df['Elv_ExitDate'] = df['U_DEF_EXT_STUDENTS.ellexitdate']

    df[df.loc[condition_3], 'Elv_ExitDate'] = np.NaN

    df['ELV_LEPStatus'] = df['Elv_ExitDate'].map(derive_from_date_exited_4y)

    df.loc[df['Elv_ExitDate'].isnull(), 'ELV_LEPStatus'] = 'ELL'

    df.loc[df['Elv_LEPStatus'] == 'ELL', 'Elv_ReceivingServices'] = 'Yes'

    df.to_csv(qualify_file_name(file, 'processed'), sep='\t', index=False)

    copy_to_processor_directory(district,
                                qualify_file_name(file, 'processed'),
                                file_watcher_type)


if __name__ == '__main__':
    main()
```
