```python
import pandas as pd
import numpy as np

export = pd.read_csv(r"C:\Users\NancyE\Downloads\EllevationExport_20170906_1602.csv", dtype=str)
processed = pd.read_csv(r"C:\Workspace\ma-springfield\student\StudentRoster-20170907-processed_with_school_codes.txt", sep='\t', dtype=str)

result = pd.merge(left=export, right=processed, left_on='Student #', right_on='Local Student #*', how='inner', indicator=True)

result.rename(columns={
        'Active': 'Active - existing',
        'Active Status' : 'Active - updated',
        'School Name' : 'School Name - existing',
        'LEA School Code*' : 'School Name - updated',
        'School LEA Code' : 'School Code - existing',
        'SIS School ID - updated' : 'School Code - updated',
        'Grade Level_x' : 'Grade Level - existing',
        'Grade Level_y' : 'Grade Level - updated',
        'ELP Designation_x' : 'ELP Designation - existing',
        'LEP Status_x' : 'LEP Status - existing',
        'Monitoring Status_x' : 'Monitoring Status - existing',
        'Receiving LEP Svcs' : 'Receiving Services - existing',
        'Elv_LEPStatus' : 'Derived LEP Status - updated'
    }, inplace=True)

df = result[[
        'Last Name',
        'First Name',
        'Test ID #',
        'Student #',
        'State Testing ID #*',
        'Local Student #*',
        'Active - existing',
        'Active - updated',
        'School Name - existing',
        'School Name - updated',
        'School Code - existing',
        'School Code - updated',
        'Grade Level - existing',
        'Grade Level - updated',
        'ELP Designation - existing',
        'LEP Status - existing',
        'Monitoring Status - existing',
        'Receiving Services - existing',
        'Derived LEP Status - updated'
    ]]

df.loc[:, 'Elv_SchoolChangeFlag'] = df.apply(
    lambda x: 'Y' if ((x.loc['School Code - updated'] != x.loc['School Code - existing']) and x.loc['School Name - updated'] != 'ARCHIVE') else np.NaN, 
    axis=1)

lep_condition_1 = df.loc[:, 'LEP Status - existing'] == 'LEP'
lep_condition_2 = df.loc[:, 'Derived LEP Status - updated'] == 'Current ELL'
opt_condition_1 = df.loc[:, 'LEP Status - existing'] == 'ELL - Opt Out'
opt_condition_2 = df.loc[:, 'Derived LEP Status - updated'] == 'Opt Out'
# flep_condition_1 = df.loc[:, 'LEP Status - existing'] == 'FLEP'
# flep_condition_2 = df.loc[:, 'Monitoring Status - existing'].str[-1] == df.loc[:, 'Derived LEP Status - updated'].str[-1]

df.loc[(lep_condition_1 & ~lep_condition_2) | (opt_condition_1 & ~opt_condition_2), 'Elv_StatusChangeFlag']  = 'Y'
print len(df)

# blank Active column and blank school in most recent file
df = df[df.loc[:, 'Active - updated'].notnull() & df.loc[:, 'School Name - updated'].notnull()]
print len(df)

# new students who have no existing records in ELV
df = df[df.loc[:, 'First Name'].notnull() & df.loc[:, 'Last Name'].notnull()]
print len(df)

# archived students
df = df[df.loc[:, 'School Name - updated'] != 'ARCHIVE']
print len(df)

df.to_csv(r"C:\Workspace\ma-springfield\student\StudentRoster-20170907-processed_with_flags_9.csv", sep=',', index=False)
```
