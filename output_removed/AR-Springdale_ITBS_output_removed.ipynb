```python
import glob
import os
import pandas as pd
import numpy as np

from pyep import qualify_file_name
from pyep.settings import WORKSPACE_ROOT

def main():
    district = 'ar-springdale'
    db_name = 'ESLREPS-AR-SPRINGDALE'

    subset = []

    for file in glob.glob('{}/{}/itbs/ITBS*.csv'.format(WORKSPACE_ROOT, district)):
        source = pd.read_csv(file, dtype=str)
        if 'Grade01' in file:
            source.loc[:, 'StudentGradeLevel'] = '1'
            source.loc[:, 'TestGradeLevel'] = '1'
        else:
            source.loc[:, 'StudentGradeLevel'] = '2'
            source.loc[:, 'TestGradeLevel'] = '2'
        source.loc[:, 'SourceFileName'] = os.path.split(file)[1]
        subset.append(source)

    source = pd.concat(subset)
    
    source.rename(columns={'stateid':'StudentTestID', 'localid':'StudentLocalID'}, inplace=True)

    test_type_mapping = {'Math Total': 'Math',
                        'Reading': 'Reading',
                        'Language': 'Language'}

    domains = ['Reading', 'Language', 'Math Total']

    subset_1 = []

    for domain in domains:
        df = source[['StudentTestID',
                     'StudentLocalID',
                     'StudentGradeLevel',
                     'TestGradeLevel',
                     '{} SS'.format(domain),
                    '{} NPR'.format(domain),
                    'SourceFileName']]
        df.loc[:, 'TestTypeName'] = 'ITBS ' + (test_type_mapping[domain])
        df.loc[:, 'TestSubjectName'] = test_type_mapping[domain]
        df.loc[:, 'Score1Value'] = source.loc[:, '{} SS'.format(domain)]
        df.loc[:, 'Score2Value'] = source.loc[:, '{} NPR'.format(domain)]
        subset_1.append(df)

    df = pd.concat(subset_1)

    df.loc[:, 'DistrictDBName'] = db_name
    df.loc[:, 'TestAdministrationDate'] = pd.to_datetime('04/01/2017')
    df.loc[:, 'TestName'] = 'ITBS'
    df.loc[:, 'TestID'] = '0'
    df.loc[:, 'TestTypeID'] = '0'
    df.loc[:, 'TestSubjectID'] = '0'
    df.loc[:, 'Score1Label'] = 'Scale Score'
    df.loc[:, 'Score1Type'] = 'Scale'
    df.loc[:, 'Score2Label'] = 'NPR'
    df.loc[:, 'Score2Type'] = 'Scale'

    headers = ['DistrictDBName',
                'StudentTestID',
                'StudentLocalID',
                'StudentGradeLevel',
                'TestAdministrationDate',
                'TestID',
                'TestName',
                'TestTypeID',
                'TestTypeName',
                'TestSubjectID',
                'TestSubjectName',
                'TestGradeLevel',
                'Score1Label',
                'Score1Type',
                'Score1Value',
                'Score2Label',
                'Score2Type',
                'Score2Value',
                'Score3Label',
                'Score3Type',
                'Score3Value',
                'Score4Label',
                'Score4Type',
                'Score4Value',
                'Score5Label',
                'Score5Type',
                'Score5Value',
                'Score6Label',
                'Score6Type',
                'Score6Value',
                'Score7Label',
                'Score7Type',
                'Score7Value',
                'Score8Label',
                'Score8Type',
                'Score8Value',
                'SourceFileName']
    
    df = df[df.loc[:, 'Score1Value'] != 'Not Tested']

    df.loc[:, headers].to_csv('{}/{}/itbs/final/{}'.format(WORKSPACE_ROOT, district, 'ELLSTDTST_AR-Springdale_ITBS_Grade1-2_2017.csv'), index=False)
    
main()
```
