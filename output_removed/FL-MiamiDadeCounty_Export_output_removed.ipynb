```python
# services - wide

import pandas as pd
import numpy as np
import glob
import os

from pyep.settings import WORKSPACE_ROOT

district = 'fl-miamidadecounty'

subset = []

for f in glob.glob('{}/{}/export/services/*.csv'.format(WORKSPACE_ROOT, district)):
    df_file = pd.read_csv(f, dtype=str)
    df_file.loc[:, 'Export_MeetingType'] = os.path.split(f)[1]
    subset.append(df_file)
    
df = pd.concat(subset)

df.drop_duplicates(inplace=True)

df.loc[:, 'Elv_Counter'] = df.groupby('Student #').cumcount(ascending=True)
df.loc[:, 'Elv_Counter'] = df.loc[:, 'Elv_Counter'].map(str)

df_pivoted = pd.pivot_table(data=df,
                                    index=['Student #', 'Services Year/Semester'],
                                    columns =['Elv_Counter'],
                                    values=['Service', 'Service Provider', 'Service Schedule', 'Service Minutes per Week'],
                                    aggfunc=np.max).reset_index(col_level=1)

df_pivoted.columns = [' '.join(col).strip() for col in df_pivoted.columns.tolist()]

df_pivoted.fillna(value=np.NaN, inplace=True)

df_columns_final = []

for col in df_pivoted.columns.tolist():
    if ('0' in col or '1' in col or '2' in col or 
        '3' in col or '4' in col or '5' in col or '6'
        in col or '7' in col or '8' in col or '9' in col or '10' in col) and 'Service' in col:
        col = col.replace(col.split(' ')[-1], str(int(col.split(' ')[-1]) + 1))
        df_columns_final.append(col)
    else:
        df_columns_final.append(col)
        
df_pivoted.columns = df_columns_final

df_pivoted = df_pivoted.replace(['/ for ', '------'], np.NaN)

df_pivoted.dropna(axis=1, how='all', inplace=True)

df_pivoted.drop_duplicates(inplace=True)

df_pivoted.to_csv(r"C:\Workspace\fl-miamidadecounty\export\services_combined_wide.csv", index=False)
```
