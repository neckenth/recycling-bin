```python
import pandas as pd
import numpy as np
from pyep import read_delimited, qualify_file_name


df = read_delimited(r"C:\Workspace\fl-miamidadecounty\student\ELLEVATION_DEMOGRAPHIC-20171103.csv",
                    sep=',',
                    castNull=True,
                    isInteractive=False)

print df.columns.tolist()

df.drop_duplicates(subset=['POST_PROGRAM_REVIEW_DATE1', 'POST_PROGRAM_REVIEW_DATE2', 'POST_PROGRAM_REVIEW_DATE3', 'POST_PROGRAM_REVIEW_DATE4'], inplace=True)

guardians = ['G1', 'G2', 'G3']

for p in guardians:
    df.loc[:, 'Elv_FatherName_{}'.format(p)] = df.apply(
        lambda x: x.loc['{}_Name'.format(p)] if x.loc['{}_Relationship'.format(p)] == 'F' else np.NaN, axis=1)
    df.loc[:, 'Elv_MotherName_{}'.format(p)] = df.apply(
        lambda x: x.loc['{}_Name'.format(p)] if x.loc['{}_Relationship'.format(p)] == 'M' else np.NaN, axis=1)
    df.loc[:, 'Elv_GuardianName_{}'.format(p)] = df.apply(
        lambda x: x.loc['{}_Name'.format(p)] if x.loc['{}_Relationship'.format(p)] == 'G' else np.NaN, axis=1)

parents = ['Father', 'Mother', 'Guardian']

for r in parents:
    df.loc[:, 'Elv_{}Name'.format(r)] = df[['Elv_{}Name_G1'.format(r),
                                            'Elv_{}Name_G2'.format(r),
                                            'Elv_{}Name_G3'.format(r)]].fillna('').sum(axis=1)

df.drop(labels=['Elv_FatherName_G1',
                'Elv_FatherName_G2',
                'Elv_FatherName_G3',
                'Elv_MotherName_G1',
                'Elv_MotherName_G2',
                'Elv_MotherName_G3',
                'Elv_GuardianName_G1',
                'Elv_GuardianName_G2',
                'Elv_GuardianName_G3'], axis=1, inplace=True)

df.to_csv(qualify_file_name(r"C:\Workspace\fl-miamidadecounty\student\ELLEVATION_DEMOGRAPHIC-20171103.csv", 'processed_for_mapping'), index=False)
```
