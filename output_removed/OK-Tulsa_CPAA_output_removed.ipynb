```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT

def main():
    
    def derive_test_name(row):
        if 'Math' in row.loc['assessmentName'] and 'SPAN' not in row.loc['assessmentName']:
            return 'CPAA Math'
        elif 'Lit' in row.loc['assessmentName'] and 'SPAN' not in row.loc['assessmentName']:
            return 'CPAA Early Literacy'
        elif 'Math' in row.loc['assessmentName'] and 'SPAN' in row.loc['assessmentName']:
            return 'CPAA-SPAN Math'
        elif 'Lit' in row.loc['assessmentName'] and 'SPAN' in row.loc['assessmentName']:
            return 'CPAA-SPAN Early Literacy'
        
    
    def pull_scores(row):
        if row.loc['TestTypeName'] == 'CPAA Math' or row.loc['TestTypeName'] == 'CPAA-SPAN Math':
            return str(row.loc['math_percentile']).rstrip('.0')
        elif row.loc['TestTypeName'] == 'CPAA Early Literacy' or row.loc['TestTypeName'] == 'CPAA-SPAN Early Literacy':
            return str(row.loc['ela_percentile']).rstrip('.0')
        
        
    def derive_test_lang(row):
        if 'SPAN' in row.loc['assessmentName']:
            return 'CPAA-SPAN '
        elif 'SPAN' not in row.loc['assessmentName']:
            return 'CPAA '

        
    test_subject_mapping = {
        'CPAA Math' : 'Math',
        'CPAA Early Literacy' : 'ELA',
        'CPAA-SPAN Math' : 'Math',
        'CPAA-SPAN Early Literacy' : 'ELA'
    }
    
    domain_test_type = {
        'ela_percentile' : 'Early Literacy',
        'math_percentile' : 'Math'
    }
        
    db_name = 'ESLREPS-OK-TULSA'
    district = 'ok-tulsa'
    source_file_name = 'Tulsa Public Schools_9206_FALL 2017-2018_ConceptNewWithPerformanceLevels_20171017.csv'
    
    file_name = '{}/{}/cpaa/{}'.format(WORKSPACE_ROOT, district, source_file_name)
    df = pd.read_csv(file_name, dtype=str)
    
    generals = ['Gr 1 SPRING CPAA', 
                'K SPRING CPAA',
                'Gr 2 SPRING CPAA',
                'K SPRING CPAA-SPAN',
                'Gr 1 SPRING CPAA-SPAN',
                'Gr 2 SPRING CPAA-SPAN']
    
    domains = ['ela', 'math']
    
    subset = []
        
    df_domain = df[df.loc[:, 'assessmentName'].isin(generals)]
    
    for domain in domains:
        df_domain = df[['studentId',
                    'grade',
                    'assessmentName',
                    'dateAssigned', 
                    '{}_percentile'.format(domain)]]

        df_domain.rename(columns={
                'studentId' : 'StudentLocalID',
                'dateAssigned' : 'TestAdministrationDate',
                'grade' : 'StudentGradeLevel'
            }, inplace=True)

        df_domain.loc[:, 'DistrictDBName'] = db_name
        df_domain.loc[:, 'StudentTestID'] = np.NaN
        df_domain.loc[:, 'TestGradeLevel'] = df_domain.loc[:, 'StudentGradeLevel']
        df_domain.loc[:, 'TestName'] = 'CPAA'
        df_domain.loc[:, 'TestID'] = '0'
        df_domain.loc[:, 'TestTypeName'] = df_domain.apply(derive_test_lang, axis=1) + domain_test_type['{}_percentile'.format(domain)]
        df_domain.loc[:, 'TestTypeID'] = '0'
        df_domain.loc[:, 'TestSubjectName'] = df_domain.loc[:, 'TestTypeName'].map(test_subject_mapping)
        df_domain.loc[:, 'TestSubjectID'] = '0'
        df_domain.loc[:, 'Score1Type'] = 'Scale'
        df_domain.loc[:, 'Score1Label'] = 'Percentile'
        df_domain.loc[:, 'Score1Value'] = df_domain.loc[:, '{}_percentile'.format(domain)]
        df_domain.loc[:, 'SourceFileName'] = source_file_name
        
        subset.append(df_domain)
    
    df = df[~df.loc[:, 'assessmentName'].isin(generals)]
    
    df.rename(columns={
            'studentId' : 'StudentLocalID',
            'dateAssigned' : 'TestAdministrationDate',
            'grade' : 'StudentGradeLevel',
        }, inplace=True)
    df.loc[:, 'DistrictDBName'] = db_name
    df.loc[:, 'StudentTestID'] = np.NaN
    df.loc[:, 'TestGradeLevel'] = df.loc[:, 'StudentGradeLevel']
    df.loc[:, 'TestName'] = 'CPAA'
    df.loc[:, 'TestID'] = '0'
    df.loc[:, 'TestTypeName'] = df.apply(derive_test_name, axis=1)
    df.loc[:, 'TestTypeID'] = '0'
    df.loc[:, 'TestSubjectName'] = df.loc[:, 'TestTypeName'].map(test_subject_mapping)
    df.loc[:, 'TestSubjectID'] = '0'
    df.loc[:, 'Score1Type'] = 'Scale'
    df.loc[:, 'Score1Label'] = 'Percentile'
    df.loc[:, 'Score1Value'] = df.apply(pull_scores, axis=1)
    df.loc[:, 'SourceFileName'] = source_file_name
    
    subset.append(df)
    
    final = pd.concat(subset)  

    headers = ['DistrictDBName',
        'StudentTestID',
        'StudentLocalID',
        'StudentGradeLevel',
        'TestAdministrationDate',
        'TestID',
        'TestName',
        'TestTypeID',
        'TestTypeName',
        'TestSubjectID',
        'TestSubjectName',
        'TestGradeLevel',
        'Score1Label',
        'Score1Type',
        'Score1Value',
        'Score2Label',
        'Score2Type',
        'Score2Value',
        'Score3Label',
        'Score3Type',
        'Score3Value',
        'Score4Label',
        'Score4Type',
        'Score4Value',
        'Score5Label',
        'Score5Type',
        'Score5Value',
        'Score6Label',
        'Score6Type',
        'Score6Value',
        'Score7Label',
        'Score7Type',
        'Score7Value',
        'Score8Label',
        'Score8Type',
        'Score8Value',
        'SourceFileName']
    
    final = final[final.loc[:, 'Score1Value'].notnull()]
    
    final.loc[:, headers].to_csv(r"C:\Workspace\ok-tulsa\cpaa\spring\ELLSTDTST_OK-Tulsa_CPAA_Spring_16_17.csv", index=False)
             
main()
```
