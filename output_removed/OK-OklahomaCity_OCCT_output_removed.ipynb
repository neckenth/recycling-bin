```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT

def main():
    db_name = 'ESLREPS-OK-OKLAHOMACITY'
    district = 'ok-oklahomacity'
    source_file_name = 'CRT Reading Scores 2016 (1).xlsx'
    
    file_name = '{}/{}/standardized tests/{}'.format(WORKSPACE_ROOT, district, source_file_name)
    source = pd.read_excel(file_name)
    
    performance_level_map = {
        'Unsatisfactory' : '1 - Unsatisfactory',
        'Limited Knowledge' : '2 - Limited Knowledge',
        'Proficient' : '3 - Proficient',
        'Advanced' : '4 - Advanced'
    }
    
    source.loc[:, 'DistrictDBName'] = db_name
    source.loc[:, 'StudentTestID'] = np.NaN
    source.loc[:, 'StudentLocalID'] = source.loc[:, 'stuid']
    source.loc[:, 'StudentGradeLevel'] = source.loc[:, 'grade16']
    source.loc[:, 'TestAdministrationDate'] = pd.to_datetime(source.loc[:, 'testdate'], format='%Y%m%d')
    source.loc[:, 'TestID'] = '0'
    source.loc[:, 'TestName'] = 'OCCT'
    source.loc[:, 'TestTypeID'] = '0'
    source.loc[:, 'TestTypeName'] = 'OCCT Reading'
    source.loc[:, 'TestSubjectID'] = '0'
    source.loc[:, 'TestSubjectName'] = 'Reading'
    source.loc[:, 'TestGradeLevel'] = source.loc[:, 'grade16']
    source.loc[:, 'Score1Label'] = 'Score'
    source.loc[:, 'Score1Type'] = 'Scale'
    source.loc[:, 'Score1Value'] = source.loc[:, 'OPI16']
    source.loc[:, 'Score2Label'] = 'Performance Level'
    source.loc[:, 'Score2Type'] = 'Level'
    source.loc[:, 'Score2Value'] = source.loc[:, 'PerformanceLevel16'].map(performance_level_map)
    source.loc[:, 'SourceFileName'] = source_file_name
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    source.loc[:, headers].to_csv(r"C:\Workspace\ok-oklahomacity\standardized tests\ELLSTDTST_OK-OklahamaCity_OCCT_2016.csv", index=False)

main()
```
