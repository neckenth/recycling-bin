```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT

def main():
    
    district = 'tx-garlandisd'
    district_db_name = 'ESLREPS-TX-GARLANDISD'
    source_file_name = '2017-18 SAT School Day Grade 11 (March 7).xlsx'
    grades_file_name = 'EllevationExport_20180508_1120.csv'
    
    scores = pd.read_excel('{}/{}/sat/{}'.format(WORKSPACE_ROOT, district, source_file_name), converters={'SECONDARY_ID':str,
                                                                                                         'LATEST_SAT_EBRW':str,
                                                                                                          'LATEST_SAT_MATH_TEST':str,
                                                                                                            'LATEST_SAT_READING':str,
                                                                                                            'LATEST_SAT_WRIT_LANG':str})
    grades = pd.read_csv('{}/{}/sat/{}'.format(WORKSPACE_ROOT, district, grades_file_name), dtype=str)
    
    scores = scores[scores.loc[:, 'SECONDARY_ID'].notnull()]
    scores = scores[scores.loc[:, 'LATEST_ASSESSMENT_DATE'].notnull()]
    grades = grades[grades.loc[:, 'Grade Level'].notnull()]

    source = pd.merge(left=scores, right=grades, left_on='SECONDARY_ID', right_on='Student #', how='inner', indicator=True)

    source.rename(columns={
            'LATEST_SAT_EBRW': 'LATEST_SAT_EBRW_SECTION',
        }, inplace=True)

    subjs = ['EBRW', 'MATH']

    subset = []

    for s in subjs:
        domain = source[[
                'SECONDARY_ID',
                'Grade Level', 
                'LATEST_ASSESSMENT_DATE',
                'LATEST_SAT_{}_SECTION'.format(s),
                'LATEST_SAT_MATH_TEST',
                'LATEST_SAT_READING', 
                'LATEST_SAT_WRIT_LANG'
            ]]

        domain.rename(columns={
            'SECONDARY_ID': 'StudentLocalID',
            'Grade Level': 'StudentGradeLevel',
            'LATEST_ASSESSMENT_DATE': 'TestAdministrationDate',
            'LATEST_SAT_{}_SECTION'.format(s): 'Score1Value',
        }, inplace=True)
        
        domain.loc[:, 'Score1Label'] = 'Section Score'
        domain.loc[:, 'Score1Type'] = 'Scale'

        if s == 'EBRW':
            domain.rename(columns={
                    'LATEST_SAT_READING': 'Score2Value',
                    'LATEST_SAT_WRIT_LANG': 'Score3Value'
                }, inplace=True)
            domain.loc[:, 'Score2Label'] = 'Reading Test Score'
            domain.loc[:, 'Score2Type'] = 'Scale'
            domain.loc[:, 'Score3Label'] = 'Writing Test Score'
            domain.loc[:, 'Score3Type'] = 'Scale'

            domain.drop(labels=['LATEST_SAT_MATH_TEST'], axis=1, inplace=True)
            
        elif s == 'MATH':
            domain.rename(columns={
                    'LATEST_SAT_MATH_TEST': 'Score2Value'
                }, inplace=True)
            domain.loc[:, 'Score2Label'] = 'Math Test Score'
            domain.loc[:, 'Score2Type'] = 'Scale'

            domain.drop(labels=['LATEST_SAT_READING',
                               'LATEST_SAT_WRIT_LANG'], axis=1, inplace=True)
            

        domain.loc[:, 'TestTypeName'] = 'SAT {}'.format('{}'.format(s))
        domain.loc[:, 'TestSubjectName'] = '{}'.format(s)

        subset.append(domain)

    df = pd.concat(subset)

    df.loc[:, 'DistrictDBName'] = district_db_name
    df.loc[:, 'TestName'] = 'SAT'
    df.loc[:, 'TestID'] = '0'
    df.loc[:, 'TestTypeName'] = df.loc[:, 'TestTypeName'].map(
        lambda x: x if x != 'SAT MATH' else 'SAT Mathematics')
    df.loc[:, 'TestSubjectName'] = df.loc[:, 'TestSubjectName'].map(
        lambda x: x if x != 'MATH' else 'Math')
    df.loc[:, 'TestTypeID'] = '0'
    df.loc[:, 'TestSubjectID'] = '0'
    df.loc[:, 'TestGradeLevel'] = df.loc[:, 'StudentGradeLevel']
    df.loc[:, 'SourceFileName'] = source_file_name
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    df.loc[:, headers].to_csv('{}/{}/sat/ELLSTDTST_TX-GARLANDISD_SAT_2018.csv'.format(WORKSPACE_ROOT, district), sep=',', index=False)

main()

```
