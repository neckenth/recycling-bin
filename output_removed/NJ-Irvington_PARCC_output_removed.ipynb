```python
import pandas as pd
import numpy as np
import datetime as dt

from pyep.settings import WORKSPACE_ROOT
from pyep import qualify_file_name
from common import is_null

def main():

    district = 'nj-irvington'

    source_file_name = 'PC_pcspr17_NJ-132330_PARCC_District_Summative_Record_File_Spring (2).csv'

    necessary_columns = [
        'StateStudentIdentifier',
        'LocalStudentIdentifier',
        'Unit1OnlineTestStartDateTime',
        'GradeLevelWhenAssessed',
        'Subject',
        'TestScaleScore',
        'TestPerformanceLevel',
        'TestReadingScaleScore',
        'TestWritingScaleScore'
    ]

#     df = pd.read_excel('{}/{}/parcc/{}'.format(WORKSPACE_ROOT, district, source_file_name), usecols=necessary_columns, converters={'TestPerformanceLevel':str,
#                                  'StateStudentIdentifier':str,
#                                  'LocalStudentIdentifier':str,
#                                  'GradeLevelWhenAssessed':str,                                                                                                  
#                                  'Unit1OnlineTestStartDateTime':str,
#                                  'TestScaleScore':str,
#                                  'TestReadingScaleScore':str,
#                                  'TestWritingScaleScore':str})
    
    df = pd.read_csv('{}/{}/parcc/{}'.format(WORKSPACE_ROOT, district, source_file_name), usecols=necessary_columns, dtype=str)

    district_db_name = 'ESLREPS-NJ-IRVINGTON'

    perf_level_mapping = {
        '1': 'Did Not Meet',
        '2': 'Partially Met',
        '3': 'Approached',
        '4': 'Met',
        '5': 'Exceeded'
    }

    df.rename(columns={
            'StateStudentIdentifier':'StudentTestID',
            'LocalStudentIdentifier':'StudentLocalID',
            'Unit1OnlineTestStartDateTime':'TestAdministrationDate',
            'GradeLevelWhenAssessed': 'StudentGradeLevel',
            'TestScaleScore':'Score1Value',
            'TestReadingScaleScore':'Score3Value',
            'TestWritingScaleScore':'Score4Value'
        }, inplace=True)

    df.loc[:, 'DistrictDBName'] = district_db_name
    df.loc[:, 'TestAdministrationDate'] = df.loc[:, 'TestAdministrationDate'].map(
        lambda x: pd.to_datetime(x)).dt.date
    df.loc[:, 'TestName'] = 'PARCC'
    df.loc[:, 'TestID'] = '0'
    df.loc[:, 'TestTypeName'] = df.loc[:, 'Subject'].map(
        lambda x: 'PARCC {}'.format(x) if x != 'English Language Arts/Literacy' else 'PARCC ELA/Literacy')
    df.loc[:, 'TestTypeID'] = '0'
    df.loc[:, 'TestSubjectName'] = df.loc[:, 'Subject'].map(
        lambda x: 'Math' if x != 'English Language Arts/Literacy' else 'ELA')
    df.loc[:, 'TestSubjectID'] = '0'
    df.loc[:, 'TestGradeLevel'] = df.loc[:, 'StudentGradeLevel']
    df.loc[:, 'Score1Label'] = 'Scale Score'
    df.loc[:, 'Score1Type'] = 'Scale'
    df.loc[:, 'Score2Label'] = 'Performance Level'
    df.loc[:, 'Score2Type'] = 'Level'
    df.loc[:, 'Score2Value'] = df.loc[:, 'TestPerformanceLevel'].map(
        lambda x: '{} - {}'.format(x, perf_level_mapping[x]) if not is_null(x) else np.NaN)
    df.loc[:, 'Score3Label'] = 'Reading Score'
    df.loc[:, 'Score3Type'] = 'Scale'
    df.loc[:, 'Score4Label'] = 'Writing Score'
    df.loc[:, 'Score4Type'] = 'Scale'
    df.loc[:, 'SourceFileName'] = source_file_name

    headers = ['DistrictDBName',
                'StudentTestID',
                'StudentLocalID',
                'StudentGradeLevel',
                'TestAdministrationDate',
                'TestID',
                'TestName',
                'TestTypeID',
                'TestTypeName',
                'TestSubjectID',
                'TestSubjectName',
                'TestGradeLevel',
                'Score1Label',
                'Score1Type',
                'Score1Value',
                'Score2Label',
                'Score2Type',
                'Score2Value',
                'Score3Label',
                'Score3Type',
                'Score3Value',
                'Score4Label',
                'Score4Type',
                'Score4Value',
                'Score5Label',
                'Score5Type',
                'Score5Value',
                'Score6Label',
                'Score6Type',
                'Score6Value',
                'Score7Label',
                'Score7Type',
                'Score7Value',
                'Score8Label',
                'Score8Type',
                'Score8Value',
                'SourceFileName']
    
    df = df[df.loc[:, 'Score1Value'].notnull()]
    df = df[df.loc[:, 'TestAdministrationDate'].notnull()]

    df.loc[:, headers].to_csv('{}/{}/parcc/ELLSTDTST_NJ-Irvington_PARCC_Spring_16_17.csv'.format(WORKSPACE_ROOT, district), index=False)

main()
```
