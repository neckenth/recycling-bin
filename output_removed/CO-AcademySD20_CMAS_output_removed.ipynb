```python
import pandas as pd
import numpy as np
import os

from pyep.settings import WORKSPACE_ROOT
from common import is_null

def main():
    
    db_name = 'ESLREPS-CO-ACADEMYSD20'
    district = 'co-academysd20'
    
    # first convert files to comma-separated to avoid issues with unicode
    
#     for file in os.listdir(WORKSPACE_ROOT + '/{}/standardized-tests/CMAS'.format(district)):
#         each_file = pd.read_excel(WORKSPACE_ROOT + '/{}/standardized-tests/CMAS/'.format(district) + file)
#         each_file.to_csv(WORKSPACE_ROOT + '/{}/standardized-tests/CMAS/'.format(district) + file)
    
    
    
    test_type_mapping = {
        'SS ' : 'CMAS Social Studies',
        'Social Studies' : 'CMAS Social Studies',
        'SCI' : 'CMAS Science',
        'Science' : 'CMAS Science'
    }
    
    test_subject_mapping = {
        'SS ' : 'History and Social Science',
        'Social Studies' : 'History and Social Science',
        'SCI' : 'Science',
        'Science' : 'Science'
    }
    
    pl_mapping = {
        '1' : '1 - Limited Command',
        '2' : '2 - Moderate Command',
        '3' : '3 - Strong Command',
        '4' : '4 - Distinguished Command'
    }
    
    subset_2015 = []
    
    for file in os.listdir(WORKSPACE_ROOT + '/{}/standardized-tests/CMAS/2015'.format(district)):
        df_2015 = pd.read_csv(WORKSPACE_ROOT + '/{}/standardized-tests/CMAS/2015/'.format(district) + file, sep=',', dtype=str)
        df_2015.loc[:, 'SourceFileName'] = file
        
        subset_2015.append(df_2015)
        
    df_2015 = pd.concat(subset_2015)
    
    df_2015 = df_2015[['SASID',
                'LASID',
                'Grade',
                'Test Date',
                'Content',
                'SS',
                'Perf LVL',
                'Std1 Perf Lvl',
                'Std2 Perf Lvl',
                'Std3 Perf Lvl',
                'Std4 Perf Lvl',
                'SourceFileName']]
        
    subset_2016 = []
        
    for file in os.listdir(WORKSPACE_ROOT + '/{}/standardized-tests/CMAS/2016'.format(district)):
        df_2016 = pd.read_csv(WORKSPACE_ROOT + '/{}/standardized-tests/CMAS/2016/'.format(district) + file, sep=',', dtype=str)
        df_2016.loc[:, 'SourceFileName'] = file
        
        subset_2016.append(df_2016)
    
    df_2016 = pd.concat(subset_2016)
    
    df_2016 = df_2016[['StateStudentIdentifier',
                   'LocalStudentIdentifier',
                    'GradeLevelWhenAssessed',
                       'AttemptCreateDate',
                   'Subject',
                   'SS',
                    'Perf LVL',
                    'Std1 Perf Lvl',
                    'Std2 Perf Lvl',
                    'Std3 Perf Lvl',
                    'Std4 Perf Lvl',
                    'SourceFileName']]
        
    source = pd.concat([df_2015, df_2016])
    
    source.loc[:, 'DistrictDBName'] = db_name
    source.loc[:, 'StudentTestID'] = source.apply(
        lambda x: x.loc['SASID'] if not is_null(x.loc['SASID']) else x.loc['StateStudentIdentifier'], axis=1)
    source.loc[:, 'StudentLocalID'] = source.apply(
        lambda x: x.loc['LASID'] if not is_null(x.loc['LASID']) else x.loc['LocalStudentIdentifier'], axis=1)
    source.loc[:, 'StudentGradeLevel'] = source.apply(
        lambda x: x.loc['Grade'].rstrip('0') if not is_null(x.loc['Grade']) else x.loc['GradeLevelWhenAssessed'].rstrip('0'), axis=1)
    source.loc[:, 'TestGradeLevel'] = source.loc[:, 'StudentGradeLevel']
    source.loc[:, 'TestAdministrationDate'] = source.apply(
        lambda x: pd.to_datetime(x.loc['Test Date'], format='%m%d%Y', errors='coerce') if not is_null(x.loc['Test Date']) else x.loc['AttemptCreateDate'], axis=1)
    source.loc[:, 'TestID'] = '0'
    source.loc[:, 'TestName'] = 'CMAS'
    source.loc[:, 'TestTypeID'] = '0'
    source.loc[:, 'TestTypeName'] = source.apply(
        lambda x: x.loc['Subject'] if not is_null(x.loc['Subject']) else x.loc['Content'], axis=1) 
    source.loc[:, 'TestTypeName'] = source.loc[:, 'TestTypeName'].map(test_type_mapping)
    source.loc[:, 'TestSubjectID'] = '0'
    source.loc[:, 'TestSubjectName'] = source.apply(
        lambda x: x.loc['Subject'] if not is_null(x.loc['Subject']) else x.loc['Content'], axis=1)
    source.loc[:, 'TestSubjectName'] = source.loc[:, 'TestSubjectName'].map(test_subject_mapping)
    source.loc[:, 'Score1Label'] = 'Overall Scale Score'
    source.loc[:, 'Score1Type'] = 'Scale'
    source.loc[:, 'Score1Value'] = source.loc[:, 'SS']
    source.loc[:, 'Score2Label'] = 'Overall Performance Level'
    source.loc[:, 'Score2Type'] = 'Level'
    source.loc[:, 'Score2Value'] = source.loc[:, 'Perf LVL'].map(pl_mapping)
    source.loc[:, 'Score3Label'] = 'Standard 1 Performance Level'
    source.loc[:, 'Score3Type'] = 'Level'
    source.loc[:, 'Score3Value'] = source.loc[:, 'Std1 Perf Lvl'].map(pl_mapping)
    source.loc[:, 'Score4Label'] = 'Standard 2 Performance Level'
    source.loc[:, 'Score4Type'] = 'Level'
    source.loc[:, 'Score4Value'] = source.loc[:, 'Std2 Perf Lvl'].map(pl_mapping)
    source.loc[:, 'Score5Label'] = 'Standard 3 Performance Level'
    source.loc[:, 'Score5Type'] = 'Level'
    source.loc[:, 'Score5Value'] = source.loc[:, 'Std3 Perf Lvl'].map(pl_mapping)
    source.loc[:, 'Score6Label'] = 'Standard 4 Performance Level'
    source.loc[:, 'Score6Type'] = 'Level'
    source.loc[:, 'Score6Value'] = source.loc[:, 'Std4 Perf Lvl'].map(pl_mapping)
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    source = source[source.loc[:, 'TestAdministrationDate'].notnull()]
    source = source[source.loc[:, 'Score1Value'].notnull()]
    source = source[source.loc[:, 'Score1Value'] != '   ']
    
    source.loc[:, headers].to_csv(r"C:\Workspace\co-academysd20\standardized-tests\CMAS\ELLSTDTST_CO-AcademySD20_CMAS_2015_2016.csv", index=False)

    
        
    
    
    
main()
```
