```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT

from common import is_null

def main():
    db_name = 'ESLREPS-MD-FREDERICKCOUNTY'
    district = 'md-frederickcounty'
    source_file_name = 'ElementaryPARCCScores20170419.csv'
    
    file_name = '{}/{}/parcc/{}'.format(WORKSPACE_ROOT, district, source_file_name)
    source = pd.read_csv(file_name)
    
    domains = ['ELA', 'MATH', 'ELA10', 'ELA11', 'ALG01', 'ALG02', 'GEO01']
    
    #print source.columns.tolist()
    
    test_type_mapping = {
        'ELA': 'ELA/Literacy',
        'ELA10': 'ELA/Literacy',
        'ELA11' : 'ELA/Literacy',
        'MATH' : 'Mathematics',
        'ALG01' : 'Algebra I',
        'ALG02' : 'Algebra II',
        'GEO01' : 'Geometry'
    }
    
    test_subject_mapping = {
        'ELA' : 'ELA',
        'ELA10' : 'ELA', 
        'ELA11' : 'ELA',
        'MATH' : 'Math',
        'ALG01' : 'Math',
        'ALG02' : 'Math', 
        'GEO01' : 'Math'
    }
    
    performance_level_mapping = {
        1. : '1 - Did not meet expectations',
        2. : '2 - Partially met expectations',
        3. : '3 - Approached expectations',
        4. : '4 - Met expectations',
        5. : '5 - Exceeded expectations',
    }
    
    subset = []
    
    for domain in domains:
        df = source[['STUDENT_ID',
                     'GRADE',
                    'TEST_DATE',
                    '{}_SCALE'.format(domain),
                     '{}_LEVEL'.format(domain),
                     '{}_LEVEL_DES'.format(domain)
                    ]]
        
        df.loc[:, 'StudentLocalID'] = source.loc[:, 'STUDENT_ID']
        df.loc[:, 'StudentGradeLevel'] = source.loc[:, 'GRADE']
        df.loc[:, 'TestGradeLevel'] = source.loc[:, 'GRADE']
        df.loc[:, 'TestAdministrationDate'] = source.loc[:, 'TEST_DATE']
        df.loc[:, 'TestTypeName'] = 'PARCC ' + (test_type_mapping[domain])
        df.loc[:, 'TestSubjectName'] = (test_subject_mapping[domain])
        df.loc[:, 'Score1Value'] = source.loc[:, '{}_SCALE'.format(domain)]
        df.loc[:, 'Score2Value'] = source.loc[:, '{}_LEVEL'.format(domain)].map(performance_level_mapping)
        
        subset.append(df)
        
    df = pd.concat(subset)
    
    print df.loc[:, 'Score2Value'].unique()
    
    df.loc[:, 'DistrictDBName'] = db_name
    df.loc[:, 'StudentTestID'] = np.NaN
    df.loc[:, 'TestID'] = '0'
    df.loc[:, 'TestName'] = 'PARCC'
    df.loc[:, 'TestTypeID'] = '0'
    df.loc[:, 'TestSubjectID'] = '0'
    df.loc[:, 'Score1Label'] = 'Score'
    df.loc[:, 'Score1Type'] = 'Scale'
    df.loc[:, 'Score2Label'] = 'Performance Level'
    df.loc[:, 'Score2Type'] = 'Level'
    df.loc[:, 'SourceFileName'] = source_file_name
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    df.loc[:, headers].to_csv(r"C:\Workspace\md-frederickcounty\parcc\ELLSTDTST_MD-FREDERICKCOUNTY_PARCC_16_17.csv", index=False)

main()
```
