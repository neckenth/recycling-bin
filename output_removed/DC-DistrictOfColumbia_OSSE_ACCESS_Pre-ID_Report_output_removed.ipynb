```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
from common import is_null

export = pd.read_csv(r"C:\Users\NancyE\Downloads\EllevationExport_20171212_1446.csv", dtype=str)
new_cols = ['{}_export'.format(x) for x in export.columns]
export.columns = new_cols

roster = pd.read_csv(r"C:\Workspace\dc-districtofcolumbia\access\Pre_ID_ACCESS_2017_DCPS_Roster.csv", dtype=str)

df = pd.merge(left=roster, right=export, left_on='District Student ID', right_on='Student #_export', how='left', indicator=True)

today = dt.now().strftime('%m/%d/%Y')

def reformat_bday(x):
    month = x.split('/')[0]
    day = x.split('/')[1]
    year = x.split('/')[2]
    if len(month) == 1:
        month = '0{}'.format(month)
    if len(day) == 1:
        day = '0{}'.format(day)
    return '{}/{}/{}'.format(month, day, year)

def derive_ethnicity(x):
    if x.loc['Ethnicity - Hispanic/Latino'] == 'Y':
        return 'Y'
    elif is_null(x.loc['Ethnicity - Hispanic/Latino']) and x.loc['Ethnicity_export'] == 'Hispanic':
        return 'Y'
    else:
        return np.NaN
    
def years_between(today, x):
    d1 = dt.strptime(today, "%m/%d/%Y")
    d2 = dt.strptime(x, "%m/%d/%Y")
    diff = (d1.year - d2.year)
    if len(str(diff)) == 2:
        return str(diff)
    elif len(str(diff)) == 1:
        return '0{}'.format(str(diff))

lep_services_description_mapping = {
    'Content-based ESL': 'EBL',
    'Push In Model': 'EBL',
    'Pull-Out Model': 'EBL',
    'Dual Language Program - One-way': 'MNL',
    'Dual Language Program - Two-way': 'MNL'
}

df.loc[:, 'Date of Birth'] = df.loc[:, 'Date of Birth'].map(
    lambda x: reformat_bday(x))

df.loc[:, 'Grade'] = df.loc[:, 'Grade'].map(
    lambda x: x if len(x) == 2 else '0{}'.format(x))

df.loc[:, 'Ethnicity - Hispanic/Latino'] = df.apply(derive_ethnicity, axis=1)

df.loc[:, 'Date First Enrolled US School'] = df.apply(
    lambda x: x.loc['Came into US_export'] if not is_null(x.loc['Came into US_export']) else x.loc['District Enrollment_export'], axis=1)

df.loc[df.loc[:, 'Entered LEP_export'].notnull(), 'Length of Time in LEP/ELL Program'] = df.loc[df.loc[:, 'Entered LEP_export'].notnull(), 'Entered LEP_export'].map(lambda x: years_between(str(today), x))

df.loc[:, 'LIEP Classification'] = df.loc[:, 'Service_export'].map(lep_services_description_mapping)

df.loc[:, 'LIEP - Parental Refusal'] = df.loc[:, 'Detailed ELP Status_export'].map(
    lambda x: 'Y' if x == 'Parental Exempt (ELL Level 1-4)' else np.NaN)

df.loc[:, 'Mode of Administration'] = df.loc[:, 'Grade'].map(
    lambda x: 'P' if x == '00' else 'O')

df.loc[:, 'Alternate ACCESS for ELLs Tester'] = df.loc[:, 'Test Type_export'].map(
    lambda x: 'Y' if x == 'Alternate ACCESS for ELLs' else np.NaN)

df.loc[:, roster.columns.tolist()].to_csv(r"C:\Workspace\dc-districtofcolumbia\access\temp\Pre_ID_ACCESS_2017_DCPS_Roster.csv", dtype=str, index=False)
```
