```python
import pandas as pd

df = pd.read_csv(r"C:\Workspace\fl-osceola\act\ELLevation-ScoresACT.csv", dtype=str)

district = 'fl-osceola'
district_db_name = 'ESLREPS-FL-OSCEOLA'

test_type_mapping = {
    'MATH_': 'Math',
    'COMP_': 'Composite',
    'ENGL(ELA)_': 'ELA',
    'ENGL_': 'English',
    'SCIENCE_': 'Science',
    'READ_': 'Reading',
    'STEAM_': 'STEAM',
    'WRITING_': 'Writing'
}

test_subj_mapping = {
    'MATH_': 'Math',
    'COMP_': 'Overall',
    'ENGL(ELA)_': 'ELA',
    'ENGL_': 'ELA',
    'SCIENCE_': 'Science',
    'READ_': 'Reading',
    'STEAM_': 'Science',
    'WRITING_': 'Writing'
}

df.rename(columns={
        'DISTRICT LOCAL ID': 'StudentLocalID',
        'STATE TESTING ID': 'StudentTestID',
        'DATE GIVEN': 'TestAdministrationDate',
        'GRADELEVEL': 'StudentGradeLevel',
        'TEST NAME': 'TestName',  
    }, inplace=True)

domains = [
    'MATH_',
    'COMP_',
    'ENGL(ELA)_',
    'ENGL_',
    'SCIENCE_',
    'READ_',
    'STEAM_',
    'WRITING_'
]

subset = []

for d in domains:
    sub = df[[
            'StudentLocalID',
            'StudentTestID',
            'TestAdministrationDate',
            'StudentGradeLevel',
            'TestName',
            '{}SCALE.SCORE'.format(d)
        ]]
    sub.loc[:, 'TestTypeName'] = 'ACT {}'.format(test_type_mapping[d])
    sub.loc[:, 'TestSubjectName'] = test_subj_mapping[d]
    sub.loc[:, 'Score1Value'] = df.loc[:, '{}SCALE.SCORE'.format(d)]
    subset.append(sub)
    
df = pd.concat(subset)

df.loc[:, 'DistrictDBName'] = district_db_name
df.loc[:, 'TestID'] = '0'
df.loc[:, 'TestTypeID'] = '0'
df.loc[:, 'TestSubjectID'] = '0'
df.loc[:, 'TestGradeLevel'] = df.loc[:, 'StudentGradeLevel']
df.loc[:, 'Score1Label'] = 'Score'
df.loc[:, 'Score1Type'] = 'Scale'
df.loc[:, 'SourceFileName'] = 'ELLevation-ScoresACT.csv'

headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']

df = df[df.loc[:, 'Score1Value'].notnull()]

df.loc[:, headers].to_csv(r"C:\Workspace\fl-osceola\act\ELLSTDTST_FL-Osceola_ACT.csv", index=False)
    
```
