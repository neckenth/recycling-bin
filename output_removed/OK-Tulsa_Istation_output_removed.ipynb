```python
import pandas as pd
import numpy as np
import glob
import os

from pyep.settings import WORKSPACE_ROOT

district = 'ok-tulsa'
district_db_name = 'ESLREPS-OK-TULSA'

subset = []

for f in glob.glob('{}/{}/istation/*.xlsx'.format(WORKSPACE_ROOT, district)):
    df = pd.read_excel(f)
    df.loc[:, 'SourceFileName'] = os.path.split(f)[1]
    subset.append(df)
    
df = pd.concat(subset)

df.rename(columns={
        'STUDENT_ID': 'StudentLocalID',
        'Test Date': 'TestAdministrationDate',
        'Percentile': 'Score1Value'
    }, inplace=True)

df.loc[:, 'DistrictDBName'] = district_db_name
df.loc[:, 'StudentTestID'] = np.NaN
df.loc[:, 'StudentGradeLevel'] = df.loc[:, 'GRADE'].map(
    lambda x: 'Pre-K' if x == -1 else ('K' if x == 0 else x))
df.loc[:, 'TestName'] = 'Istation'
df.loc[:, 'TestID'] = '0'
df.loc[:, 'TestTypeName'] = df.loc[:, 'SourceFileName'].map(
    lambda x: 'Istation' if 'Anderson' in x else 'Istation Espanol') + ' Overall Reading'
df.loc[:, 'TestTypeID'] = '0'
df.loc[:, 'TestSubjectName'] = 'Reading'
df.loc[:, 'TestSubjectID'] = '0'
df.loc[:, 'TestGradeLevel'] = df.loc[:, 'StudentGradeLevel']
df.loc[:, 'Score1Label'] = 'Percentile Rank'
df.loc[:, 'Score1Type'] = 'Scale'

headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']

df.loc[:, headers].to_csv('{}/{}/istation/ELLSTDTST_OK-Tulsa_Istation_Spring_2018.csv'.format(WORKSPACE_ROOT, district), sep=',', index=False)
```
