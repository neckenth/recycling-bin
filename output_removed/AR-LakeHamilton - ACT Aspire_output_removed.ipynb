```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT


def main():
    # fill out the following three district-specific fields
    db_name = 'ESLREPS-AR-LAKEHAMILTON'
    district = 'ar-lakehamilton'
    source_file_name = 'LHSD ESOL 2016 Aspire Results.csv'

    file_name = '{}/{}/ACT Aspire/{}'.format(WORKSPACE_ROOT, district, source_file_name)
    source = pd.read_csv(file_name, dtype=str)

    score_map = {
        'Benchmark' : {'Y' : '2 - Yes',
                       'N' : '1 - No'},
        'Readiness Level' : {'N' : '1 - In Need of Support',
                            'C' : '2 - Close',
                            'R' :  '3 - Ready',
                            'E' : '4 - Exceeding'}}
    
    # first, loop through only score columns for the following 5 domains - all have the same scores/standard columns
    domains = ['English', 'Reading', 'Mathematics', 'Writing', 'Science']

    subset = []

    for domain in domains:

        df_domain = source[['State Student ID',
                    'Actual Grade',
                    '{} Test Administration Date'.format(domain),
                    '{} Tested Grade'.format(domain),
                    '{} Scale Score'.format(domain), 
                    '{} ACT Readiness Benchmark'.format(domain),
                    '{} Readiness Level'.format(domain),
                    '{} National Percentile Rank'.format(domain)]]

        df_domain.loc[:, 'StudentTestID'] = df_domain.loc[:, 'State Student ID']
        df_domain.loc[:, 'TestTypeName'] = 'ACT ASPIRE {}'.format(domain)
        df_domain.loc[:, 'TestSubjectName'] = '{}'.format(domain)
        df_domain.loc[:, 'TestAdministrationDate'] = df_domain.loc[:, '{} Test Administration Date'.format(domain)]
        df_domain.loc[:, 'StudentGradeLevel'] = df_domain.loc[:, 'Actual Grade']
        df_domain.loc[:, 'TestGradeLevel'] = df_domain.loc[:, '{} Tested Grade'.format(domain)]
        df_domain.loc[:, 'Score1Label'] = 'Score'
        df_domain.loc[:, 'Score1Type'] = 'Scale'
        df_domain.loc[:, 'Score1Value'] = df_domain.loc[:, '{} Scale Score'.format(domain)]
        df_domain.loc[:, 'Score2Label'] = 'Benchmark'
        df_domain.loc[:, 'Score2Type'] = 'Level'
        df_domain.loc[:, 'Score2Value'] = df_domain.loc[:, '{} ACT Readiness Benchmark'.format(domain)].map(score_map['Benchmark'])
        df_domain.loc[:, 'Score3Label'] = 'Readiness Level'
        df_domain.loc[:, 'Score3Type'] = 'Level'
        df_domain.loc[:, 'Score3Value'] = df_domain.loc[:, '{} Readiness Level'.format(domain)].map(score_map['Readiness Level'])
        df_domain.loc[:, 'Score4Label'] = 'National Percentile Rank'
        df_domain.loc[:, 'Score4Type'] = 'Scale'
        df_domain.loc[:, 'Score4Value'] = df_domain.loc[:, '{} National Percentile Rank'.format(domain)]

        df_domain = df_domain[[
                'StudentTestID',
                'StudentGradeLevel',
                'TestAdministrationDate',
                'TestTypeName',
                'TestSubjectName',
                'TestGradeLevel',
                'Score1Label',
                'Score1Type',
                'Score1Value',
                'Score2Label',
                'Score2Type',
                'Score2Value',
                'Score3Label',
                'Score3Type',
                'Score3Value',
                'Score4Label',
                'Score4Type',
                'Score4Value'
            ]]

        subset.append(df_domain)
        
    # read in score columns for composite test separately - only 1 score
    df_composite = source[['State Student ID',
                          'Actual Grade', 
                          'ACT Aspire Composite Score']]

    df_composite.loc[:, 'StudentTestID'] = df_composite.loc[:, 'State Student ID']
    df_composite.loc[:, 'TestTypeName'] = 'ACT ASPIRE Composite'
    df_composite.loc[:, 'TestSubjectName'] = 'Overall'
    df_composite.loc[:, 'TestAdministrationDate'] = np.NaN
    df_composite.loc[:, 'StudentGradeLevel'] = df_composite.loc[:, 'Actual Grade']
    df_composite.loc[:, 'TestGradeLevel'] = df_composite.loc[:, 'Actual Grade']
    df_composite.loc[:, 'Score1Label'] = 'Score'
    df_composite.loc[:, 'Score1Type'] = 'Scale'
    df_composite.loc[:, 'Score1Value'] = df_composite.loc[:, 'ACT Aspire Composite Score']

    subset.append(df_composite)

    # read in score columns for ELA test separately - only 2 scores, different order
    df_ela = source[['State Student ID',
                    'Actual Grade', 
                     'ELA Score',
                     'ELA Readiness']]

    df_ela.loc[:, 'StudentTestID'] = df_ela.loc[:, 'State Student ID']
    df_ela.loc[:, 'TestTypeName'] = 'ACT ASPIRE ELA'
    df_ela.loc[:, 'TestSubjectName'] = 'Overall'
    df_ela.loc[:, 'TestAdministrationDate'] = np.NaN
    df_ela.loc[:, 'StudentGradeLevel'] = df_ela.loc[:, 'Actual Grade']
    df_ela.loc[:, 'TestGradeLevel'] = df_ela.loc[:, 'Actual Grade']
    df_ela.loc[:, 'Score1Label'] = 'Score'
    df_ela.loc[:, 'Score1Type'] = 'Scale'
    df_ela.loc[:, 'Score1Value'] = df_ela.loc[:, 'ELA Score']
    df_ela.loc[:, 'Score2Label'] = 'Readiness'
    df_ela.loc[:, 'Score2Type'] = 'Level'
    df_ela.loc[:, 'Score2Value'] = df_ela.loc[:, 'ELA Readiness'].map(score_map['Benchmark'])

    subset.append(df_ela)
    
    # concatenate all domain data frames together
    df = pd.concat(subset)

    # set processed file default columns - these are the same for all records regardless of domain
    df.loc[:, 'DistrictDBName'] = db_name
    df.loc[:, 'StudentLocalID'] = np.NaN
    df.loc[:, 'TestName'] = 'ACT Aspire'
    df.loc[:, 'TestID'] = '0'
    df.loc[:, 'TestTypeID'] = '0'
    df.loc[:, 'TestSubjectID'] = '0'
    df.loc[:, 'SourceFileName'] = source_file_name

    # headers specific to standardized test score format
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    # transform Test Dates into datetime format
    df['TestAdministrationDate'] = pd.to_datetime(df['TestAdministrationDate'], format='%m%d%Y', errors='coerce')
    
    # English test scores have ELA as subject, not English
    df.loc[df['TestTypeName'] == 'ACT ASPIRE English', 'TestSubjectName'] = 'ELA'
    
    # Math test scores have Math as subject, not Mathematics
    df.loc[df['TestSubjectName'] == 'Mathematics', 'TestSubjectName'] = 'Math'
    
    # forward-fill blank Test Administration Dates - there were no dates provided for ELA or Composite scores
    df['TestAdministrationDate'] = df['TestAdministrationDate'].fillna(method='pad')

    df.loc[:, headers].to_csv(r"C:\Workspace\ar-lakehamilton\ACT Aspire\ELLSTDTST_AR-LakeHamilton_ACTAspire_2016.csv", index=False)
    
main()







```
