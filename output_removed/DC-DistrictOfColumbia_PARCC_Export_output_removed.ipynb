```python
import pandas as pd

students = pd.read_excel(r"C:\Users\NancyE\Downloads\Monitoring In Question Need PARCC Scores 9-21-17.xlsx", converters={'ID#' : str})

scores = pd.read_csv("C:\Workspace\dc-districtofcolumbia\standardized-tests\parcc\ELLSTDTST_DC-DistrictOfColumbia_PARCC_2017.csv", dtype=str, usecols=['StudentLocalID','TestSubjectName', 'Score2Value'])

df = pd.melt(frame=scores,
           id_vars=['StudentLocalID', 'TestSubjectName'],
           value_vars=['Score2Value'])

results = pd.merge(left=students, right=df, left_on='ID#', right_on='StudentLocalID', how='inner')

results_ela = results[results.loc[:, 'TestSubjectName'] == 'ELA']
results_math = results[results.loc[:, 'TestSubjectName'] == 'Math']

results_ela.loc[:, 'PARCC Reading'] = results_ela.loc[:, 'value']
results_math.loc[:, 'PARCC Math'] = results_math.loc[:, 'value']

results_ela.drop(labels=['variable', 'value', 'StudentLocalID', 'TestSubjectName'], inplace=True, axis=1)
results_math.drop(labels=['variable', 'value', 'StudentLocalID', 'TestSubjectName'], inplace=True, axis=1)

def helper(group):
    group.loc[:, 'PARCC Math'] = group.loc[:, 'PARCC Math'].max()
    return group

new_df = results_ela.combine_first(results_math)

new_df = new_df.groupby('ID#').apply(helper)

new_df = new_df[new_df.loc[:, 'PARCC Reading'].notnull()]

new_df.to_csv(r"C:\Workspace\dc-districtofcolumbia\standardized-tests\parcc\PARCC Recommendations 2017_populated.csv", index=False)
```
