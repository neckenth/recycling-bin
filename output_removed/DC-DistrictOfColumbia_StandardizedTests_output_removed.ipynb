```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT
from pyep import qualify_file_name
from common import is_null


def main_moy():

    district = 'dc-districtofcolumbia'
    db_name = 'ESLREPS-DC-DISTRICTOFCOLUMBIA'
    source_file_name = 'SY2017-18 MOY Assessment Data 3-23-18.xlsx'

    f = r"C:\Workspace\dc-districtofcolumbia\standardized-tests\SY2017-18 MOY Assessment Data 3-23-18.xlsx"

    xl = pd.ExcelFile(f)

    data_frames = {}

    for sheet in xl.sheet_names:
        data_frames[sheet] = xl.parse(xl.sheet_names.index(sheet))

    for frame in data_frames.keys():
        if '- Student' not in frame:
            data_frames.pop(frame)
        else:
            new_frame = frame.split(' ')[2]
            data_frames[new_frame.encode('utf-8')] = data_frames.pop(frame)
    
    headers = ['DistrictDBName',
                'StudentTestID',
                'StudentLocalID',
                'StudentGradeLevel',
                'TestAdministrationDate',
                'TestID',
                'TestName',
                'TestTypeID',
                'TestTypeName',
                'TestSubjectID',
                'TestSubjectName',
                'TestGradeLevel',
                'Score1Label',
                'Score1Type',
                'Score1Value',
                'Score2Label',
                'Score2Type',
                'Score2Value',
                'Score3Label',
                'Score3Type',
                'Score3Value',
                'Score4Label',
                'Score4Type',
                'Score4Value',
                'Score5Label',
                'Score5Type',
                'Score5Value',
                'Score6Label',
                'Score6Type',
                'Score6Value',
                'Score7Label',
                'Score7Type',
                'Score7Value',
                'Score8Label',
                'Score8Type',
                'Score8Value',
                'SourceFileName']

    ##################################DIBELS################################

    dibels_benchmark_level = {
        'Well Below Benchmark' : '1 - Well Below Benchmark',
        'Benchmark' : '3 - At or Above Benchmark',
        'Below Benchmark' : '2 - Below Benchmark'
    }

    dibels = data_frames['DIBELS-']

    dibels.rename(columns={
            'Studentid' : 'StudentLocalID',
            'Grade' : 'StudentGradeLevel',
            'Test Date' : 'TestAdministrationDate',
            'Assessment Measure-Composite Score-Score' : 'Score1Value'
        }, inplace=True)

    dibels.loc[:, 'DistrictDBName'] = db_name
    dibels.loc[:, 'StudentTestID'] = np.NaN
    dibels.loc[:, 'TestGradeLevel'] = dibels.loc[:, 'StudentGradeLevel']
    dibels.loc[:, 'TestID'] = '0'
    dibels.loc[:, 'TestName'] = 'DIBELS'
    dibels.loc[:, 'TestTypeID'] = '0'
    dibels.loc[:, 'TestTypeName'] = 'DIBELS Composite'
    dibels.loc[:, 'TestSubjectID'] = '0'
    dibels.loc[:, 'TestSubjectName'] = 'Reading'
    dibels.loc[:, 'Score1Label'] = 'Score'
    dibels.loc[:, 'Score1Type'] = 'Scale'
    dibels.loc[:, 'Score2Label'] = 'Benchmark Level'
    dibels.loc[:, 'Score2Type'] = 'Level'
    dibels.loc[:, 'Score2Value'] = dibels.loc[:, 'Assessment Measure-Composite Score-Levels'].map(dibels_benchmark_level)
    dibels.loc[:, 'SourceFileName'] = qualify_file_name(source_file_name, 'DIBELS')

    dibels.loc[:, headers].to_csv('{}/{}/standardized-tests/ELLSTDTST_DC-DistrictOfColumbia_DIBELS_MOY_17-18.csv'.format(WORKSPACE_ROOT, district), index=False)

    ########################################TRC##########################################
    trc_proficiency_level = {
        'Far Below Proficient' : '1 - Far Below Proficient',
        'Below Proficient' : '2 - Below Proficient',
        'Proficient' : '3 - Proficient',
        'Above Proficient': '4 - Above Proficient'
    }

    trc = data_frames['TRC-']

    trc.rename(columns={
            'Studentid' : 'StudentLocalID',
            'Grade' : 'StudentGradeLevel',
            'Test Date': 'TestAdministrationDate'
        }, inplace=True)

    trc.loc[:, 'DistrictDBName'] = db_name
    trc.loc[:, 'StudentTestID'] = np.NaN
    trc.loc[:, 'TestGradeLevel'] = trc.loc[:, 'StudentGradeLevel']
    trc.loc[:, 'TestID'] = '0'
    trc.loc[:, 'TestName'] = 'TRC'
    trc.loc[:, 'TestTypeID'] = '0'
    trc.loc[:, 'TestTypeName'] = 'TRC'
    trc.loc[:, 'TestSubjectID'] = '0'
    trc.loc[:, 'TestSubjectName'] = 'Reading'
    trc.loc[:, 'Score1Label'] = 'Proficiency Level'
    trc.loc[:, 'Score1Type'] = 'Level'
    trc.loc[:, 'Score1Value'] = trc.loc[:, 'Assessment Measure-TRC Proficiency Level-Levels'].map(trc_proficiency_level)
    trc.loc[:, 'SourceFileName'] = qualify_file_name(source_file_name, 'TRC')

    trc.loc[:, headers].to_csv('{}/{}/standardized-tests/ELLSTDTST_DC-DistrictOfColumbia_TRC_MOY_17-18.csv'.format(WORKSPACE_ROOT, district), index=False)

    ##############################################iReady#########################################

    iready_relative_placement_mapping = {
        '1 Grade Level Below': '1 level below',
        '2+ Grade Levels Below': '2 or more levels below',
        'On/Above Grade Level': 'On/Above level'
    }

    def format_placement(row):
        if 'Level' in row.loc['performancelevel'].encode('utf8'):
            return row.loc['performancelevel'].encode('utf8').split(' ')[1]
        elif 'Level' not in row.loc['performancelevel'].encode('utf8'):
            return '{} - {}'.format(row.loc['performancelevel'].encode('utf8').split(' ')[1],
                                    row.loc['performancelevel'].encode('utf8').split(' ')[0])

    iready = data_frames['iReady-']

    iready.rename(columns={
            'studentid': 'StudentLocalID',
            'grade': 'StudentGradeLevel',
            'testdate': 'TestAdministrationDate',
            'scalescore': 'Score3Value',
        }, inplace=True)

    iready.loc[:, 'DistrictDBName'] = db_name
    iready.loc[:, 'StudentTestID'] = np.NaN
    iready.loc[:, 'TestGradeLevel'] = iready.loc[:, 'StudentGradeLevel']
    iready.loc[:, 'TestID'] = '0'
    iready.loc[:, 'TestName'] = 'i-Ready'
    iready.loc[:, 'TestTypeID'] = '0'
    iready.loc[:, 'TestTypeName'] = 'i-Ready Diagnostic ELA'
    iready.loc[:, 'TestSubjectID'] = '0'
    iready.loc[:, 'TestSubjectName'] = 'Reading'
    iready.loc[:, 'Score1Label'] = 'Relative Placement'
    iready.loc[:, 'Score1Type'] = 'Level'
    iready.loc[:, 'Score1Value'] = iready.loc[:, 'ongrdlvl'].map(iready_relative_placement_mapping)
    iready.loc[:, 'Score2Label'] = 'Placement'
    iready.loc[:, 'Score2Type'] = 'Level'
    iready.loc[:, 'Score2Value'] = iready.apply(format_placement, axis=1)
    iready.loc[:, 'Score3Label'] = 'Scale Score'
    iready.loc[:, 'Score3Type'] = 'Scale'
    iready.loc[:, 'SourceFileName'] = qualify_file_name(source_file_name, 'iReady')

    iready.loc[:, headers].to_csv('{}/{}/standardized-tests/ELLSTDTST_DC-DistrictOfColumbia_iReady_MOY_17-18.csv'.format(WORKSPACE_ROOT, district), index=False)

    ##################################################MAP##########################################
    mp = data_frames['MAP']

    mp.rename(columns={
            'studentid': 'StudentLocalID',
            'grade': 'StudentGradeLevel',
            'teststartdate': 'TestAdministrationDate',
            'testritscore': 'Score1Value'
        }, inplace=True)

    mp.loc[:, 'DistrictDBName'] = db_name
    mp.loc[:, 'StudentTestID'] = np.NaN
    mp.loc[:, 'TestGradeLevel'] = mp.loc[:, 'StudentGradeLevel']
    mp.loc[:, 'TestID'] = '0'
    mp.loc[:, 'TestName'] = 'NWEA MAP'
    mp.loc[:, 'TestTypeID'] = '0'
    mp.loc[:, 'TestTypeName'] = mp.loc[:, 'testname'].map(
        lambda x: 'NWEA MAP Spanish Mathematics' if 'Spanish' in x else 'NWEA MAP Mathematics')
    mp.loc[:, 'TestSubjectID'] = '0'
    mp.loc[:, 'TestSubjectName'] = 'Math'
    mp.loc[:, 'Score1Label'] = 'RIT'
    mp.loc[:, 'Score1Type'] = 'Scale'
    mp.loc[:, 'Score2Label'] = 'Percentile'
    mp.loc[:, 'Score2Type'] = 'Scale'
    mp.loc[:, 'Score2Value'] = mp.loc[:, 'testpercentile'].map(
        lambda x: str(x).split('.')[0] if not is_null(x) else np.NaN)
    mp.loc[:, 'SourceFileName'] = qualify_file_name(source_file_name, 'MAP')
        
    mp = mp[mp.loc[:, 'StudentGradeLevel'] != 'OT']

    mp.loc[:, headers].to_csv('{}/{}/standardized-tests/ELLSTDTST_DC-DistrictOfColumbia_NWEA_MAP_MOY_17-18.csv'.format(WORKSPACE_ROOT, district), index=False)

main_moy()
```
