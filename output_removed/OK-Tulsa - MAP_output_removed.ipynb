```python
import pandas as pd
import numpy as np
import os

from pyep.settings import WORKSPACE_ROOT


def main():
        
        
    def derive_test_type(row):
        if 'Math' in row.loc['SourceFileName']:
            return 'MAP Mathematics'
        elif 'Reading' in row.loc['SourceFileName']:
            return 'MAP Reading'

    
    def derive_test_subject(row):
        if 'Math' in row.loc['SourceFileName']:
            return 'Math'
        elif 'Reading' in row.loc['SourceFileName']:
            return 'Reading'
    
    
    correct_grade_levels = {
        '0' : 'K',
        '-1' : 'Pre-K'
    }
    
    db_name = 'ESLREPS-OK-TULSA'
    district = 'ok-tulsa'

    subset = []

    for file in os.listdir('{}/{}/spring tests/map/'.format(WORKSPACE_ROOT, district)):
        df = pd.read_csv('{}/{}/spring tests/map/'.format(WORKSPACE_ROOT, district) + file, dtype=str)
        df.loc[:, 'SourceFileName'] = file
        subset.append(df)

    df = pd.concat(subset)

    df.rename(columns={
            'Student ID' : 'StudentLocalID',
            'Grade Level' : 'StudentGradeLevel',
            'Spring Test' : 'Score1Value',
            'Spring Percentile' : 'Score2Value'
        },
             inplace=True)

    df.loc[:, 'DistrictDBName'] = db_name
    df.loc[:, 'StudentTestID'] = np.NaN
    df.loc[:, 'StudentGradeLevel'] = df.loc[:, 'StudentGradeLevel'].map(
        lambda x: x if not x in ['0', '-1'] else correct_grade_levels[x])
    df.loc[:, 'TestGradeLevel'] = df.loc[:, 'StudentGradeLevel']
    df.loc[:, 'TestID'] = 0
    df.loc[:, 'TestName'] = 'NWEA MAP'
    df.loc[:, 'TestTypeID'] = 0
    df.loc[:, 'TestTypeName'] = df.apply(derive_test_type, axis=1)
    df.loc[:, 'TestSubjectID'] = 0
    df.loc[:, 'TestSubjectName'] = df.apply(derive_test_subject, axis=1)
    df.loc[:, 'Score1Label'] = 'RIT'
    df.loc[:, 'Score1Type'] = 'Scale'
    df.loc[:, 'Score2Label'] = 'Percentile'
    df.loc[:, 'Score2Type'] = 'Scale'

    headers = ['DistrictDBName',
                'StudentTestID',
                'StudentLocalID',
                'StudentGradeLevel',
                'TestAdministrationDate',
                'TestID',
                'TestName',
                'TestTypeID',
                'TestTypeName',
                'TestSubjectID',
                'TestSubjectName',
                'TestGradeLevel',
                'Score1Label',
                'Score1Type',
                'Score1Value',
                'Score2Label',
                'Score2Type',
                'Score2Value',
                'Score3Label',
                'Score3Type',
                'Score3Value',
                'Score4Label',
                'Score4Type',
                'Score4Value',
                'Score5Label',
                'Score5Type',
                'Score5Value',
                'Score6Label',
                'Score6Type',
                'Score6Value',
                'Score7Label',
                'Score7Type',
                'Score7Value',
                'Score8Label',
                'Score8Type',
                'Score8Value',
                'SourceFileName']

    df.loc[:, headers].to_csv(r"C:\Workspace\ok-tulsa\spring tests\ELLSTDTST_OK-Tulsa_MAP_Spring_16_17.csv", index=False)
    
main()
    
    
```
