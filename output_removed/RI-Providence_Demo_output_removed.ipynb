```python
import pandas as pd
import numpy as np

from common import is_null

from pyep import read_delimited, qualify_file_name
from pycron import fetch_from_sftp, copy_to_processor_directory


def main():
    # Edit these variables
    district = 'ri-providence'
    filename = 'Demographics.csv'
    remote_directory = '/data/'
    data_type = 'student'
    file_watcher_type = 'SIS'

    file = fetch_from_sftp(district_name=district,
                           file_name=filename,
                           remote_directory=remote_directory,
                           data_type=data_type,
                           add_datestamp=True)

    df = read_delimited(file, castNull=True, isInteractive=False)
    
    print df.columns.tolist()
    
    df['Elv_LEPStatus'] = df.apply(derive_elv_lep_status, axis=1)

    df['Elv_DateEnteredLEP'] = df.apply(
        lambda x: x['Start Date'] if not x['Elv_LEPStatus'] == 'TDNQ' else np.NaN, axis=1)

    df['Elv_Initial_Waiver_Date'] = df['Initial Waiver Date'].map(
        lambda x: np.NaN if is_null(x) else pd.to_datetime(x))

    df['Elv_Other_Waiver_Date'] = df['Other Waiver Date'].map(
        lambda x: np.NaN if is_null(x) else pd.to_datetime(x))

    df['Elv_DateRefused'] = df.apply(
        lambda x: max(pd.to_datetime(x['Elv_Initial_Waiver_Date']), pd.to_datetime(x['Elv_Other_Waiver_Date'])),
        axis=1)

    df['Elv_BirthCountry'] = df['Birth Place'].map(
        lambda x: x if 'USA' not in str(x) else 'USA')

     # handle duplicate records
    # identify duplicate records
    dups = df.loc[df.loc[:, 'Other ID'].duplicated(), 'Other ID']
    # add new column to store Start Date while changing date info
    df.loc[:, 'Elv_Start Date'] = pd.to_datetime(df.loc[:, 'Start Date'])
    # for students whose records are duplicated _and_ whose 'Start Date' values are blank, put in a placeholder date in the past
    df.loc[(df.loc[:, 'Other ID'].isin(dups) & df.loc[:, 'Start Date'].isnull()), 'Elv_Start Date'] = pd.to_datetime('1/1/1900')
    # sort dataframe from smallest to largest based on 'Other ID' and 'Elv_Start Date'
    df = df.sort_values(by = ['Elv_Start Date', 'Other ID'], ascending=True)
    # drop_duplicates using the 'keep'=last parameter of dropping the first record for each duplicate (drop record where 'Elv_Start Date' is earlier and keep the record with the most recent 'Elv_Start Date')
    # this 'Elv_Start Date' does not get mapped to anything
    df.drop_duplicates('Other ID', keep='last', inplace=True)
    # replace the placeholder 'Elv_Start Date' values with np.NaN values to make them blank again
    df.loc[:, 'Elv_Start Date'].replace(to_replace=pd.to_datetime('1/1/1900'), value=np.NaN, inplace=True)

    df.to_csv(qualify_file_name(file, 'processed'), index=False)

    copy_to_processor_directory(district,
                                qualify_file_name(file, 'processed'),
                                file_watcher_type)


def derive_elv_lep_status(row):
    condition_1 = row.loc['Current Status Desc'] == 'Yes'
    condition_1a = row.loc['Current Status Desc'] == 'No'
    condition_2 = row.loc['Current Status Desc'] == 'Monitor Year 1'
    condition_3 = row.loc['Current Status Desc'] == 'Monitor Year 2'
    condition_7 = row.loc['Student is ELL Tested not Qualified *If NO*'] == 'Yes'
    condition_8 = row.loc['Student is ELL Tested not Qualified *If NO*'] == 'No'
    if condition_1 & condition_7:
        return 'ELL'
    elif condition_7 & condition_2:
        return 'M1'
    elif condition_7 & condition_3:
        return 'M2'
    elif condition_8 & condition_1a:
        return 'TDNQ'
    elif condition_7 & condition_1a:
        return 'Fully Exited/Ever EL'
    else:
        return 'Unknown'

if __name__ == '__main__':
    main()
```
