```python
import pandas as pd
import numpy as np
import os

from pyep.settings import WORKSPACE_ROOT

def main():
    db_name = 'ESLREPS-RI-PROVIDENCE'
    district = 'ri-providence'
#     source_file_name = 'PPSD_PARCC_SY1516_ELLCENSUS_MATCH.csv'
    
    file_subset = []
    
    for f in os.listdir('{}/{}/standardized-tests/star_update/'.format(WORKSPACE_ROOT, district)):
        source_file = pd.read_csv(r'{}/{}/standardized-tests/star_update/'.format(WORKSPACE_ROOT, district) + f, dtype=str)
        source_file.columns = map(str.lower, source_file.columns)
        source_file.loc[:, 'SourceFileName'] = f
        file_subset.append(source_file)
        
    source = pd.concat(file_subset)

    test_dates = ['_boy', '_moy', '_eoy']
    file_specific_test_dates = []
    
    subset = []
    
#     for test_date in test_dates:
#         if test_date in ''.join(source.columns.tolist()):
#             print ''.join(source.columns.tolist())
# #             print test_date
#             file_specific_test_dates.append(test_date)
#             print file_specific_test_dates
    for test_date in test_dates:
        df = source[['sasid',
                    'grade',
                    'scaledscore{}'.format(test_date),
                    'benchmark{}'.format(test_date),
                    'stardate{}'.format(test_date),
                    'SourceFileName']]

        df.loc[:, 'StudentTestID'] = source.loc[:, 'sasid']
        df.loc[:, 'StudentGradeLevel'] = source.loc[:, 'grade']
        df.loc[:, 'StudentGradeLevel'] = df.loc[:, 'StudentGradeLevel'].replace(' ', np.NaN)
        df.loc[:, 'TestGradeLevel'] = source.loc[:, 'grade']
        df.loc[:, 'TestGradeLevel'] = df.loc[:, 'TestGradeLevel'].replace(' ', np.NaN)
        df.loc[:, 'TestAdministrationDate'] = source.loc[:, 'stardate{}'.format(test_date)]
        df.loc[:, 'TestAdministrationDate'] = df.loc[:, 'TestAdministrationDate'].replace(' ', np.NaN)
        df.loc[:, 'TestTypeName'] = 'STAR ' + source.apply(derive_test_type, axis=1)
        df.loc[:, 'TestSubjectName'] = source.apply(derive_test_subject, axis=1)
        df.loc[:, 'Score1Value'] = source.loc[:, 'scaledscore{}'.format(test_date)]
        df.loc[:, 'Score2Value'] = source.loc[:, 'benchmark{}'.format(test_date)]
        df.loc[:, 'SourceFileName'] = source.loc[:, 'SourceFileName']

        subset.append(df)
        
    df = pd.concat(subset)
    
    df.loc[:, 'DistrictDBName'] = db_name
    df.loc[:, 'StudentLocalID'] = np.NaN
    df.loc[:, 'TestID'] = 0
    df.loc[:, 'TestName'] = 'STAR'
    df.loc[:, 'TestTypeID'] = 0
    df.loc[:, 'TestSubjectID'] = 0
    df.loc[:, 'Score1Label'] = 'Scaled Score'
    df.loc[:, 'Score1Type'] = 'Scale'
    df.loc[:, 'Score2Label'] = 'Benchmark'
    df.loc[:, 'Score2Type'] = 'Level'
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    print len(df)
    
    df = df[df.loc[:, 'TestAdministrationDate'].notnull()]
    df = df[df.loc[:, 'StudentGradeLevel'].notnull()]
    
    print len(df)

    df.loc[:, headers].to_csv(r"C:\Workspace\ri-providence\standardized-tests\star_final\test.csv", index=False)
    
def derive_test_type(row):
    if 'EARLYLIT' in row.loc['SourceFileName']:
        return 'Early Literacy'
    elif 'READ' in row.loc['SourceFileName']:
        return 'Reading'
    elif 'MATH' in row.loc['SourceFileName']:
        return 'Math'
    
def derive_test_subject(row):
    if 'EARLYLIT' in row.loc['SourceFileName'] or 'READ' in row.loc['SourceFileName']:
        return 'ELA'
    else:
        return 'Math'
main()
```
