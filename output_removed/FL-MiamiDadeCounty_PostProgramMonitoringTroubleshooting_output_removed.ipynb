```python
import pandas as pd
import numpy as np
import datetime as dt

from pyep import read_delimited
from common import is_null

TODAY_MP = dt.datetime.now().date()

MARKING_PERIODS = \
    {
        '1': {'start':'2017-08-21',
              'end': '2017-10-25'},
        '2': {'start': '2017-10-26',
              'end': '2018-03-21'},
        '3': {'start': '2018-03-22',
              'end': '2018-05-31'},
        '4': {'start': '2018-06-01',
              'end': '2018-06-08'}
    }


def derive_marking_period(x):
    if pd.to_datetime(MARKING_PERIODS['1']['start']).date() <= x <= pd.to_datetime(MARKING_PERIODS['1']['end']).date():
        return '1'
    if pd.to_datetime(MARKING_PERIODS['2']['start']).date() <= x <= pd.to_datetime(MARKING_PERIODS['2']['end']).date():
            return '2'
    elif pd.to_datetime(MARKING_PERIODS['3']['start']).date() <= x <= pd.to_datetime(MARKING_PERIODS['3']['end']).date():
        return '3'
    elif pd.to_datetime(MARKING_PERIODS['4']['start']).date() <= x <= pd.to_datetime(MARKING_PERIODS['4']['end']).date():
        return '4'


def derive_meeting_date(row, mp):
    date_columns = ['POST_PROGRAM_REVIEW_DATE1',
                    'POST_PROGRAM_REVIEW_DATE2',
                    'POST_PROGRAM_REVIEW_DATE3',
                    'POST_PROGRAM_REVIEW_DATE4']
    possible_dates = []
    for col in date_columns:
        if not (is_null(row.loc[col])):
            if pd.to_datetime(MARKING_PERIODS['{}'.format(mp)]['start']) <= pd.to_datetime(row.loc[col]) <= pd.to_datetime(MARKING_PERIODS['{}'.format(mp)]['end']):
                possible_dates.append(pd.to_datetime(row.loc[col]))
    if len(possible_dates) > 0:
        return max(possible_dates)
    else:
        return np.NaN


def name_next_reclass_meeting(row):
    meeting_dates = ['1', '2', '3', '4']
    for date in meeting_dates:
        if pd.to_datetime(row.loc['POST_PROGRAM_REVIEW_DATE{}'.format(date)]) == row.loc['Elv_NextPostPrgReclassDate']:
            return 'POST_PROGRAM_REVIEW_DATE{}'.format(date)

df = read_delimited(r"C:\Workspace\fl-miamidadecounty\student\ELLEVATION_DEMOGRAPHIC-20180323.csv",
                    sep=',',
                    castNull=True,
                    isInteractive=False,
                    inferDialect=False)

current_marking_period = derive_marking_period(x=TODAY_MP)

df.loc[:, 'Elv_NextPostPrgReclassDate'] = df.apply(
    lambda x: derive_meeting_date(x, mp=current_marking_period), axis=1)

df.loc[:, 'Elv_NextPostPrgReclassDate_Flag'] = df.apply(
    lambda x: name_next_reclass_meeting(x), axis=1)

df.loc[:, 'Elv_Comment'] = df.apply(
    lambda x: '{}, {}, {}, {}'.format(x.loc['POST_PROGRAM_REVIEW_DATE1'], x.loc['POST_PROGRAM_REVIEW_DATE2'],
                                      x.loc['POST_PROGRAM_REVIEW_DATE3'], x.loc['POST_PROGRAM_REVIEW_DATE4'])
    if not is_null(x.loc['POST_PROGRAM_REVIEW_DATE1']) and not is_null(x.loc['POST_PROGRAM_REVIEW_DATE2']) and
    not is_null(x.loc['POST_PROGRAM_REVIEW_DATE3']) and not is_null(x.loc['POST_PROGRAM_REVIEW_DATE4']) else np.NaN, axis=1)

guardians = ['G1', 'G2', 'G3']

for p in guardians:
    df.loc[:, 'Elv_FatherName_{}'.format(p)] = df.apply(
        lambda x: x.loc['{}_Name'.format(p)] if x.loc['{}_Relationship'.format(p)] == 'F' else np.NaN, axis=1)
    df.loc[:, 'Elv_MotherName_{}'.format(p)] = df.apply(
        lambda x: x.loc['{}_Name'.format(p)] if x.loc['{}_Relationship'.format(p)] == 'M' else np.NaN, axis=1)
    df.loc[:, 'Elv_GuardianName_{}'.format(p)] = df.apply(
        lambda x: x.loc['{}_Name'.format(p)] if x.loc['{}_Relationship'.format(p)] == 'G' else np.NaN, axis=1)

parents = ['Father', 'Mother', 'Guardian']

for r in parents:
    df.loc[:, 'Elv_{}Name'.format(r)] = df[['Elv_{}Name_G1'.format(r),
                                            'Elv_{}Name_G2'.format(r),
                                            'Elv_{}Name_G3'.format(r)]].fillna('').sum(axis=1)

df.drop(labels=['Elv_FatherName_G1',
                'Elv_FatherName_G2',
                'Elv_FatherName_G3',
                'Elv_MotherName_G1',
                'Elv_MotherName_G2',
                'Elv_MotherName_G3',
                'Elv_GuardianName_G1',
                'Elv_GuardianName_G2',
                'Elv_GuardianName_G3'], axis=1, inplace=True)

df.to_csv(r"C:\Workspace\fl-miamidadecounty\student\ELLEVATION_DEMOGRAPHIC-20180323_processed.csv", index=False)
```
