```python
# Assessment Fact - NYSITELL

import pandas as pd
import numpy as np
import os
import re

from pyep import read_delimited
from pyep.settings import WORKSPACE_ROOT

df = read_delimited(r"C:\Workspace\ny-johnsoncitycentral\elp-test-score\NY031502_AssessmentFact_SFTP_201612090132.csv")

def helper(group):
    group.loc[:, 'Elv_GradeLevel'] = group.loc[:, 'Elv_GradeLevel'].max()
    return group

df['ITEM DESCRIPTION'].str.extract('(\w{8}:\s)(\w{5}\s\w{1,4}\s)(.-?\d{0,2}?\s)(\w{5,}\s?\w{0,5})')

df.loc[df['ITEM DESCRIPTION'].str.contains('Total Score'), 'Elv_GradeLevel'] = df['ITEM DESCRIPTION'].str.extract('\w{8}:\s\w{5}\s\w{1,4}\s(.-?\d{0,2}?\s)\w{5,}\s?\w{0,5}')

df = df.groupby('STUDENT ID').apply(helper)

df['Elv_Domain'] = df['ITEM DESCRIPTION'].str.extract('\w{8}:\s\w{5}\s\w{1,4}\s.-?\d{0,2}?\s(\w{5,}\s?\w{0,5})')

df['Elv_Domain'].unique()

df['Elv_Level'] = df['ITEM DESCRIPTION'].str.extract('\w{8}:\s(\w{5}\s\w{1,4})\s.-?\d{0,2}?\s\w{5,}\s?\w{0,5}')

df=df.pivot_table(index=['STUDENT ID', 'TEST DATE', 'Elv_GradeLevel', 'Elv_Level'],
                  columns='Elv_Domain',
                 values=['NUMERIC SCORE', 'STANDARD ACHIEVED CODE'],
                 aggfunc=np.max).reset_index(col_level=1)

df.columns=[' '.join(col).strip() for col in df.columns.tolist()]

print len(df)

df.to_csv(r"C:\Workspace\ny-johnsoncitycentral\elp-test-score\NY031502_AssessmentFact_SFTP_201612090132_processed.csv", index=False)
```

```python
#nyseslat
import pandas as pd
import numpy as np
import re
import os

from pyep.settings import WORKSPACE_ROOT

subset = []

for file in os.listdir(WORKSPACE_ROOT + '/ny-johnsoncitycentral/elp-test-score/nyseslat/'):
    df = pd.read_csv(WORKSPACE_ROOT + '/ny-johnsoncitycentral/elp-test-score/nyseslat/' + file)
    subset.append(df)

df = pd.concat(subset)

df['Subtest Identifier'].str.extract('(\w{8}:\s)(.-?\d{0,2}?\s)(\w{2,}\s?\w{5}?)')

df['Elv_Domain'] = df['Subtest Identifier'].str.extract('\w{8}:\s.-?\d{0,2}?\s(\w{2,}\s?\w{5}?)')

df.loc[df['Year'] == '12-13', 'Elv_DateGiven'] = pd.to_datetime('04-10-2013')

df.loc[df['Year'] == '13-14', 'Elv_DateGiven'] = pd.to_datetime('04-09-2014')

df.loc[df['Year'] == '14-15', 'Elv_DateGiven'] = pd.to_datetime('04-14-2015')

df.loc[df['Year'] == '15-16', 'Elv_DateGiven'] = pd.to_datetime('04-12-2016')

df = df.pivot_table(index=['Student ID', 'Elv_DateGiven', 'Grade Level'],
                  columns='Elv_Domain',
                 values=['Raw Score', 'Scaled Score', 'Achievement Level'],
                 aggfunc=np.max).reset_index(col_level=1)

df.columns=[' '.join(col).strip() for col in df.columns.tolist()]

df = df.replace('Not available', np.NaN)

df_post_2015 = df[df.loc[:, 'Elv_DateGiven'] > pd.to_datetime('01-01-2015')]

df_pre_2015 = df[df.loc[:, 'Elv_DateGiven'] < pd.to_datetime('01-01-2015')]

df_pre_2015.to_csv(r"C:\Workspace\ny-johnsoncitycentral\elp-test-score\nyseslat\combined_processed_pre_2015.csv", index=False)

df_post_2015.to_csv(r"C:\Workspace\ny-johnsoncitycentral\elp-test-score\nyseslat\combined_processed_post_2015.csv", index=False)

```
