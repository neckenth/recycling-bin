```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT

from common import is_null

def main():
    db_name = 'ESLREPS-TX-GARLANDISD'
    district = 'tx-garlandisd'
    source_file_name = 'MOY MAP 2018.csv'

    file_name = '{}/{}/map_nwea/{}'.format(WORKSPACE_ROOT, district, source_file_name)
    source = pd.read_csv(file_name, dtype=str)
    
    test_subject_mapping = {
        'NWEA MAP Mathematics' : 'Math',
        'NWEA MAP Reading' : 'Reading',
        'NWEA MAP Spanish Mathematics' : 'Math'
    }
    
    source.rename(columns={
            'StudentID':'StudentLocalID',
            'StudentGrade':'StudentGradeLevel'
        }, inplace=True)
    
    source['Test Version'] = source['Test Version'].apply(lambda x: x.decode('unicode_escape').encode('ascii', 'ignore').strip())

    source['MOY Date'] = source['MOY Date'].apply(lambda x: x.decode('unicode_escape').encode('ascii', 'ignore').strip())

    
    source.loc[:, 'DistrictDBName'] = db_name
    source.loc[:, 'StudentTestID'] = np.NaN
    source.loc[:, 'TestGradeLevel'] = source.loc[:, 'StudentGradeLevel']
    source.loc[:, 'TestAdministrationDate'] = source.loc[:, 'MOY Date']
    source.loc[:, 'TestID'] = '0'
    source.loc[:, 'TestName'] = 'NWEA MAP'
    source.loc[:, 'TestTypeID'] = '0'
    source.loc[:, 'TestTypeName'] = 'NWEA MAP ' + source.apply(derive_test_type, axis=1)
    source.loc[:, 'TestSubjectID'] = '0'
    source.loc[:, 'TestSubjectName'] = source.loc[:, 'TestTypeName'].map(test_subject_mapping)
    source.loc[:, 'Score1Label'] = 'RIT'
    source.loc[:, 'Score1Type'] = 'Scale'
    source.loc[:, 'Score1Value'] = source.loc[:, 'MOY RIT']
    source.loc[:, 'Score2Label'] = 'Percentile'
    source.loc[:, 'Score2Type'] = 'Scale'
    source.loc[:, 'Score2Value'] = source.loc[:, 'MOY Perc']
    source.loc[:, 'SourceFileName'] = source_file_name
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    source = source[source.loc[:, 'StudentGradeLevel'].notnull()]
    
    source.loc[:, headers].to_csv(r"C:\Workspace\tx-garlandisd\map_nwea\ELLSTDTST_TX-GarlandISD_MAP_MOY_17_18.csv", index=False, encoding='utf-8')

def derive_test_type(row):
    if 'Reading' in str(row.loc['Test Version']):
        return 'Reading'
    elif 'Math' in str(row.loc['Test Version']) and 'Spanish' not in str(row.loc['Test Version']):
        return 'Mathematics'
    elif 'Spanish' in str(row.loc['Test Version']):
        return 'Spanish Mathematics'
    
main()
```
