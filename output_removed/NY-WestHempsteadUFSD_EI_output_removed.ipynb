```python
import pandas as pd
import numpy as np
import os

from pyep.settings import WORKSPACE_ROOT

def main():
    db_name = 'ESLREPS-NY-WESTHEMPSTEADUFSD'
    district = 'ny-westhempsteadufsd'
    
    subset = []
    
    for file in os.listdir(WORKSPACE_ROOT + '/' + district + '/standardizedtests/'):
        source = pd.read_excel(WORKSPACE_ROOT + '/' + district + '/standardizedtests/' + file)
        source.loc[:, 'SourceFileName'] = file
        subset.append(source)
        
    df = pd.concat(subset)
    print len(df)
    df = df[~df.loc[:, 'ITEM_DESC'].str.contains('Regents')]
    df = df[~df.loc[:, 'ITEM_DESC'].str.contains('NYSAA')]
    df = df[~df.loc[:, 'ITEM_DESC'].str.contains('NYSESLAT')]
    print len(df)
    
    test_type_map = {
        '2015-2016 Grades 3-8 ELA Elementary_Middle Level.xlsx' : 'EI ELA',
        '2015-2016 Grades 3-8 MATH Elementary_Middle Level.xlsx' : 'EI Math',
        '2015-2016 Grades 3-8 Science Elementary_Middle Level.xlsx' : 'EI Science'
    }
    
    test_subject_map = {
        '2015-2016 Grades 3-8 ELA Elementary_Middle Level.xlsx' : 'ELA',
        '2015-2016 Grades 3-8 MATH Elementary_Middle Level.xlsx' : 'Math',
        '2015-2016 Grades 3-8 Science Elementary_Middle Level.xlsx' : 'Science'
    }
    
    df.loc[:, 'DistrictDBName'] = db_name
    df.loc[:, 'StateTestID'] = np.NaN
    df.loc[:, 'StudentLocalID'] = df.loc[:, 'STUDENT_ID']
    df.loc[:, 'StudentGradeLevel'] = df.loc[:, 'CURR_GRADE_LVL']
    df.loc[:, 'TestAdministrationDate'] = pd.to_datetime('4/5/2016')
    df.loc[:, 'TestID'] = '0'
    df.loc[:, 'TestName'] = 'NY EI'
    df.loc[:, 'TestTypeID'] = '0'
    df.loc[:, 'TestTypeName'] = df.loc[:, 'SourceFileName'].map(test_type_map)
    df.loc[:, 'TestSubjectID'] = '0'
    df.loc[:, 'TestSubjectName'] = df.loc[:, 'SourceFileName'].map(test_subject_map)
    df.loc[:, 'TestGradeLevel'] = df.loc[:, 'CURR_GRADE_LVL']
    df.loc[:, 'Score1Label'] = df.loc[:, 'TestSubjectName'].map(
        lambda x: 'Scale Score' if not x == 'Science' else 'Raw Score')
    df.loc[:, 'Score1Type'] = 'Scale'
    df.loc[:, 'Score1Value'] = df.loc[:, 'NUMERIC_SCORE']
    df.loc[:, 'Score2Label'] = 'Performance Level'
    df.loc[:, 'Score2Type'] = 'Level'
    df.loc[:, 'Score2Value'] = df.loc[:, 'STANDARD_ACHIEVED']
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    df.loc[:, headers].to_csv(r"C:\Workspace\ny-westhempsteadufsd\standardizedtests\ELLSTDTST_NY-WestHempsteadUFSD_EI_2016.csv", index=False)

main()
```
