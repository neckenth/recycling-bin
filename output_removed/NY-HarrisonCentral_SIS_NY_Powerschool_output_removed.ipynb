```python
# LEP - student lite and program fact

import sys
import datetime as dt
import fnmatch
import json
import os
import pysftp
import shutil
import time
import ftplib
import socket
import ssl

from common import is_null
from pyep import qualify_file_name, read_delimited
from pyep.transform import derive_from_date_exited
from pycron import copy_to_processor_directory, fetch_from_sftp
from pyep.settings import WORKSPACE_ROOT


def main():
    # Edit these variables
    district = 'ny-harrisoncentral'
#     today = dt.datetime.now().strftime('%Y%m%d')
    filename1 = 'NY660501_StudentLite_SFTP_201710131447.csv'
    filename2 = 'NY660501_ProgramFact_SFTP_201710131448.csv'
    remote_directory = '/data/'
    data_type = 'student'
    file_watcher_type = 'SIS'

    student_lite = fetch_from_sftp(district_name=district,
                                        file_name=filename1,
                                        remote_directory=remote_directory,
                                        data_type=data_type,
                                        add_datestamp=False)

    program_fact = fetch_from_sftp(district_name=district,
                                        file_name=filename2,
                                        remote_directory=remote_directory,
                                        data_type=data_type,
                                        add_datestamp=False)

    student = read_delimited(student_lite, castNull=True)
    program = read_delimited(program_fact, castNull=True)
    
    program = program[program.loc[:, 'PROGRAMS CODE'].isin(['5709'])]

    merged = student.merge(
        program[['STUDENT ID', '*PROGRAM SERVICE NAME', 'ENDING DATE',
                 '*PROGRAM SERVICE CATEGORY', '*EXIT REASON']],
        left_on='StudentId',
        right_on='STUDENT ID',
        how='inner')

    merged['Elv_LEPStatus'] = merged['ENDING DATE'].map(
        lambda x: derive_from_date_exited(x) if not is_null(x) else 'LEP')
    
    merged = merged[~merged.loc[:, 'LocationCode'].isin(['001', 'OP06'])]

    merged.to_csv(qualify_file_name(student_lite, 'processed'), index=False)

    copy_to_processor_directory(district,
                                qualify_file_name(student_lite, 'processed'),
                                file_watcher_type)


if __name__ == '__main__':
    main()
```
