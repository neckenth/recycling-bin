```python
# 14-15
import pandas as pd
import numpy as np
import os
import re

from pyep import read_delimited
from pyep.settings import WORKSPACE_ROOT

df = read_delimited(r"C:\Workspace\ny-ramapocentral\NYSESLAT\di-7423\NYSITELL\NYSITELL 1415 Ramapo.csv", dtype=str)

def helper(group):
    group.loc[:, 'Elv_GradeLevel'] = group.loc[:, 'Elv_GradeLevel'].max()
    return group

df['Assessment'].str.extract('(\w{8}:\s)(\w{5}\s\w{1,4}\s)(.-?\d{0,2}?\s)(\w{2,}\s?\w{5}?)')

df.loc[df['Assessment'].str.contains('Total Score'), 'Elv_GradeLevel'] = df.loc[:, 'Assessment'].str.extract('\w{8}:\s\w{5}\s\w{1,4}\s(.-?\d{0,2}?\s)\w{2,}\s?\w{5}?')

df = df.groupby('STUDENT_ID').apply(helper)

df['Elv_Domain'] = df['Assessment'].str.extract('\w{8}:\s\w{5}\s\w{1,4}\s.-?\d{0,2}?\s(\w{2,}\s?\w{5}?)')

df['Elv_Level'] = df['Assessment'].str.extract('\w{8}:\s(\w{5}\s\w{1,4}\s).-?\d{0,2}?\s\w{2,}\s?\w{5}?')

df=df.pivot_table(index=['STUDENT_ID', 'Assessment Date', 'Elv_GradeLevel', 'Elv_Level'],
                  columns='Elv_Domain',
                 values='Numeric Score',
                 aggfunc=np.max).reset_index(col_level=1)

df.columns=[''.join(col).strip() for col in df.columns.tolist()]

print len(df)

#df.to_csv(r"C:\Workspace\ny-ramapocentral\NYSESLAT\di-7423\NYSITELL\NYSITELL 1415 Ramapo_processed.csv", index=False)
```
