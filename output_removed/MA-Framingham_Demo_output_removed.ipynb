```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT
from pyep.transform import iderive_lep_status
from pycron import fetch_from_sftp, copy_to_processor_directory

def main():
    # Edit these variables
    district = 'ma-framingham'
    filename_1 = 'DistrictELLs.xlsx'
    filename_2 = 'FLEP.xlsx'
    remote_directory = '/data/'
    data_type = 'student'
    file_watcher_type = 'SIS'

    file_1 = fetch_from_sftp(district_name=district,
                           file_name=filename_1,
                           remote_directory=remote_directory,
                           data_type=data_type,
                           add_datestamp=True)

    file_2 = fetch_from_sftp(district_name=district,
                             file_name=filename_2,
                             remote_directory = remote_directory,
                             data_type = data_type,
                             add_datestamp = True)

    root = WORKSPACE_ROOT + '/{}/{}/'.format(district, data_type)

    subset = [file_1, file_2]

    subset_processed = []

    # convert each of the files from excel to pandas dataframes
    for file in subset:
        # add converters parameters to prevent int inputs from being converted into floats
        df_excel = pd.read_excel(file, convert_float=True, converters={'SASID': str, 'LASID': str, 'Grade': str})
        subset_processed.append(df_excel)

    # concatenate three files together
    df = pd.concat(subset_processed, join='outer', ignore_index=True)

    # proceed with regular pre-processing code
    df.drop_duplicates(subset='SASID', inplace=True)

    # convert dates to string using strftime for the purposes of using iderive_lep_status
    df['Elv_DateEntered'] = df['_Bil_Entry_Date'].map(lambda x: x.strftime('%Y-%m-%d') if not pd.isnull(x) else np.NaN)
    
    df['Elv_DateExited'] = df['_Bil_Exit_Date'].map(lambda x: x.strftime('%Y-%m-%d') if not pd.isnull(x) else np.NaN)

    df.loc[:, 'Elv_LEPStatus'] = df.apply(iderive_lep_status,
                                          axis=1,
                                          dateEnteredLep='Elv_DateEntered',
                                          dateExitedLep='Elv_DateExited')

    df.loc[df['ELL Program (026)'] == 'Opt-out', 'Elv_LEPStatus'] = 'Opt-out'

    df.to_csv(root + 'combined-processed.csv', index=False, sep=',')

    copy_to_processor_directory(district,
                                (root + 'combined-processed.csv'),
                                file_watcher_type)

if __name__ == '__main__':
    main()
```
