```python
# student
import pandas as pd
import numpy as np


def main():
    
    binary_mapping = {
        'No' : '0',
        'Yes' : '1'
    }
    
    df = pd.read_csv(r"C:\Users\NancyE\Downloads\EllevationExport_20170418_1521.csv", dtype=str)

    df.loc[:, 'Elv_Counter'] = df.groupby('Student #').cumcount(ascending=True)
    df.loc[:, 'Elv_Counter'] = df.loc[:, 'Elv_Counter'].map(str)

    df_1 = pd.pivot_table(data=df,
                   index=['Student #'],
                   columns = ['Elv_Counter'],
                   values=['Service', 'Service Provider', 'Service Schedule', 'Service Minutes per Week'],
                   aggfunc=np.max).reset_index(col_level=1)

    df_1.columns = [' '.join(col).strip() for col in df_1.columns.tolist()]

    service_info = ['Service', 'Service Provider', 'Service Schedule', 'Service Minutes per Week']

    for item in service_info:
        df_1['{} 4'.format(item)].rename('{} 5'.format(item), inplace=True)
        df_1['{} 3'.format(item)].rename('{} 4'.format(item), inplace=True)
        df_1['{} 2'.format(item)].rename('{} 3'.format(item), inplace=True)
        df_1['{} 1'.format(item)].rename('{} 2'.format(item), inplace=True)
        df_1['{} 0'.format(item)].rename('{} 1'.format(item), inplace=True)
    
    df = pd.merge(df, df_1, on='Student #', how='outer')
    
    column_mapping = {
        'Student #' : 'STUDENT #',
        'Birthday' : 'DATE OF BIRTH', 
        'City/Town of Birth' : 'PLACE OF BIRTH',
        'Native Language' : 'NATIVE LANGUAGE',
        'Birth Country' : 'BIRTH COUNTRY',
        'ELP Designation' : 'ELP DESIGNATION',
        'LEP Status' : 'LEP STATUS',
        'Monitoring Status' : 'MONITORING STATUS',
        'School LEA Code' : 'LEACODE',
        'Teacher' : 'TEACHER',
        'Grade Level' : 'GRADE LEVEL',
        'Yrs in US School' : 'YRS IN US SCHOOL',
        'Entered LEP' : 'ENTERED LEP',
        'LEP Exit Date' : 'EXITED LEP',
        '(K-12) US School Entry' : 'ENROLLED IN US',
        'LEA Enrollment' : 'DISTRICT ENROLLMENT',
        'Monitored since' : 'MONITORED SINCE',
        'Refugee' : 'REFUGEE',
        'Service 1' : 'LEP SERVICE 1',
        'Service Schedule 1' : 'SERVICE 1 FREQUENCY',
        'Service Minutes per Week 1' : 'SERVICE 1 DURATION',
        'Service Provider 1' : 'SERVICE 1 TEACHER',
        'Service 2' : 'LEP SERVICE 2',
        'Service Schedule 2' : 'SERVICE 2 FREQUENCY',
        'Service Minutes per Week 2' : 'SERVICE 2 DURATION',
        'Service Provider 2' : 'SERVICE 2 TEACHER'
    }
    
    df = df.rename(columns=column_mapping)
    
    df.loc[:, 'MONITORING STATUS'] = df.loc[:, 'MONITORING STATUS'].map(binary_mapping)
    df.loc[:, 'NOM'] = df.loc[:, 'NOM'].map(binary_mapping)
    df.loc[:, 'REFUGEE'] = df.loc[:, 'REFUGEE'].map(binary_mapping)
    df.loc[:, 'SERVICE 1 FREQUENCY'] = df.loc[:, 'SERVICE 1 FREQUENCY'].map(
        lambda x: x if x not in ['/ for ', '------'] else np.NaN)
    df.loc[:, 'SERVICE 2 FREQUENCY'] = df.loc[:, 'SERVICE 2 FREQUENCY'].map(
        lambda x: x if x not in ['/ for ', '------'] else np.NaN)
    
    df = df[['STUDENT #',
             'SSN',
             'DATE OF BIRTH',
             'PLACE OF BIRTH',
             'NATIVE LANGUAGE',
             'BIRTH COUNTRY',
             'ELP DESIGNATION',
             'LEP STATUS',
             'MONITORING STATUS',
             'LEACODE',
             'TEACHER',
             'GRADE LEVEL',
             'NOM',
             'YRS IN US SCHOOL',
             'ENTERED LEP',
             'EXITED LEP',
             'ENROLLED IN US',
             'DISTRICT ENROLLMENT',
             'MONITORED SINCE',
             'REFUGEE',
             'LEP SERVICE 1',
             'SERVICE 1 FREQUENCY',
             'SERVICE 1 DURATION',
             'SERVICE 1 TEACHER',
             'LEP SERVICE 2',
             'SERVICE 2 FREQUENCY',
             'SERVICE 2 DURATION',
             'SERVICE 2 TEACHER']]  

    df.to_csv(r"C:\Workspace\nc-charlotte-mecklenburg\export\student\test_3.csv", index=False)
    
main()



```
