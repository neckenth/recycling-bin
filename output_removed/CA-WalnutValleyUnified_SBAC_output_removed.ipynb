```python
import pandas as pd
import numpy as np

from pyep import read_delimited, qualify_file_name
from pyep.settings import WORKSPACE_ROOT


def main():
    ncesid = '0641280'
    district = 'ca-walnutvalleyunified'
    source_file_name = 'SBACData.csv'

    file_name = '{}/{}/sbac/{}'.format(WORKSPACE_ROOT, district,
                                       source_file_name)
    source = read_delimited(file_name, inferDialect=False, sep='^',
                            isInteractive=False, castNull=True)

    df = source[['RecordType',
                 'Statewide Student_Identifier(SSID)',
                 'CALPADS Grade',
                 'GradeAssessed',
                 'PaperandPencilTestCompletionDate(PAPER)',
                 'TestStartDateandTime1',
                 'TestStartDateandTime2',
                 'TestMode-OnlineorPaper',
                 'SmarterClaim1PerformanceLevel',
                 'SmarterClaim2PerformanceLevel',
                 'SmarterClaim3PerformanceLevel',
                 'SmarterClaim4PerformanceLevel',
                 'ScaleScore',
                 'PerformanceLevel',
                 'AchievementLevels']]

    test_type_map = {
        '01': 'SBAC ELA',
        '02': 'SBAC Math',
        '03': 'CAA for ELA',
        '04': 'CAA for Mathematics',
        '05': 'CST for Science',
        '06': 'CMA for Science',
        '07': 'CAPA for Science',
        '08': 'STS for RLA'
    }

    score_map = {
        'PL': {'5': '5 - Advanced',
               '4': '4 - Proficient',
               '3': '3 - Basic',
               '2': '2 - Below Basic',
               '1': '1 - Far Below Basic'},
        'AL': {'4': '4 - Standard Exceeded',
               '3': '3 - Standard Met',
               '2': '2 - Standard Nearly Met',
               '1': '1 - Standard Not Met'},
        'SCPL': {'1': '1 - Below Standard',
                 '2': '2 - At or Near Standard',
                 '3': '3 - Above Standard'}
    }

    df.loc[:, 'StudentTestID'] = df.loc[:,
                                 'Statewide Student_Identifier(SSID)']
    df.loc[:, 'TestTypeName'] = df.loc[:, 'RecordType'].map(test_type_map)
    df.loc[:, 'StudentGradeLevel'] = df.loc[:, 'GradeAssessed']
    df.loc[:, 'TestGradeLevel'] = df.loc[:, 'GradeAssessed']

    df.loc[:, 'NCESID'] = ncesid
    df.loc[:, 'StudentLocalID'] = np.NaN
    df.loc[:, 'TestID'] = '0'
    df.loc[:, 'TestName'] = 'SBAC'
    df.loc[:, 'TestTypeID'] = '0'
    df.loc[:, 'TestSubjectID'] = '0'
    df.loc[:, 'Score1Label'] = 'Scale Score'
    df.loc[:, 'Score1Type'] = 'Scale'
    df.loc[:, 'Score2Type'] = 'Level'
    df.loc[:, 'SourceFileName'] = source_file_name
    df.loc[:, 'Score7Label'] = np.NaN
    df.loc[:, 'Score7Type'] = np.NaN
    df.loc[:, 'Score7Value'] = np.NaN
    df.loc[:, 'Score8Label'] = np.NaN
    df.loc[:, 'Score8Type'] = np.NaN
    df.loc[:, 'Score8Value'] = np.NaN

    df.loc[:, 'Score1Value'] = df.loc[:, 'ScaleScore']

    df.loc[df.loc[:, 'RecordType'].isin(
        ['01', '03', '08']), 'TestSubjectName'] = 'ELA'
    df.loc[
        df.loc[:, 'RecordType'].isin(['02', '04']), 'TestSubjectName'] = 'Math'
    df.loc[df.loc[:, 'RecordType'].isin(
        ['05', '06', '07']), 'TestSubjectName'] = 'Science'

    df.loc[df.loc[:, 'RecordType'].isin(
        ['01', '02', '03', '04']), 'Score2Label'] = 'Achievement Level'
    df.loc[df.loc[:, 'RecordType'].isin(
        ['05', '06', '07', '08']), 'Score2Label'] = 'Performance Level'
    df.loc[df.loc[:, 'RecordType'].isin(
        ['01', '02', '03', '04']), 'Score2Value'] = df.loc[:,
                                                    'AchievementLevels']
    df.loc[df.loc[:, 'RecordType'].isin(
        ['05', '06', '07', '08']), 'Score2Value'] = df.loc[:,
                                                    'PerformanceLevel']

    df.loc[df.loc[:, 'RecordType'].isin(
        ['01']), 'Score3Label'] = 'Claim 1 - Reading'
    df.loc[df.loc[:, 'RecordType'].isin(
        ['02']), 'Score3Label'] = 'Claim 1 - Concepts and Procedures'
    df.loc[df.loc[:, 'RecordType'].isin(['01', '02']), 'Score3Type'] = 'Level'
    df.loc[df.loc[:, 'RecordType'].isin(['01', '02']), 'Score3Value'] = df.loc[
                                                                        :,
                                                                        'SmarterClaim1PerformanceLevel']

    df.loc[df.loc[:, 'RecordType'].isin(
        ['01']), 'Score4Label'] = 'Claim 2 - Writing'
    df.loc[df.loc[:, 'RecordType'].isin(
        ['02']), 'Score4Label'] = 'Claim 2 - Problem Solving and Modeling'
    df.loc[df.loc[:, 'RecordType'].isin(['01', '02']), 'Score4Type'] = 'Level'
    df.loc[df.loc[:, 'RecordType'].isin(['01', '02']), 'Score4Value'] = df.loc[
                                                                        :,
                                                                        'SmarterClaim2PerformanceLevel']

    df.loc[df.loc[:, 'RecordType'].isin(
        ['01']), 'Score5Label'] = 'Claim 3 - Listening'
    df.loc[df.loc[:, 'RecordType'].isin(
        ['02']), 'Score5Label'] = 'Claim 3 - Communicating Reasoning'
    df.loc[df.loc[:, 'RecordType'].isin(['01', '02']), 'Score5Type'] = 'Level'
    df.loc[df.loc[:, 'RecordType'].isin(['01', '02']), 'Score5Value'] = df.loc[
                                                                        :,
                                                                        'SmarterClaim3PerformanceLevel']

    df.loc[(df.loc[:,
            'RecordType'] == '01'), 'Score6Label'] = 'Claim 4 - Research Inquiry'
    df.loc[(df.loc[:, 'RecordType'] == '01'), 'Score6Type'] = 'Level'
    df.loc[(df.loc[:, 'RecordType'] == '01'), 'Score6Value'] = df.loc[:,
                                                               'SmarterClaim4PerformanceLevel']

    df.loc[df.loc[:, 'Score2Label'] == 'Achievement Level', 'Score2Value'] = \
    df.loc[df.loc[:, 'Score2Label'] == 'Achievement Level', 'Score2Value'].map(
        score_map['AL'])
    df.loc[df.loc[:, 'Score2Label'] == 'Performance Level', 'Score2Value'] = \
    df.loc[df.loc[:, 'Score2Label'] == 'Performance Level', 'Score2Value'].map(
        score_map['PL'])

    score_fields = ['Score3Value', 'Score4Value', 'Score5Value', 'Score6Value']

    for field in score_fields:
        df.loc[df.loc[:, 'TestTypeName'].isin(
            ['SBAC ELA', 'SBAC Math']
        ), field] = df.loc[df.loc[:, 'TestTypeName'].isin(
            ['SBAC ELA', 'SBAC Math']), field].map(score_map['SCPL'])

    df.loc[(df.loc[:,
            'TestMode-OnlineorPaper'] == 'P'), 'TestAdministrationDate'] = pd.to_datetime(
        df.loc[(df.loc[:,
                'TestMode-OnlineorPaper'] == 'P'), 'PaperandPencilTestCompletionDate(PAPER)'],
        format='%m%d%Y',
        errors='coerce')

    df.loc[(df.loc[:, 'TestMode-OnlineorPaper'] == 'O'),
           'TestAdministrationDate'] = pd.to_datetime(
        df.loc[(df.loc[:, 'TestMode-OnlineorPaper'] == 'O'),
               'TestStartDateandTime1'],
        format='%Y-%m-%d%H:%M:%S',
        errors='coerce')

    standardized_test_score_header = ['NCESID',
                                      'StudentTestID',
                                      'StudentLocalID',
                                      'StudentGradeLevel',
                                      'TestAdministrationDate',
                                      'TestID',
                                      'TestName',
                                      'TestTypeID',
                                      'TestTypeName',
                                      'TestSubjectID',
                                      'TestSubjectName',
                                      'TestGradeLevel',
                                      'Score1Label',
                                      'Score1Type',
                                      'Score1Value',
                                      'Score2Label',
                                      'Score2Type',
                                      'Score2Value',
                                      'Score3Label',
                                      'Score3Type',
                                      'Score3Value',
                                      'Score4Label',
                                      'Score4Type',
                                      'Score4Value',
                                      'Score5Label',
                                      'Score5Type',
                                      'Score5Value',
                                      'Score6Label',
                                      'Score6Type',
                                      'Score6Value',
                                      'Score7Label',
                                      'Score7Type',
                                      'Score7Value',
                                      'Score8Label',
                                      'Score8Type',
                                      'Score8Value',
                                      'SourceFileName']

    df = df[~df.loc[:, 'Score1Value'].isnull()]
    df.drop_duplicates(inplace=True)
    df = df[(
        ~df.loc[:, 'StudentGradeLevel'].isnull() &
        ~df.loc[:, 'TestGradeLevel'].isnull() &
        ~df.loc[:, 'TestAdministrationDate'].isnull()
    )]

    df[standardized_test_score_header].to_csv(qualify_file_name(file_name,
                                                                'processed'),
                                              index=False)


if __name__ == '__main__':
    main()
```
