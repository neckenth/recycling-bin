```python
import pandas as pd
import numpy as np

from pyep import qualify_file_name
from pyep.settings import WORKSPACE_ROOT

pd.set_option('precision', 0)

def main():
    district = 'ar-littlerocksd'
    db_name = 'ESLREPS-AR-LITTLEROCKSD'
    source_file_name = 'ITBS 16 17.xlsx'
    
    df = pd.read_excel('{}/{}/itbs/{}'.format(WORKSPACE_ROOT, district, source_file_name))
    
    print df.loc[:, 'SCHOOL_YEAR'].unique()
    
    df.rename(columns={
            'STUDENT_ID':'StudentLocalID',
            'STUDENT_GRADE_LEVEL':'StudentGradeLevel'
        }, inplace=True)
    
    df.loc[:, 'DistrictDBName'] = db_name
    df.loc[:, 'StudentTestID'] = np.NaN
    df.loc[:, 'TestGradeLevel'] = df.loc[:, 'StudentGradeLevel']

    test_type_mapping = {'MATH_': 'Math',
                        'READING_': 'Reading',
                         'LANGUAGE_': 'Language',
                        'SCIENCE_': 'Science'}
    
    test_subject_mapping = {'MATH_': 'Math',
                        'READING_': 'Reading',
                         'LANGUAGE_': 'ELA',
                        'SCIENCE_': 'Science'}

    domains = ['READING_', 'LANGUAGE_', 'MATH_', 'SCIENCE_']

    subset_1 = []

    for domain in domains:
        df1 = df[[
                'SCHOOL_YEAR',
                'DistrictDBName',
                'StudentTestID',
                'StudentLocalID',
                'StudentGradeLevel',
                'TestGradeLevel',
                '{}SCALED_SCORE'.format(domain),
                '{}NATIONAL_PERCENTILE_RANK'.format(domain),
                    ]]
        df1.loc[:, 'TestAdministrationDate'] = df.loc[:, 'SCHOOL_YEAR'].map(
            lambda x: pd.to_datetime('5/30/2016') if x == 'LRSD_16' else pd.to_datetime('5/30/2017'))
        df1.loc[:, 'TestTypeName'] = 'ITBS ' + (test_type_mapping[domain])
        df1.loc[:, 'TestSubjectName'] = test_subject_mapping[domain]
        df1.loc[df.loc[:, '{}SCALED_SCORE'.format(domain)].notnull(), 'Score1Value'] = df.loc[df.loc[:, '{}SCALED_SCORE'.format(domain)].notnull(), '{}SCALED_SCORE'.format(domain)].map(
            lambda x: str(int(float(x))))
        df1.loc[df.loc[:, '{}NATIONAL_PERCENTILE_RANK'.format(domain)].notnull(), 'Score2Value'] = df.loc[df.loc[:, '{}NATIONAL_PERCENTILE_RANK'.format(domain)].notnull(), '{}NATIONAL_PERCENTILE_RANK'.format(domain)].map(
            lambda x: str(int(float(x))))
        subset_1.append(df1)

    final = pd.concat(subset_1)

    final.loc[:, 'TestName'] = 'ITBS'
    final.loc[:, 'TestID'] = '0'
    final.loc[:, 'TestTypeID'] = '0'
    final.loc[:, 'TestSubjectID'] = '0'
    final.loc[:, 'Score1Label'] = 'Scale Score'
    final.loc[:, 'Score1Type'] = 'Scale'
    final.loc[:, 'Score2Label'] = 'NPR'
    final.loc[:, 'Score2Type'] = 'Scale'
    final.loc[:, 'SourceFileName'] = source_file_name
    
#     final.loc[:, 'Score1Value'] = final.loc[:, 'Score1Value'].map(
#         lambda x: str(x))
    
#     final.loc[final.loc[:, 'Score1Value'].notnull(), 'Score1Value'] = final.loc[final.loc[:, 'Score1Value'].notnull(), 'Score1Value'].map(
#         lambda x: str(int(float(x))))
#     print final.loc[:, 'Score1Value'].unique()

    headers = ['DistrictDBName',
                'StudentTestID',
                'StudentLocalID',
                'StudentGradeLevel',
                'TestAdministrationDate',
                'TestID',
                'TestName',
                'TestTypeID',
                'TestTypeName',
                'TestSubjectID',
                'TestSubjectName',
                'TestGradeLevel',
                'Score1Label',
                'Score1Type',
                'Score1Value',
                'Score2Label',
                'Score2Type',
                'Score2Value',
                'Score3Label',
                'Score3Type',
                'Score3Value',
                'Score4Label',
                'Score4Type',
                'Score4Value',
                'Score5Label',
                'Score5Type',
                'Score5Value',
                'Score6Label',
                'Score6Type',
                'Score6Value',
                'Score7Label',
                'Score7Type',
                'Score7Value',
                'Score8Label',
                'Score8Type',
                'Score8Value',
                'SourceFileName']
    
    final = final[final.loc[:, 'Score1Value'].notnull()]
    
    

    final.loc[:, headers].to_csv('{}/{}/itbs/{}'.format(WORKSPACE_ROOT, district, 'ELLSTDTST_AR-LittleRockSD_ITBS_17_18.csv'), index=False)
    
main()
```
