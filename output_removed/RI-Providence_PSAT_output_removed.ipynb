```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT

def main():
    db_name = 'ESLREPS-RI-PROVIDENCE'
    district = 'ri-providence'
    source_file_name = 'PPSD_PSAT_SY1516_ELLCENSUS_MATCH.csv'
    
    file_name = '{}/{}/standardized-tests/{}'.format(WORKSPACE_ROOT, district, source_file_name)
    source = pd.read_csv(file_name)
    
    test_types = ['LATEST_PSAT_TOTAL', 'LATEST_PSAT_MATH', 'LATEST_PSAT_EBRW']
    
    subset = []
    
    test_type_mapping = {
        'LATEST_PSAT_TOTAL': 'Total Score',
        'LATEST_PSAT_MATH' : 'Math Section Score',
        'LATEST_PSAT_EBRW' : 'Evidence-Based Reading and Writing Section Score'
    }
    
    test_subject_mapping = {
        'LATEST_PSAT_TOTAL' : 'Overall',
        'LATEST_PSAT_MATH' : 'Math',
        'LATEST_PSAT_EBRW' : 'ELA'
    }
    
    for test_type in test_types:
        df = source[['LASID',
                    'SASID',
                    'Grade',
                    'PSATDATE',
                    '{}'.format(test_type)]]
    
        df.loc[:, 'StudentTestID'] = df.loc[:, 'SASID']
        df.loc[:, 'StudentGradeLevel'] = df.loc[:, 'Grade']
        df.loc[:, 'TestAdministrationDate'] = df.loc[:, 'PSATDATE']
        df.loc[:, 'TestGradeLevel'] = df.loc[:, 'Grade']
        df.loc[:, 'TestTypeName'] = 'PSAT ' + (test_type_mapping[test_type])
        df.loc[:, 'TestSubjectName'] = (test_subject_mapping[test_type])
        df.loc[:, 'Score1Value'] = df.loc[:, '{}'.format(test_type)]
        
        subset.append(df)
    
    df = pd.concat(subset)
    
    df.loc[:, 'DistrictDBName'] = db_name
    df.loc[:, 'StudentLocalID'] = np.NaN
    df.loc[:, 'TestID'] = 0
    df.loc[:, 'TestName'] = 'PSAT'
    df.loc[:, 'TestTypeID'] = 0
    df.loc[:, 'TestSubjectID'] = 0
    df.loc[:, 'Score1Label'] = 'Score'
    df.loc[:, 'Score1Type'] = 'Scale'
    df.loc[:, 'SourceFileName'] = source_file_name
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    df.loc[:, headers].to_csv(r"C:\Workspace\ri-providence\standardized-tests\ELLSTDTST_RI-Providence_PSAT_15-16.csv", index=False)
    
main()
    
    
```
