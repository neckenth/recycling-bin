```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT

def main():
    db_name = 'ESLREPS-NY-WESTHEMPSTEADUFSD'
    district = 'ny-westhempsteadufsd'
    source_file_name = 'NY280227_AssessmentFact_SFTP_201703100918.csv'
    
    file_name = '{}/{}/regents/{}'.format(WORKSPACE_ROOT, district, source_file_name)
    source = pd.read_csv(file_name)
    
    source = source[source.loc[:, 'TEST DESCRIPTION'] == 'Regents']
    
    source.loc[:, 'TestTypeName'] = source.loc[:, 'ITEM DESCRIPTION'].map(
        lambda x: str(x).split(' -')[0])
    
    test_type_map = {
        'Regents Algebra2/Trigonometry' : 'Regents Algebra 2/Trigonometry',
        'Regents Phy Set/Chemistry' : 'Regents Chemistry',
        'Regents Global History' : 'Regents Global History & Geography',
        'Regents Living Environment' : 'Regents Living Environment',
        'Regents Common Core Geometry' : 'Regents Geometry',
        'Regents Common Core Algebra II' : 'Regents Algebra II (Common Core)',
        'Regents Common Core ELA' : 'Regents ELA (Common Core)',
        'Regents US History&Gov\'t' : 'Regents U.S. History & Government',
        'Regents Phy Set/Earth Sci' : 'Regents Earth Science',
        'Regents Common Core Algebra I' : 'Regents Algebra I (Common Core)'
        }
    
    test_subject_map = {
        'Regents Algebra 2/Trigonometry' : 'Math',
        'Regents Chemistry' : 'Science',
        'Regents Global History & Geography' : 'History and Social Science',
        'Regents Living Environment' : 'Science',
        'Regents Geometry' : 'Math',
        'Regents Algebra II (Common Core)' : 'Math',
        'Regents ELA (Common Core)' : 'ELA',
        'Regents U.S. History & Government' : 'History and Social Science',
        'Regents Earth Science' : 'Science',
        'Regents Algebra I (Common Core)' : 'Math'
    }
    
    source.loc[:, 'DistrictDBName'] = db_name
    source.loc[:, 'StudentTestID'] = np.NaN
    source.loc[:, 'StudentLocalID'] = source.loc[:, 'STUDENT ID']
    source.loc[:, 'StudentGradeLevel'] = source.loc[:, '*STUDENT CURRENT GRADE']
    source.loc[:, 'TestAdministrationDate'] = source.loc[:, 'TEST DATE']
    source.loc[:, 'TestID'] = '0'
    source.loc[:, 'TestName'] = 'NY Regents'
    source.loc[:, 'TestTypeID'] = '0'
    source.loc[:, 'TestTypeName'] = source.loc[:, 'TestTypeName'].map(test_type_map)
    source.loc[:, 'TestSubjectID'] = '0'
    source.loc[:, 'TestSubjectName'] = source.loc[:, 'TestTypeName'].map(test_subject_map)
    source.loc[:, 'TestGradeLevel'] = source.loc[:, '*STUDENT CURRENT GRADE']
    source.loc[:, 'Score1Label'] = 'Raw Score'
    source.loc[:, 'Score1Type'] = 'Scale'
    source.loc[:, 'Score1Value'] = source.loc[:, 'NUMERIC SCORE']
    source.loc[:, 'SourceFileName'] = source_file_name
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    source.loc[:, headers].to_csv(r"C:\Workspace\ny-westhempsteadufsd\regents\ELLSTDTST_NY-WestHempsteadUFSD_Regents_2016.csv", index=False)
    
main()
```
