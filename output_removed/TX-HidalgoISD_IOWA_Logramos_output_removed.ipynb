```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT

def main():
    district = 'tx-hidalgoisd'
    district_db_name = 'ESLREPS-TX-HIDALGOISD'
    source_file_name = 'IOWA Test - English scores.csv'

    df = pd.read_csv('{}/{}/IOWA/{}'.format(WORKSPACE_ROOT, district, source_file_name), dtype=str)

    df.rename(columns={
            'STUDENT ID 1': 'StudentLocalID',
            'TEST DATE': 'TestAdministrationDate',
            'GRADE': 'StudentGradeLevel',
        }, inplace=True)

    subjs = ['READING', 'LANGUAGE']

    subset = []

    for s in subjs:
        sub = df[['StudentLocalID',
                 'TestAdministrationDate',
                 'StudentGradeLevel',
                 '{} TOTAL - NCE'.format(s)]]
        sub.rename(columns={
                '{} TOTAL - NCE'.format(s): 'Score1Value'
            }, inplace=True)
        sub.loc[:, 'TestTypeName'] = 'IOWA ' + '{}{}'.format(s[0], s[1:].lower())
        sub.loc[:, 'TestSubjectName'] = sub.loc[:, 'TestTypeName'].map(
            lambda x: 'Reading' if s == 'READING' else 'ELA')
        subset.append(sub)

    df = pd.concat(subset)
    
    df.loc[:, 'DistrictDBName'] = district_db_name
    df.loc[:, 'StudentLocalID'] = df.loc[:, 'StudentLocalID'].str.lstrip(' ').str.lstrip('0')
    df.loc[:, 'TestAdministrationDate'] = df.loc[:, 'TestAdministrationDate'].str.rstrip(' ')
    df.loc[:, 'TestName'] = 'Iowa Assessments'
    df.loc[:, 'TestID'] = '0'
    df.loc[:, 'TestTypeID'] = '0'
    df.loc[:, 'TestSubjectID'] = '0'
    df.loc[:, 'TestGradeLevel'] = df.loc[:, 'StudentGradeLevel']
    df.loc[:, 'Score1Label'] = 'NCE'
    df.loc[:, 'Score1Type'] = 'Scale'
    df.loc[:, 'SourceFileName'] = source_file_name

    headers = ['DistrictDBName',
                'StudentTestID',
                'StudentLocalID',
                'StudentGradeLevel',
                'TestAdministrationDate',
                'TestID',
                'TestName',
                'TestTypeID',
                'TestTypeName',
                'TestSubjectID',
                'TestSubjectName',
                'TestGradeLevel',
                'Score1Label',
                'Score1Type',
                'Score1Value',
                'Score2Label',
                'Score2Type',
                'Score2Value',
                'Score3Label',
                'Score3Type',
                'Score3Value',
                'Score4Label',
                'Score4Type',
                'Score4Value',
                'Score5Label',
                'Score5Type',
                'Score5Value',
                'Score6Label',
                'Score6Type',
                'Score6Value',
                'Score7Label',
                'Score7Type',
                'Score7Value',
                'Score8Label',
                'Score8Type',
                'Score8Value',
                'SourceFileName']
    
    df = df[df.loc[:, 'Score1Value'].notnull()]

    df.loc[:, headers].to_csv('{}/{}/IOWA/ELLSTDTST_TX-HidalgoISD_IOWA_2018.csv'.format(WORKSPACE_ROOT, district), sep=',', index=False)

main()
```
