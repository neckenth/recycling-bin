```python
import pandas as pd
from pyep.settings import WORKSPACE_ROOT
from pyep import qualify_file_name
import numpy as np
df = pd.read_csv(r"C:\Workspace\dc-districtofcolumbia\standardized-tests\parcc\DC PARCC 2017 + MSAA Student-level Master File.csv", dtype=str)

def main():
    db_name = 'ESLREPS-DC-DISTRICTOFCOLUMBIA'
    district = 'dc-districtofcolumbia'
    source_file_name = 'DC PARCC 2017 + MSAA Student-level Master File.csv'
    
    file_name = '{}/{}/standardized-tests/parcc/{}'.format(WORKSPACE_ROOT, district, source_file_name)
    source = pd.read_csv(file_name, dtype=str)
    
    performance_level_mapping = {
        '1' : '1 - Did not meet expectations',
        '2' : '2 - Partially met expectations',
        '3' : '3 - Approached expectations',
        '4' : '4 - Met expectations',
        '5' : '5 - Exceeded expectations',
    }
    
    def find_test_type(x):
        if pd.isnull(x):
            return np.NaN
        else:
            try:
                int(x)
                return 'PARCC Mathematics' 
            except ValueError:
                return 'PARCC ' + str(x)
            pass
        
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    subset = []
    
    domains = ['ela_', 'math_']
    
    for domain in domains:
    # split ELA and Math into two dfs
        source.rename(columns={
                'Student ID' : 'StudentLocalID',
                'grade' : 'StudentGradeLevel',
                '{}scale'.format(domain) : 'Score1Value {}'.format(domain),
                '{}perf'.format(domain) : 'Score2Value {}'.format(domain),
                '{}subclaim1'.format(domain) : 'Score3Value {}'.format(domain),
                '{}subclaim2'.format(domain) : 'Score4Value {}'.format(domain),
                '{}subclaim3'.format(domain) : 'Score5Value {}'.format(domain),
                '{}subclaim4'.format(domain) : 'Score6Value {}'.format(domain)
            }, inplace=True)
        
    source.loc[:, 'DistrictDBName'] = db_name
    source.loc[:, 'StudentTestID'] = np.NaN
    source.loc[:, 'TestAdministrationDate'] = pd.to_datetime('4/25/2017')
    source.loc[:, 'TestGradeLevel'] = source.loc[:, 'StudentGradeLevel']
    source.loc[:, 'TestID'] = 0
    source.loc[:, 'TestName'] = 'PARCC'
    source.loc[:, 'TestTypeID'] = 0
    source.loc[:, 'TestSubjectID'] = 0
    source.loc[:, 'SourceFileName'] = source_file_name
    
    ela = source[['DistrictDBName',
                  'StudentLocalID',
                  'StudentTestID',
                  'TestAdministrationDate',
                  'TestID',
                  'TestName',
                  'TestTypeID',
                  'TestSubjectID',
                 'StudentGradeLevel',
                  'TestGradeLevel',
                 'Score1Value ela_',
                 'Score2Value ela_',
                 'Score3Value ela_',
                 'Score4Value ela_',
                 'Score5Value ela_',
                 'Score6Value ela_',
                 'SourceFileName',
                 'ela_assessment']]

    ela.loc[:, 'TestTypeName'] = 'PARCC ELA'
    ela.loc[:, 'TestSubjectName'] = 'ELA'
    ela.loc[:, 'Score1Label'] = 'Scale Score'
    ela.loc[:, 'Score1Type'] = 'Scale'
    ela.loc[:, 'Score1Value'] = source.loc[:, 'Score1Value ela_']
    ela.loc[:, 'Score2Label'] = 'Performance Level'
    ela.loc[:, 'Score2Type'] = 'Level'
    ela.loc[:, 'Score2Value'] = source.loc[:, 'Score2Value ela_'].map(performance_level_mapping)
    ela.loc[:, 'Score3Label'] = 'Subclaim 1'
    ela.loc[:, 'Score3Type'] = 'Level'
    ela.loc[:, 'Score3Value'] = source.loc[:, 'Score3Value ela_']
    ela.loc[:, 'Score4Label'] = 'Subclaim 2'
    ela.loc[:, 'Score4Type'] = 'Level'
    ela.loc[:, 'Score4Value'] = source.loc[:, 'Score4Value ela_']
    ela.loc[:, 'Score5Label'] = 'Subclaim 3'
    ela.loc[:, 'Score5Type'] = 'Level'
    ela.loc[:, 'Score5Value'] = source.loc[:, 'Score5Value ela_']
    ela.loc[:, 'Score6Label'] = 'Subclaim 4'
    ela.loc[:, 'Score6Type'] = 'Level'
    ela.loc[:, 'Score6Value'] = source.loc[:, 'Score6Value ela_']
    ela.loc[:, 'Score7Label'] = 'Subclaim 5'
    ela.loc[:, 'Score7Type'] = 'Level'
    ela.loc[:, 'Score7Value'] = source.loc[:, 'ela_subclaim5']
    
    ela = ela[ela.loc[:, 'ela_assessment'] == 'PARCC Only - ELA']
    
    subset.append(ela)
    
    math = source[['DistrictDBName',
              'StudentLocalID',
              'StudentTestID',
              'TestAdministrationDate',
              'TestID',
              'TestName',
              'TestTypeID',
              'TestSubjectID',
             'StudentGradeLevel',
              'TestGradeLevel',
             'Score1Value math_',
             'Score2Value math_',
             'Score3Value math_',
             'Score4Value math_',
             'Score5Value math_',
             'SourceFileName',
                  'math_assessment']]

    math.loc[:, 'TestTypeName'] = source.loc[:, 'math_subject'].map(
        lambda x: find_test_type(x))
    math.loc[:, 'TestSubjectName'] = 'Math'
    math.loc[:, 'Score1Label'] = 'Scale Score'
    math.loc[:, 'Score1Type'] = 'Scale'
    math.loc[:, 'Score1Value'] = source.loc[:, 'Score1Value math_']
    math.loc[:, 'Score2Label'] = 'Performance Level'
    math.loc[:, 'Score2Type'] = 'Level'
    math.loc[:, 'Score2Value'] = source.loc[:, 'Score2Value math_'].map(performance_level_mapping)
    math.loc[:, 'Score3Label'] = 'Subclaim 1'
    math.loc[:, 'Score3Type'] = 'Level'
    math.loc[:, 'Score3Value'] = source.loc[:, 'Score3Value math_']
    math.loc[:, 'Score4Label'] = 'Subclaim 2'
    math.loc[:, 'Score4Type'] = 'Level'
    math.loc[:, 'Score4Value'] = source.loc[:, 'Score4Value math_']
    math.loc[:, 'Score5Label'] = 'Subclaim 3'
    math.loc[:, 'Score5Type'] = 'Level'
    math.loc[:, 'Score5Value'] = source.loc[:, 'Score5Value math_']
    math.loc[:, 'Score6Label'] = 'Subclaim 4'
    math.loc[:, 'Score6Type'] = 'Level'
    math.loc[:, 'Score6Value'] = source.loc[:, 'Score6Value math_']
    
    math = math[math.loc[:, 'math_assessment'] == 'PARCC Only - Math']
    
    subset.append(math)
    
    df = pd.concat(subset)
    
    processed_file = qualify_file_name(source_file_name, 'processed')
    
    condition_1 = df.loc[:, 'Score1Value'].notnull()
    condition_2 = df.loc[:, 'Score2Value'].notnull()
    condition_3 = df.loc[:, 'Score3Value'].notnull()
    condition_4 = df.loc[:, 'Score4Value'].notnull()
    condition_5 = df.loc[:, 'Score5Value'].notnull()
    condition_6 = df.loc[:, 'Score6Value'].notnull()
    
    df = df[(condition_1 & condition_2 & condition_3 & condition_4 & condition_5 & condition_6)]
       
    df.loc[:, headers].to_csv('{}/{}/standardized-tests/parcc/ELLSTDTST_DC-DistrictOfColumbia_PARCC_2017.csv'.format(WORKSPACE_ROOT, district, processed_file), index=False)
    
main()
    
```
