```python
# services

import pandas as pd
import numpy as np

df_services = pd.read_csv(r"C:\Workspace\ok-tulsa\export\EllevationExport_EdFi ODS - services-20180425_1341.csv", dtype=str)

df_services.drop(labels=['Service Provider', 'Service Schedule', 'Service Minutes per Week'], inplace=True, axis=1)

df_services.loc[:, 'Elv_Counter'] = df_services.groupby('Student #').cumcount(ascending=True)
df_services.loc[:, 'Elv_Counter'] = df_services.loc[:, 'Elv_Counter'].map(str)


df_services_pivoted = pd.pivot_table(data=df_services,
                                    index=['Student #', 'Services Year/Semester'],
                                    columns =['Elv_Counter'],
                                    values=['Service', 'Service Provider', 'Service Schedule', 'Service Minutes per Week'],
                                    aggfunc=np.max).reset_index(col_level=1)

df_services_pivoted.columns = [' '.join(col).strip() for col in df_services_pivoted.columns.tolist()]

df_services_pivoted.fillna(value=np.NaN, inplace=True)

df_columns_final = []

for col in df_services_pivoted.columns.tolist():
    if ('0' in col or '1' in col or '2' in col or '3' in col or '4' in col or '5' in col) and 'Service' in col:
        col = col.replace(col.split(' ')[-1], str(int(col.split(' ')[-1]) + 1))
        df_columns_final.append(col)
    elif ('Minutes' in col or 'Provider' in col or 'Schedule' in col):
        df_services_pivoted.drop(col, inplace=True)
    else:
        df_columns_final.append(col)
        
df_services_pivoted.columns = df_columns_final

df_services_pivoted.to_csv(r"C:\Workspace\ok-tulsa\export\EllevationExport_EdFi ODS - services-20180425_1341_processed.csv", index=False)
```
