```python
import pandas as pd

df = pd.read_excel(r"C:\Users\NancyE\Downloads\DCPS Students to Update with ACCESS Scores 9-29-17.xlsx")

df = df[df.loc[:, 'Detailed ELP Status'].isin(['LES', 'NES'])]

df.loc[df.loc[:, 'Composite/Overall Proficiency Level'].notnull(), 'Elv_DetailedELPStatus'] = df.loc[df.loc[:, 'Composite/Overall Proficiency Level'].notnull(), 'Composite/Overall Proficiency Level'].map(
    lambda x: 'FLEP 1' if int(x) >= 5.0 else 'Unknown')
    
condition_1 = df.loc[:, 'Elv_DetailedELPStatus'] == 'FLEP 1'

df.loc[condition_1, 'Elv_ELPDes'] = 'Not ELL'
df.loc[condition_1, 'Elv_LEPStatus'] = 'FLEP'
df.loc[condition_1, 'Elv_MonStatus'] = 'FLEP Year 1'
df.loc[condition_1, 'Elv_DateExited'] = pd.to_datetime('6/14/17')
df.loc[condition_1, 'Elv_MonSinceDate'] = pd.to_datetime('8/21/17')
df.loc[condition_1, 'Elv_ReceivingServices'] = 'No'
    
df.to_csv(r"C:\Workspace\dc-districtofcolumbia\access\NES_LES_updates.csv", index=False)

```
