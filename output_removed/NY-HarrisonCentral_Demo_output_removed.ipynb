```python
import sys

from common import is_null
from pyep import qualify_file_name, read_delimited
from pyep.transform import derive_from_date_exited
from pycron import fetch_from_sftp, copy_to_processor_directory


def main():
    # Edit these variables
    district = 'ny-harrisoncentral'
    filename1 = sys.argv[1]
    filename2 = sys.argv[2]
    remote_directory = '/data/'
    data_type = 'student'
    file_watcher_type = 'SIS'

    student_lite = fetch_from_sftp(district_name=district,
                                   file_name=filename1,
                                   remote_directory=remote_directory,
                                   data_type=data_type,
                                   add_datestamp=False)

    program_fact = fetch_from_sftp(district_name=district,
                                   file_name=filename2,
                                   remote_directory=remote_directory,
                                   data_type=data_type,
                                   add_datestamp=False)

    student = read_delimited(r'C:\Workspace\ny-harrisoncentral\student\NY660501_StudentLite_SFTP_201610181123.csv', 
                             castNull=True)
    program = read_delimited(r"C:\Workspace\ny-harrisoncentral\student\NY660501_ProgramFact_SFTP_201610181126.csv", 
                             castNull=True)

    merged = student.merge(
        program[['STUDENT ID', '*PROGRAM SERVICE NAME', 'ENDING DATE',
                 '*PROGRAM SERVICE CATEGORY', '*EXIT REASON']],
        left_on='StudentId',
        right_on='STUDENT ID',
        how='inner')

    merged['Elv_LEPStatus'] = merged['ENDING DATE'].map(
        lambda x: derive_from_date_exited(x) if not is_null(x) else 'LEP')

    merged.to_csv(qualify_file_name(student_lite, 'processed'), index=False)

#     copy_to_processor_directory(district,
#                                 qualify_file_name(student_lite, 'processed'),
#                                 file_watcher_type)


if __name__ == '__main__':
    main()
```

```python
import pandas as pd
import numpy as np

from pyep.transform import derive_from_date_exited_4y

from common import is_null

df = pd.read_csv(r"C:\Workspace\ny-harrisoncentral\student\Former ELLs.csv", dtype=str)

df['Program Start Date'] = pd.to_datetime(df['Program Start Date'], errors='coerce')

df['Elv_DateExited'] = df['Program End Date'].map(
    lambda x: np.NaN if is_null(x) else pd.to_datetime(x))

condition_1 = df.loc[:, 'Program End Date'].notnull()

df.loc[condition_1, 'Elv_LEPStatus'] = df.loc[condition_1, 'Elv_DateExited'].map(
    lambda x: derive_from_date_exited_4y(str(x)))

df['Elv_LEPStatus'].unique()

df.to_csv(r"C:\Workspace\ny-harrisoncentral\student\Former ELLs_processed.csv", index=False)
```
