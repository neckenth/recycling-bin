```python
import pandas as pd
import numpy as np
import datetime as dt
from common import is_null

def derive_marking_period(x, marking_periods):
#     if pd.to_datetime(marking_periods['1']['start']) <= x <= pd.to_datetime(marking_periods['1']['end']):
#         return '1'
    if pd.to_datetime(marking_periods['2']['start']) <= x <= pd.to_datetime(marking_periods['2']['end']):
            return '2'
    elif pd.to_datetime(marking_periods['3']['start']) <= x <= pd.to_datetime(marking_periods['3']['end']):
        return '3'
    elif pd.to_datetime(marking_periods['4']['start']) <= x <= pd.to_datetime(marking_periods['4']['end']):
        return '2'

def derive_meeting_date(x, mp):
    mp = derive_marking_period(today, marking_periods)
    date_columns = ['POST_PROGRAM_REVIEW_DATE1',
                    'POST_PROGRAM_REVIEW_DATE2',
                    'POST_PROGRAM_REVIEW_DATE3',
                    'POST_PROGRAM_REVIEW_DATE4']
    for col in date_columns:
        if not (is_null(x.loc[col])):
            if pd.to_datetime(marking_periods['{}'.format(mp)]['start']) <= pd.to_datetime(x.loc[col]) <= pd.to_datetime(marking_periods['{}'.format(mp)]['end']):
                return x.loc[col]

def name_next_reclass_meeting(x):
    meeting_dates = ['1', '2', '3', '4']
    for date in meeting_dates:
        if x.loc['POST_PROGRAM_REVIEW_DATE{}'.format(date)] == x.loc['Elv_NextPostPrgReclassDate']:
            return 'POST_PROGRAM_REVIEW_DATE{}'.format(date)


df = pd.read_csv(r"C:\Workspace\fl-miamidadecounty\student\Demographic with Reclassification date.csv", sep=',', dtype=str)

today = dt.datetime.now()

marking_periods = \
    {
#     '1': {'start':'UNKNOWN',
#             'end': 'UNKNOWN'},
    '2': {'start': '2018-01-01',
            'end': '2018-03-23'},
    '3': {'start': '2018-03-23',
            'end': '2018-05-31'},
    '4': {'start': '2018-06-01',
            'end': '2018-06-08'}
    }

current_marking_period = derive_marking_period(x=today, marking_periods=marking_periods)

df.loc[:, 'Elv_NextPostPrgReclassDate'] = df.apply(
    lambda x: derive_meeting_date(x, mp=current_marking_period), axis=1)

df.loc[:, 'Elv_NextPostPrgReclassDate_Flag'] = df.apply(name_next_reclass_meeting, axis=1)

df.loc[:, 'Elv_Comment'] = df.apply(
    lambda x: '{}, {}, {}, {}'.format(x.loc['POST_PROGRAM_REVIEW_DATE1'], x.loc['POST_PROGRAM_REVIEW_DATE2'],
                                      x.loc['POST_PROGRAM_REVIEW_DATE3'], x.loc['POST_PROGRAM_REVIEW_DATE4'])
    if not is_null(x.loc['POST_PROGRAM_REVIEW_DATE1']) and not is_null(x.loc['POST_PROGRAM_REVIEW_DATE2']) and
    not is_null(x.loc['POST_PROGRAM_REVIEW_DATE3']) and not is_null(x.loc['POST_PROGRAM_REVIEW_DATE4']) else np.NaN, axis=1)

guardians = ['G1', 'G2', 'G3']

for p in guardians:
    df.loc[:, 'Elv_FatherName_{}'.format(p)] = df.apply(
        lambda x: x.loc['{}_Name'.format(p)] if x.loc['{}_Relationship'.format(p)] == 'F' else np.NaN, axis=1)
    df.loc[:, 'Elv_MotherName_{}'.format(p)] = df.apply(
        lambda x: x.loc['{}_Name'.format(p)] if x.loc['{}_Relationship'.format(p)] == 'M' else np.NaN, axis=1)
    df.loc[:, 'Elv_GuardianName_{}'.format(p)] = df.apply(
        lambda x: x.loc['{}_Name'.format(p)] if x.loc['{}_Relationship'.format(p)] == 'G' else np.NaN, axis=1)

parents = ['Father', 'Mother', 'Guardian']

for r in parents:
    df.loc[:, 'Elv_{}Name'.format(r)] = df[['Elv_{}Name_G1'.format(r),
                                            'Elv_{}Name_G2'.format(r),
                                            'Elv_{}Name_G3'.format(r)]].fillna('').sum(axis=1)

df.drop(labels=['Elv_FatherName_G1',
                'Elv_FatherName_G2',
                'Elv_FatherName_G3',
                'Elv_MotherName_G1',
                'Elv_MotherName_G2',
                'Elv_MotherName_G3',
                'Elv_GuardianName_G1',
                'Elv_GuardianName_G2',
                'Elv_GuardianName_G3'], axis=1, inplace=True)

df.to_csv(r"C:\Workspace\fl-miamidadecounty\student\Demographic with Reclassification date_updatedfix_test.csv", index=False)

#         copy_to_processor_directory(district,
#                                     f,
#                                     file_watcher_type=file_watcher_type)

#     end_time = time.time()
#     elapsed_time = end_time - start_time
#     message = 'Elapsed time for chunk and copy demographics is {} seconds'.format(elapsed_time)
#     logging.info(msg=message)


# def find_next_future_date(x):
#     # instead of find_next_future_date - store all 4 marking periods in 1, 2, 3, 4
#     # derive current marking period based on today's date
#     # find the date where the date falls within the _current_ marking period
#     # return that column header
#     # change over the summer for next marking periods
#     date_columns = ['POST_PROGRAM_REVIEW_DATE1',
#                     'POST_PROGRAM_REVIEW_DATE2',
#                     'POST_PROGRAM_REVIEW_DATE3',
#                     'POST_PROGRAM_REVIEW_DATE4']
#     remaining_date_columns = []
#     for col in date_columns:
#         if not (is_null(x.loc[col])):
#             if (pd.to_datetime(x.loc[col]) > dt.datetime.now()):
#                 remaining_date_columns.append(col)
#     # if all dates are in the past (before current marking period's start date)
#     # return most recent date and its column header (probably always 4)
#     try:
#         return min(x.loc[remaining_date_columns])
#     except ValueError:
#         return max(x.loc[date_columns])
#
#     # use once to derive current mp
#     # use a second time to derive column header

# def derive_marking_period(x, marking_periods):
#     if pd.to_datetime(marking_periods['1']['start']) <= x <= pd.to_datetime(marking_periods['1']['end']):
#         return '1'
#     elif pd.to_datetime(marking_periods['2']['start']) <= x <= pd.to_datetime(marking_periods['2']['end']):
#             return '2'
#     elif pd.to_datetime(marking_periods['3']['start']) <= x <= pd.to_datetime(marking_periods['3']['end']):
#         return '3'
#     elif pd.to_datetime(marking_periods['4']['start']) <= x <= pd.to_datetime(marking_periods['4']['end']):
#         return '2'

# def derive_meeting_date(x, mp):
#     mp = derive_marking_period(today)
#     date_columns = ['POST_PROGRAM_REVIEW_DATE1',
#                     'POST_PROGRAM_REVIEW_DATE2',
#                     'POST_PROGRAM_REVIEW_DATE3',
#                     'POST_PROGRAM_REVIEW_DATE4']
#     for col in date_columns:
#         if not (is_null(x.loc[col])):
#             if pd.to_datetime(marking_periods['{}'.format(mp)]['start']) <= pd.to_datetime(x.loc[col] <= pd.to_datetime(marking_periods['{}'.format(mp)]['end'])):
#                 return x.loc[col]

# def name_next_reclass_meeting(x):
#     meeting_dates = ['1', '2', '3', '4']
#     for date in meeting_dates:
#         if x.loc['POST_PROGRAM_REVIEW_DATE{}'.format(date)] == x.loc['Elv_NextPostPrgReclassDate']:
#             return 'POST_PROGRAM_REVIEW_DATE{}'.format(date)
```
