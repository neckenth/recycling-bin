```python
import sys
import os
import numpy as np
import pandas as pd
import glob

from pyep import qualify_file_name, read_delimited
from pyep.settings import WORKSPACE_ROOT
from common import is_null
from pycron import fetch_from_sftp, copy_to_processor_directory


def main():
    
    district = 'tx-planoisd'
    
    # shortens Test Forms to allow for additional forms to be added over time
    def get_form(text):
        extras = ('Form', '(')
        for e in extras:
            try:
                return text[:text.index(e)].rstrip()
            except ValueError:
                continue

    # creates unique column headers from duplicate pairs of columns
    def df_column_uniquify(df):
        df_columns = df.columns
        new_columns = []
        for item in df_columns:
            counter = 0
            newitem = item
            while newitem in new_columns:
                counter += 1
                newitem = "{}_{}".format(item, counter)
            new_columns.append(newitem)
        df.columns = new_columns
        return df

    # read in files
    subset = []
    
    for f in glob.glob('{}/{}/ipt/*.csv'.format(WORKSPACE_ROOT, district)):
        df = pd.read_csv(f, dtype=str)
        subset.append(df)
    
    df = pd.concat(subset)

    # pivot file to create 1 row/student with Oral Des and Level scores for each Test Form they took
    df_1 = pd.pivot_table(data=df,
                          index=['Student_ID', 'Grade', 'Oral_Test_Date'],
                          columns='Oral_Test_Name',
                          values=['Oral_Designation', 'Oral_Score_Level'],
                          aggfunc=np.max).reset_index(col_level=1)

    df_1.columns = [' '.join(col).strip() for col in df_1.columns.tolist()]

    # identify all columns where scores are kept - these are long at this point and contain 'editions' that are variable
    form_cols = df_1.columns[
        df_1.columns.str.startswith('Oral_Designation') | df_1.columns.str.startswith('Oral_Score')]

    shortened_form_cols = []

    # shorten these score columns to reduce risk that additional forms will appear and cause code to fail
    # keep only the first piece of the score ending in 'English' or 'Spanish'
    for col in form_cols:
        shortened_form_cols.append(get_form(col))

    # rename the columns with these new shortened names --> this will create multiple columns with the same name
    # i.e. for all tests that had multiple forms per language:
    # Pre-IPT English Form G and Pre-IPT English Form H will now both be called Pre-IPT English
    col_dict = dict(zip(form_cols, shortened_form_cols))
    df_1.rename(columns=col_dict, inplace=True)

    # because of this duplication, we need to append _1 on all duplicate column headers to 'uniquify' them
    df_1 = df_column_uniquify(df_1)

    # for score columns that now have uniquified pairs, the scores for the same test may be present in one
    # but missing from another --> use fillna to combine these columns
    # Pre-IPT Spanish only has 1 form, so this is excluded so as to not hit an indexing error by a try...except
    
    df_1.to_csv(r"C:\Workspace\tx-planoisd\ipt\test.csv", index=False)
    
    for col in df_1.columns:
        try:
            if '_1' not in col:
                if col.startswith('Oral_Designation') or col.startswith('Oral_Score'):
                    df_1[str(col)].fillna(df_1[str(col) + '_1'], inplace=True)
                    df_1[str(col)].fillna(df_1[str(col) + '_2'], inplace=True)
        except KeyError:
            print col

    # now each column in the uniquified pair has the same content -->
    # drop the columns with _1 appended to reduce redundancy and confusion
    final_cols = [c for c in df_1.columns if '_1' not in c and '_2' not in c]
    df_1 = df_1[final_cols]
    
    df_1.drop_duplicates(inplace=True)
    
    df_1.to_csv(r"C:\Workspace\tx-planoisd\ipt\Oral_export_2018_6_22_processed.csv", index=False)


if __name__ == '__main__':
    main()
```
