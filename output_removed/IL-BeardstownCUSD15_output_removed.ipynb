```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT

def main():
    district = 'il-beardstowncusd15'
    source_file_name = 'schedules for kindergarten thru 6th grade.csv'
    
    file_name = '{}/{}/schedule/{}'.format(WORKSPACE_ROOT, district, source_file_name)
    source = pd.read_csv(file_name, dtype=str)
    
    periods = ['P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7']
    
    subset = []
    
    for period in periods:
        
        df_constant = source[['ID',
                             'Grade',
                             '{}SecCd'.format(period),
                             '{}SubjCd'.format(period),
                             '{}SubjTitl'.format(period),
                             '{}Teach'.format(period),]]
        df_constant.loc[:, 'Elv_Period'] = str(period).split('P')[1]
        df_constant.loc[:, 'ID'] = source.loc[:, 'ID']
        df_constant.loc[:, 'Grade'] = source.loc[:, 'Grade']
        df_constant.loc[:, 'Elv_CourseName'] = source.loc[:, '{}SubjTitl'.format(period)]
        df_constant.loc[:, 'Elv_CourseCode'] = source.loc[:, '{}SubjCd'.format(period)]
        df_constant.loc[:, 'Elv_Section'] = source.loc[:, '{}SecCd'.format(period)]
        df_constant.loc[:, 'Elv_StaffName'] = source.loc[:, '{}Teach'.format(period)]
        
        subset.append(df_constant)
        
    df = pd.concat(subset)
    
    df = df[['ID',
            'Grade',
            'Elv_CourseName',
            'Elv_CourseCode',
            'Elv_Section',
             'Elv_Period',
            'Elv_StaffName']]
        
    df.to_csv(r"C:\Workspace\il-beardstowncusd15\schedule\schedules for kindergarten thru 6th grade_processed.csv", index=False)
main()
```
