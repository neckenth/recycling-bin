```python
import pandas as pd
import numpy as np
import glob
import os
import re

from pyep.settings import WORKSPACE_ROOT

district = 'ma-holyoke'
district_db_name = 'ESLREPS-MA-HOLYOKE'

subset_el = []
subset_read = []
subset_math = []

el_literacy_skills = {
    'Early Emergent':'1 - Early Emergent',
    'Late Emergent':'2 - Late Emergent',
    'Transitional':'3 - Transitional',
    'Probable':'4 - Probable'
}

for file in glob.glob('{}/{}/star/*/*.csv'.format(WORKSPACE_ROOT, district)):
    df = pd.read_csv(file, dtype=str, skiprows=1)
    df.loc[:, 'DistrictDBName'] = district_db_name
    df.loc[:, 'StudentTestID'] = df.loc[:, 'StateStudentID']
    df.loc[:, 'StudentLocalID'] = df.loc[:, 'StudentId']
    df.loc[:, 'StudentGradeLevel'] = df.loc[:, 'AssessmentGrade']
    df.loc[:, 'TestAdministrationDate'] = df.loc[:, 'DateTaken']
    df.loc[:, 'TestID'] = '0'
    df.loc[:, 'TestName'] = 'STAR'
    df.loc[:, 'TestTypeID'] = '0'
    df.loc[:, 'TestSubjectID'] = '0'
    df.loc[:, 'TestGradeLevel'] = df.loc[:, 'AssessmentGrade']
    df.loc[:, 'SourceFileName'] = os.path.split(file)[1]
    if 'Early' in file:
        df.loc[:, 'TestTypeName'] = 'STAR Early Literacy'
        df.loc[:, 'TestSubjectName'] = 'ELA'
        df.loc[:, 'Score1Label'] = 'Percentile Rank'
        df.loc[:, 'Score1Type'] = 'Scale'
        df.loc[:, 'Score1Value'] = df.loc[:, 'PR']
        df.loc[:, 'Score2Label'] = 'Literacy Skills'
        df.loc[:, 'Score2Type'] = 'Level'
        df.loc[:, 'Score2Value'] = df.loc[:, 'LiteracySkillsClassification'].map(
            lambda x: ' '.join(re.findall('[A-Z][^A-Z]*', x))).map(el_literacy_skills)
        df.loc[:, 'Score3Label'] = 'Scale Score'
        df.loc[:, 'Score3Type'] = 'Scale'
        df.loc[:, 'Score3Value'] = df.loc[:, 'SS']
        subset_el.append(df)
    elif 'Read' in file:
        df.loc[:, 'TestTypeName'] = 'STAR Reading'
        df.loc[:, 'TestSubjectName'] = 'ELA'
        df.loc[:, 'Score1Label'] = 'Percentile Rank'
        df.loc[:, 'Score1Type'] = 'Scale'
        df.loc[:, 'Score1Value'] = df.loc[:, 'PR']
        df.loc[:, 'Score2Label'] = 'Grade Equivalency'
        df.loc[:, 'Score2Type'] = 'Scale'
        df.loc[:, 'Score2Value'] = df.loc[:, 'GE']
        df.loc[:, 'Score3Label'] = 'Scale Score'
        df.loc[:, 'Score3Type'] = 'Scale'
        df.loc[:, 'Score3Value'] = df.loc[:, 'SS']
        subset_read.append(df)
    elif 'Math' in file:
        df.loc[:, 'TestTypeName'] = 'STAR Math'
        df.loc[:, 'TestSubjectName'] = 'Math'
        df.loc[:, 'Score1Label'] = 'Percentile Rank'
        df.loc[:, 'Score1Type'] = 'Scale'
        df.loc[:, 'Score1Value'] = df.loc[:, 'PR']
        df.loc[:, 'Score2Label'] = 'Grade Equivalency'
        df.loc[:, 'Score2Type'] = 'Scale'
        df.loc[:, 'Score2Value'] = df.loc[:, 'GE']
        df.loc[:, 'Score3Label'] = 'Scale Score'
        df.loc[:, 'Score3Type'] = 'Scale'
        df.loc[:, 'Score3Value'] = df.loc[:, 'SS']
        subset_math.append(df)

df_el = pd.concat(subset_el)
df_read = pd.concat(subset_read)
df_math = pd.concat(subset_math)

final_df = pd.concat(objs=[df_el, df_read, df_math])

final_df.loc[:, 'Score1Value'] = final_df.apply(
    lambda x: np.NaN if str(x.loc['Score1Value']) == '0' and str(x.loc['Score2Value'][0]) != '0' else x.loc['Score1Value'], axis=1)

headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']

final_df.loc[:, headers].to_csv('{}/{}/star/ELLSTDTST_MA-Holyoke_STAR_Fall_2017.csv'.format(WORKSPACE_ROOT, district), index=False)

```
