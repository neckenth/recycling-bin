```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT

from common import is_null

def main():
    db_name = 'ESLREPS-AR-SPRINGDALE'
    district = 'ar-springdale'
    source_file_name = 'AssessmentResults.csv'
    grades_file = 'StudentsBySchool.csv'
    
    file_name = '{}/{}/map/{}'.format(WORKSPACE_ROOT, district, source_file_name)
    grades = '{}/{}/map/{}'.format(WORKSPACE_ROOT, district, grades_file)
    source = pd.read_csv(file_name, dtype=str)
    print len(source)
    grades = pd.read_csv(grades, dtype=str)
    print len(grades)
    
    source = pd.merge(source, grades, on='StudentID', how='inner')
    print len(source)
    
    test_subject_mapping = {
        'MAP Language Usage' : 'ELA',
        'MAP Reading' : 'Reading',
        'MAP Mathematics' : 'Math',
        'MAP Science' : 'Science'
    }
    
    source.rename(columns={
            'StudentID': 'StudentTestID',
            'Grade': 'StudentGradeLevel',
            'TestStartDate':'TestAdministrationDate',
            'TestRITScore':'Score1Value',
            'TestPercentile':'Score2Value'
        }, inplace=True)
    
    source.loc[:, 'DistrictDBName'] = db_name
    source.loc[:, 'StudentLocalID'] = np.NaN
    source.loc[:, 'TestGradeLevel'] = source.loc[:, 'StudentGradeLevel']
    source.loc[:, 'TestID'] = '0'
    source.loc[:, 'TestName'] = 'NWEA MAP'
    source.loc[:, 'TestTypeID'] = '0'
    source.loc[:, 'TestTypeName'] = 'MAP ' + source.loc[:, 'MeasurementScale']
    source.loc[:, 'TestSubjectID'] = '0'
    source.loc[:, 'TestSubjectName'] = source.loc[:, 'TestTypeName'].map(test_subject_mapping)
    source.loc[:, 'Score1Label'] = 'RIT'
    source.loc[:, 'Score1Type'] = 'Scale'
    source.loc[:, 'Score2Label'] = 'Percentile'
    source.loc[:, 'Score2Type'] = 'Scale'
    source.loc[:, 'SourceFileName'] = source_file_name
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    source.loc[:, headers].to_csv(r"C:\Workspace\ar-springdale\map\ELLSTDTST_AR-Springdale_MAP_Fall_2017_2018.csv", index=False)

main()
```
