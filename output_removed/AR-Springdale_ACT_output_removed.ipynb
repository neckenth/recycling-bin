```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT

def main():
    db_name = 'ESLREPS-AR-SPRINGDALE'
    district = 'ar-springdale'
    source_file_name = 'ACT-DATA_Springdale March 2016 (1).csv'
    
    file_name = '{}/{}/act/{}'.format(WORKSPACE_ROOT, district, source_file_name)
    source = pd.read_csv(file_name)
    
    domains = ['English', 'Mathematics', 'Reading', 'Science', 'Composite']
    
    subset = []
    
    for domain in domains:
        df = source[['State ID Number',
                    'Grade',
                    '{}'.format(domain)]]
        df.loc[:, 'StudentTestID'] = df.loc[:, 'State ID Number']
        df.loc[:, 'StudentGradeLevel'] = df.loc[:, 'Grade']
        df.loc[:, 'TestTypeName'] = 'ACT ' + '{}'.format(domain)
        df.loc[:, 'TestGradeLevel'] = df.loc[:, 'Grade']
        df.loc[:, 'TestSubjectName'] = '{}'.format(domain)
        df.loc[:, 'Score1Value'] = df.loc[:, '{}'.format(domain)]
        
        subset.append(df)
    
    df = pd.concat(subset)
    
    df.loc[:, 'DistrictDBName'] = db_name
    df.loc[:, 'StudentLocalID'] = np.NaN
    df.loc[:, 'TestAdministrationDate'] = pd.to_datetime('03/01/2016')
    df.loc[:, 'TestID'] = 0
    df.loc[:, 'TestName'] = 'ACT'
    df.loc[:, 'TestTypeID'] = 0
    df.loc[:, 'TestSubjectID'] = 0
    df.loc[:, 'Score1Label'] = 'Score'
    df.loc[:, 'Score1Type'] = 'Scale'
    df.loc[:, 'SourceFileName'] = source_file_name
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    df.loc[df['TestTypeName'] == 'ACT English', 'TestSubjectName'] = 'ELA'
    df.loc[df['TestTypeName'] == 'ACT Composite', 'TestSubjectName'] = 'Overall'
    df.loc[df['TestTypeName'] == 'ACT Mathematics', 'TestSubjectName'] = 'Math'
    
    df = df[df.loc[:, 'Score1Value'] != '--']
    
    df.loc[:, headers].to_csv(r"C:\Workspace\ar-springdale\act\ELLSTDTST_AR-Springdale_ACT_2016.csv", index=False)        

    
main()
```
