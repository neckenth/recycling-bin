```python
import pandas as pd
from pyep.settings import WORKSPACE_ROOT

district_db_name = 'ESLREPS-FL-PINELLASCOUNTY'
district = 'fl-pinellascounty'
source_file_name = 'ellevation_fsa_scores.tab'

source = pd.read_csv('{}\{}\standardized-tests\{}'.format(WORKSPACE_ROOT, district, source_file_name), dtype=str, sep='\t')

domains = ['Reading_', 'Math_']

test_type_mapping = {
    'Reading_': 'FSA ELA Reading',
    'Math_': 'FSA Math'
}

achievement_level_mapping = {
    '1': '1 - Inadequate',
    '2': '2 - Below Satisfactory',
    '3': '3 - Satisfactory',
    '4': '4 - Proficient',
    '5': '5 - Mastery'
}

subset = []

for domain in domains:
    
    df = source[[
            'student_id',
            'admdate',
            'admgrade',
            '{}Total_AL'.format(domain),
            '{}Total_SS'.format(domain),
            '{}Total_RS'.format(domain)
        ]]
    df.rename(columns = {
            'student_id': 'StudentLocalID',
            'admdate': 'TestAdministrationDate',
        }, inplace=True)
    df.loc[:, 'TestTypeName'] = test_type_mapping[domain]
    df.loc[:, 'TestSubjectName'] = test_type_mapping[domain].split(' ')[-1]
    df.loc[:, 'Score1Label'] = 'Achievement Level'
    df.loc[:, 'Score1Type'] = 'Level'
    df.loc[:, 'Score1Value'] = source.loc[:, '{}Total_AL'.format(domain)].map(achievement_level_mapping)
    df.loc[:, 'Score2Label'] = 'Scale Score'
    df.loc[:, 'Score2Type'] = 'Scale'
    df.loc[:, 'Score2Value'] = source.loc[:, '{}Total_SS'.format(domain)]
    df.loc[:, 'Score3Label'] = 'Raw Score'
    df.loc[:, 'Score3Type'] = 'Scale'
    df.loc[:, 'Score3Value'] = source.loc[:, '{}Total_RS'.format(domain)]
    subset.append(df)
    
df = pd.concat(subset)

df.loc[:, 'DistrictDBName'] = district_db_name
df.loc[:, 'StudentGradeLevel'] = df.loc[:, 'admgrade'].map(
    lambda x: 'K' if x == '00' else x.lstrip('0'))
df.loc[:, 'TestID'] = '0'
df.loc[:, 'TestName'] = 'FSA'
df.loc[:, 'TestTypeID'] = '0'
df.loc[:, 'TestSubjectID'] = '0'
df.loc[:, 'TestGradeLevel'] = df.loc[:, 'StudentGradeLevel']
df.loc[:, 'SourceFileName'] = source_file_name

headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']

df = df[df.loc[:, 'Score1Value'].notnull()]

df.loc[:, headers].to_csv('{}\{}\standardized-tests\ELLSTDTST_FL-Pinellas_FSA.csv'.format(WORKSPACE_ROOT, district), sep=',', index=False)
```
