```python
import pandas as pd
from pyep.settings import WORKSPACE_ROOT
from pyep import qualify_file_name
import numpy as np

def main():
    district = 'fl-browardcounty'
    district_db_name = 'ESLREPS-FL-BROWARDCOUNTY'
    source_file_name = 'BCPS_Stdt_FCAT-20180228.txt'
    
    f = r'{}/{}/standardized-tests/{}'.format(WORKSPACE_ROOT, district, source_file_name)
    
    df = pd.read_csv(f, dtype=str, sep='\t')
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    df.rename(columns={
            'External Number':'StudentTestID',
            'Student Number':'StudentLocalID',
            'Test Date':'TestAdministrationDate',
            'Test Grade Lvl':'StudentGradeLevel',
            'Sss Read Devel Score':'Score1Value',
            'Sss Read Achv Lvl':'Score2Value'
        }, inplace=True)
    
    df.loc[:, 'DistrictDBName'] = district_db_name
    df.loc[:, 'TestID'] = '0'
    df.loc[:, 'TestName'] = 'FCAT 2.0'
    df.loc[:, 'TestTypeID'] = '0'
    df.loc[:, 'TestTypeName'] = 'FCAT Reading'
    df.loc[:, 'TestSubjectID'] = '0'
    df.loc[:, 'TestSubjectName'] = 'Reading'
    df.loc[:, 'TestGradeLevel'] = df.loc[:, 'StudentGradeLevel']
    df.loc[:, 'Score1Label'] = 'Scale Score'
    df.loc[:, 'Score1Type'] = 'Scale'
    df.loc[:, 'Score2Label'] = 'Achievement Level'
    df.loc[:, 'Score2Type'] = 'Level'
    df.loc[:, 'SourceFileName'] = source_file_name
    print len(df)
    
    df = df[pd.to_datetime(df.loc[:, 'TestAdministrationDate']) >= pd.to_datetime('1/1/2016')]
    print len(df)
    
    df.loc[:, headers].to_csv('{}/{}/standardized-tests/ELLSTDTST_FL_BrowardCounty_FCAT_processed.csv'.format(WORKSPACE_ROOT, district), sep=',', index=False)
    
main()

```
