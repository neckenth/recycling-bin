```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT

from common import is_null

def main():
    db_name = 'ESLREPS-TX-PLANOISD'
    district = 'tx-planoisd'
    source_file_name = 'ComboStudentAssessment_20171204.csv'
    grades_file = 'StudentBySchool_20170503.csv'
    
    file_name = '{}/{}/map/{}'.format(WORKSPACE_ROOT, district, source_file_name)
#     grades = '{}/{}/map/{}'.format(WORKSPACE_ROOT, district, grades_file)
    source = pd.read_csv(file_name, dtype=str)
#     grades = pd.read_csv(grades)
    
#     source = pd.merge(source, grades, on='StudentID', how='inner')
    
    test_subject_mapping = {
        'MAP Language Usage' : 'ELA',
        'MAP Reading' : 'Reading',
        'MAP Mathematics' : 'Math',
        'MAP Science' : 'Science'
    }
    
    source.loc[:, 'DistrictDBName'] = db_name
    source.loc[:, 'StudentTestID'] = np.NaN
    source.loc[:, 'StudentLocalID'] = source.loc[:, 'StudentID']
    source.loc[:, 'StudentGradeLevel'] = source.loc[:, 'Grade'].map(
        lambda x: x if not x=='KG' else 'K')
    source.loc[:, 'TestGradeLevel'] = source.loc[:, 'StudentGradeLevel']
    source.loc[:, 'TestAdministrationDate'] = source.loc[:, 'TestStartDate']
    source.loc[:, 'TestID'] = '0'
    source.loc[:, 'TestName'] = 'NWEA MAP'
    source.loc[:, 'TestTypeID'] = '0'
    source.loc[:, 'TestTypeName'] = 'MAP ' + source.loc[:, 'MeasurementScale'].map(
        lambda x: x if not x == 'Science - General Science' else 'Science')
    source.loc[:, 'TestSubjectID'] = '0'
    source.loc[:, 'TestSubjectName'] = source.loc[:, 'TestTypeName'].map(test_subject_mapping)
    source.loc[:, 'Score1Label'] = 'RIT'
    source.loc[:, 'Score1Type'] = 'Scale'
    source.loc[:, 'Score1Value'] = source.loc[:, 'TestRITScore']
    source.loc[:, 'Score2Label'] = 'Percentile'
    source.loc[:, 'Score2Type'] = 'Scale'
    source.loc[:, 'Score2Value'] = source.loc[:, 'TestPercentile']
    source.loc[:, 'SourceFileName'] = source_file_name
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
#     source.replace('*', np.NaN, inplace=True)
    
    source.loc[:, headers].to_csv(r"C:\Workspace\tx-planoisd\map\ELLSTDTST_TX-PlanoISD_MAP_Fall_17_18_2.csv", index=False)

main()
```

```python
## as Custom ELP, not Standardized Test

import pandas as pd
import numpy as np
import glob
import datetime as dt

from pyep.settings import WORKSPACE_ROOT

necessary_columns = ['StudentID',
                    'Grade',
                    'MeasurementScale',
                    'TestStartDate',
                    'TestRITScore',
                    'TestPercentile']

def fix_dates(x):
    if len(x.split('/')[0]) != 2:
        if len(x.split('/')[1]) != 2:
            return '0{}/0{}/{}'.format(x.split('/')[0], x.split('/')[1], x.split('/')[2])
        else:
            return '0{}/{}/{}'.format(x.split('/')[0], x.split('/')[1], x.split('/')[2])
    else:
        return x

subset = []

for f in glob.glob('{}/tx-planoisd/map/*.csv'.format(WORKSPACE_ROOT)):
    df = pd.read_csv(f, dtype=str, usecols=necessary_columns)
    subset.append(df)
    
df = pd.concat(subset)

df.loc[df.loc[:, 'TestStartDate'].notnull(), 'TestStartDate'] = df.loc[df.loc[:, 'TestStartDate'].notnull(), 'TestStartDate'].map(
    lambda x: fix_dates(x))

df.drop_duplicates(subset=None, keep='first', inplace=True)

df_read = df[df.loc[:, 'MeasurementScale'] == 'Reading']

df_lang = df[df.loc[:, 'MeasurementScale'] == 'Language Usage']

df_read.to_csv(r"C:\Workspace\tx-planoisd\map\ComboStudentAssessment_20180622_reading.csv", index=False)

df_lang.to_csv(r"C:\Workspace\tx-planoisd\map\ComboStudentAssessment_20180622_language_usage.csv", index=False)






```
