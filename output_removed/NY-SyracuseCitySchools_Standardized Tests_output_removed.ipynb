```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT

def main():
    db_name = 'ESLREPS-NY-SYRACUSECITYSCHOOLS'
    district = 'ny-syracusecityschools'
    source_file_name = 'SCSD_Ellevation_TestData.csv'
    
    file_name = '{}/{}/standardized-tests/{}'.format(WORKSPACE_ROOT, district, source_file_name)
    source = pd.read_csv(file_name, dtype=str)
    
    condition_1 = source.loc[:, 'TEST_DESCRIPTION'].str.contains('Regents')
    
    regents_grades = ['9', '10', '11', '12']
    
    regents = source.loc[condition_1]
    
    regents.loc[regents['GRADE_ON_TEST_DATE'] == 'T+', 'GRADE_ON_TEST_DATE'] = '12'
    
    regents.loc[:, 'TestTypeName'] = regents.loc[:, 'TEST_DESCRIPTION'].map(
        lambda x: str(x).split(' (')[0])
    
    regents_test_type_map = {
        'Regents - Living Environment': 'Regents Living Environment',
        'Regents - Algebra' : 'Regents Integrated Algebra',
        'Regents - Global History' : 'Regents Global History & Geography',
        'Regents - Comp English' : 'Regents Comprehensive English',
        'Regents - Algebra CC' : 'Regents Algebra I (Common Core)',
        'Regents - US History' : 'Regents U.S. History & Government',
        'Regents - Geometry' : 'Regents Geometry',
        'Regents - Earth Science' : 'Regents Earth Science',
        'Regents - ELA CC' : 'Regents ELA (Common Core)',
        'Regents - Math_A' : 'Regents Mathematics A',
        'Regents - Chemistry' : 'Regents Chemistry',
        'Regents - Physics' : 'Regents Physics',
        'Regents - Algebra2/Trig' : 'Regents Algebra 2/Trigonometry',
        'Regents - Math_B' : 'Regents Mathematics B',
        'Regents - Spanish' : 'Regents Spanish',
        'Regents - Algebra II Common Core' : 'Regents Algebra II (Common Core)',
        'Regents - Geometry Common Core' : 'Regents Geometry (Common Core)'
    }
    
    regents_test_subject_map = {
        'Regents Algebra 2/Trigonometry' : 'Math',
        'Regents Chemistry' : 'Science',
        'Regents Global History & Geography' : 'History and Social Science',
        'Regents Living Environment' : 'Science',
        'Regents Geometry' : 'Math',
        'Regents Algebra II (Common Core)' : 'Math',
        'Regents ELA (Common Core)' : 'ELA',
        'Regents U.S. History & Government' : 'History and Social Science',
        'Regents Earth Science' : 'Science',
        'Regents Algebra I (Common Core)' : 'Math',
        'Regents Integrated Algebra' : 'Math',
        'Regents Comprehensive English' : 'ELA',
        'Regents Mathematics A' : 'Math',
        'Regents Mathematics B' : 'Math',
        'Regents Physics' : 'Science',
        'Regents Spanish' : 'World Languages',
        'Regents Algebra II (Common Core)' : 'Math',
        'Regents Geometry (Common Core)' : 'Math'
    }
    
    regents_performance_level_map = {
        '1' : 'Level 1',
        '2' : 'Level 2',
        '3' : 'Level 3',
        '4' : 'Level 4',
        '5' : 'Level 5',
        '22' : 'Unknown' #manually delete after - data inconsistency
    }
    
    regents.loc[:, 'DistrictDBName'] = db_name
    regents.loc[:, 'StudentTestID'] = np.NaN
    regents.loc[:, 'StudentLocalID'] = regents.loc[:, 'STUDENT_ID'].map(
        lambda x: str(x).rstrip(' '))
    regents.loc[:, 'StudentGradeLevel'] = regents.loc[:, 'GRADE_ON_TEST_DATE'].map(
        lambda x: x if x in regents_grades else 'X')
    regents.loc[:, 'TestAdministrationDate'] = regents.loc[:, 'TEST_DATE']
    regents.loc[:, 'TestID'] = '0'
    regents.loc[:, 'TestName'] = 'NY Regents'
    regents.loc[:, 'TestTypeID'] = '0'
    regents.loc[:, 'TestTypeName'] = regents.loc[:, 'TestTypeName'].map(regents_test_type_map)
    regents.loc[:, 'TestSubjectID'] = '0'
    regents.loc[:, 'TestSubjectName'] = regents.loc[:, 'TestTypeName'].map(regents_test_subject_map)
    regents.loc[:, 'TestGradeLevel'] = regents.loc[:, 'GRADE_ON_TEST_DATE'].map(
        lambda x: x if x in regents_grades else 'X')
    regents.loc[:, 'Score1Label'] = 'Raw Score'
    regents.loc[:, 'Score1Type'] = 'Scale'
    regents.loc[:, 'Score1Value'] = regents.loc[:, 'RAW_SCORE']
    regents.loc[:, 'Score2Label'] = 'Scale Score'
    regents.loc[:, 'Score2Type'] = 'Scale',
    regents.loc[:, 'Score2Value'] = regents.loc[:, 'SCALE_SCORE']
    regents.loc[:, 'Score3Label'] = 'Performance Level'
    regents.loc[:, 'Score3Type'] = 'Level'
    regents.loc[:, 'Score3Value'] = regents.loc[:, 'PERF_SCORE'].map(regents_performance_level_map)
    regents.loc[:, 'SourceFileName'] = source_file_name
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    regents = regents[~regents.loc[:, 'StudentGradeLevel'].isin(['X'])]
    regents = regents[~regents.loc[:, 'Score3Value'].isin(['Unknown'])]
    
#     regents = regents[pd.to_datetime(regents.loc[:, 'TestAdministrationDate']) > pd.to_datetime('01/01/2016')]
    
    regents.drop_duplicates(inplace=True)
    
    regents.loc[:, headers].to_csv(r"C:\Workspace\ny-syracusecityschools\standardized-tests\ELLSTDTST_NY-SyracuseCitySchools_Regents_historical.csv", index=False)

main()
```

```python
def main():
    db_name = 'ESLREPS-NY-SYRACUSECITYSCHOOLS'
    district = 'ny-syracusecityschools'
    source_file_name = 'SCSD_Ellevation_TestData.csv'
    
    file_name = '{}/{}/standardized-tests/{}'.format(WORKSPACE_ROOT, district, source_file_name)
    source = pd.read_csv(file_name, dtype=str)
    
    condition_1 = source.loc[:, 'TEST_DESCRIPTION'].str.contains('Regents')
    
    ei = source.loc[~condition_1]
    
    ei.loc[:, 'TestTypeName_1'] = ei.loc[:, 'TEST_DESCRIPTION'].map(
        lambda x: str(x).split(' ')[0])
    
    ei_test_type_map = {
        'ELA' : 'EI ELA',
        'Math' : 'EI Math'
    }
    
    ei_performance_level_mapping = {
        '21' : 'Level 1',
        '22' : 'Level 2', 
        '23' : 'Level 3', 
        '24' : 'Level 4'
    }
    
    ei.loc[:, 'DistrictDBName'] = db_name
    ei.loc[:, 'StudentTestID'] = np.NaN
    ei.loc[:, 'StudentLocalID'] = ei.loc[:, 'STUDENT_ID'].map(
        lambda x: str(x).rstrip(' '))
    ei.loc[:, 'StudentGradeLevel'] = ei.loc[:, 'GRADE_ON_TEST_DATE']
    ei.loc[:, 'TestAdministrationDate'] = ei.loc[:, 'TEST_DATE']
    ei.loc[:, 'TestID'] = '0'
    ei.loc[:, 'TestName'] = 'NY EI'
    ei.loc[:, 'TestTypeID'] = '0'
    ei.loc[:, 'TestTypeName'] = ei.loc[:, 'TestTypeName_1'].map(ei_test_type_map)
    ei.loc[:, 'TestSubjectID'] = '0'
    ei.loc[:, 'TestSubjectName'] = ei.loc[:, 'TestTypeName_1']
    ei.loc[:, 'TestGradeLevel'] = ei.loc[:, 'GRADE_ON_TEST_DATE']
    ei.loc[:, 'Score1Label'] = 'Raw Score'
    ei.loc[:, 'Score1Type'] = 'Scale'
    ei.loc[:, 'Score1Value'] = ei.loc[:, 'RAW_SCORE']
    ei.loc[:, 'Score2Label'] = 'Scale Score'
    ei.loc[:, 'Score2Type'] = 'Scale',
    ei.loc[:, 'Score2Value'] = ei.loc[:, 'SCALE_SCORE']
    ei.loc[:, 'Score3Label'] = 'Performance Level'
    ei.loc[:, 'Score3Type'] = 'Level'
    ei.loc[:, 'Score3Value'] = ei.loc[:, 'PERF_SCORE'].map(ei_performance_level_mapping)
    ei.loc[:, 'SourceFileName'] = source_file_name
    
    headers = ['DistrictDBName',
        'StudentTestID',
        'StudentLocalID',
        'StudentGradeLevel',
        'TestAdministrationDate',
        'TestID',
        'TestName',
        'TestTypeID',
        'TestTypeName',
        'TestSubjectID',
        'TestSubjectName',
        'TestGradeLevel',
        'Score1Label',
        'Score1Type',
        'Score1Value',
        'Score2Label',
        'Score2Type',
        'Score2Value',
        'Score3Label',
        'Score3Type',
        'Score3Value',
        'Score4Label',
        'Score4Type',
        'Score4Value',
        'Score5Label',
        'Score5Type',
        'Score5Value',
        'Score6Label',
        'Score6Type',
        'Score6Value',
        'Score7Label',
        'Score7Type',
        'Score7Value',
        'Score8Label',
        'Score8Type',
        'Score8Value',
        'SourceFileName']
        
#     ei = ei[pd.to_datetime(ei.loc[:, 'TestAdministrationDate']) > pd.to_datetime('01/01/2016')]
    
    ei.loc[:, headers].to_csv(r"C:\Workspace\ny-syracusecityschools\standardized-tests\ELLSTDTST_NY-SyracuseCitySchools_EI_historical.csv", index=False)
    
main()
```
