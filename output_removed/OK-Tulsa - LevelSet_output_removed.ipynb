```python
import pandas as pd
import numpy as np

from pyep.settings import WORKSPACE_ROOT


def main():
    db_name = 'ESLREPS-OK-TULSA'
    district = 'ok-tulsa'
    source_file_name = 'Tulsa LevelSet Results.xlsx'
    
    file_name = '{}/{}/spring tests/{}'.format(WORKSPACE_ROOT, district, source_file_name)
    
    source = pd.read_excel(file_name, converters={'Student ID':str, 
                                                  'Grade':str,
                                                  'Pre-Test Lexile Level':str,
                                                  'Interim Test Lexile Level':str,
                                                  'Post-Test Lexile Level':str})

    tests = ['Pre-Test', 'Interim Test', 'Post-Test']

    source.rename(columns = {'Student ID' : 'StudentLocalID',
                        'Grade' : 'StudentGradeLevel'},
             inplace=True)

    subset = []

    for test in tests:
        df_test = source[['StudentLocalID',
                 'StudentGradeLevel',
                 '{} Date'.format(test),
                 '{} Lexile Level'.format(test)]]

        df_test.loc[:, 'TestAdministrationDate'] = source.loc[:, '{} Date'.format(test)]
        df_test.loc[:, 'Score1Value'] = source.loc[:, '{} Lexile Level'.format(test)]
        
        subset.append(df_test)

    df = pd.concat(subset)

    df.loc[:, 'DistrictDBName'] = db_name
    df.loc[:, 'StudentTestID'] = np.NaN
    df.loc[:, 'TestGradeLevel'] = df.loc[:, 'StudentGradeLevel']
    df.loc[:, 'TestID'] = '0'
    df.loc[:, 'TestName'] = 'LevelSet'
    df.loc[:, 'TestTypeID'] = '0'
    df.loc[:, 'TestTypeName'] = 'LevelSet Reading'
    df.loc[:, 'TestSubjectID'] = '0'
    df.loc[:, 'TestSubjectName'] = 'ELA'
    df.loc[:, 'Score1Label'] = 'Lexile Level'
    df.loc[:, 'Score1Type'] = 'Scale'
    df.loc[:, 'SourceFileName'] = source_file_name
    
    headers = ['DistrictDBName',
            'StudentTestID',
            'StudentLocalID',
            'StudentGradeLevel',
            'TestAdministrationDate',
            'TestID',
            'TestName',
            'TestTypeID',
            'TestTypeName',
            'TestSubjectID',
            'TestSubjectName',
            'TestGradeLevel',
            'Score1Label',
            'Score1Type',
            'Score1Value',
            'Score2Label',
            'Score2Type',
            'Score2Value',
            'Score3Label',
            'Score3Type',
            'Score3Value',
            'Score4Label',
            'Score4Type',
            'Score4Value',
            'Score5Label',
            'Score5Type',
            'Score5Value',
            'Score6Label',
            'Score6Type',
            'Score6Value',
            'Score7Label',
            'Score7Type',
            'Score7Value',
            'Score8Label',
            'Score8Type',
            'Score8Value',
            'SourceFileName']
    
    print len(df)
    
    df = df[df.loc[:, 'TestAdministrationDate'].notnull()]
    
    print len(df)
    
    df = df[pd.to_datetime(df.loc[:, 'TestAdministrationDate']) >= pd.to_datetime('02/15/2017')]
    
    print len(df)
    
#     df.loc[:, headers].to_csv(r"C:\Workspace\ok-tulsa\spring tests\ELLSTDTST_OK-Tulsa_LevelSet_Spring_16_17.csv", index=False)
    
main()
    
```
