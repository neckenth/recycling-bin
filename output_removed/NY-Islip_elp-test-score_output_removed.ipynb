```python
import pandas as pd

df = pd.read_csv(r"C:\Workspace\ny-islip\elp-test-score\ENL Students - NYSESLAT and NYSITELL.csv", sep=',', dtype=str)

df.head()

df_nyseslat = df[df.loc[:, 'testDetail.name'].str.contains('NYSESLAT')]

print len(df_nyseslat)

df_nyseslat = df_nyseslat[~df_nyseslat.loc[:, 'testDetail.name'].str.contains('xNYSESLAT')]

print len(df_nyseslat)

df_nyseslat.head()

# print len(df_nyseslat)
df_nysitell = df[df.loc[:, 'testDetail.name'].str.contains('NYSITELL')]
```

```python
#nyseslat

import pandas as pd
import numpy as np
import os
import re

from pyep import read_delimited
from pyep.settings import WORKSPACE_ROOT

df = df_nyseslat

df['testDetail.name'].str.extract('(\w{8}:\s)(.{1,2}\s)(\w{5}\s\w{5})')

df['testDetail.name'].unique()

df['Elv_GradeLevel'] = df['testDetail.name'].str.extract('\w{8}:\s(.{1,2}\s)\w{5}\s\w{5}')

df['Elv_Domain'] = df['testDetail.name'].str.extract('\w{8}:\s.{1,2}\s(\w{5}\s\w{5})')

df = df.pivot_table(index=['student.studentNumber', 'testDetail.dateTaken', 'Elv_GradeLevel'],
                  columns='Elv_Domain',
                 values=['testDetail.rawScore', 'testDetail.scaleScore', 'testDetail.resultLevel'],
                 aggfunc=np.max).reset_index(col_level=1)

df.columns=[' '.join(col).strip() for col in df.columns.tolist()]

df['testDetail.dateTaken'] = pd.to_datetime(df['testDetail.dateTaken'], format='%m/%d/%Y', errors='coerce')

df_nyseslat_pre_2015 = df[(df['testDetail.dateTaken'] <= pd.to_datetime('01/01/2015'))]

print len(df_nyseslat_pre_2015)

df_nyseslat_post_2015 = df[(df['testDetail.dateTaken'] > pd.to_datetime('2015/01/01'))]

print len(df_nyseslat_post_2015)

df_nyseslat_pre_2015.to_csv(r"C:\Workspace\ny-islip\elp-test-score\NYSESLAT_processed_pre_2015.csv", index=False)
df_nyseslat_post_2015.to_csv(r"C:\Workspace\ny-islip\elp-test-score\NYSESLAT_processed_post_2015.csv", index=False)
```

```python
#nysitell

import pandas as pd
import numpy as np
import os
import re

from pyep import read_delimited
from pyep.settings import WORKSPACE_ROOT

df = df_nysitell

def helper(group):
    group.loc[:, 'Elv_GradeLevel'] = group.loc[:, 'Elv_GradeLevel'].max()
    return group

df['testDetail.name'].str.extract('(\w{8}:\s)(\w{5}\s\w{1,4}\s?)(.-?\d?\s)(\w{5,}\s?\w*)')

df.loc[df['testDetail.name'].str.contains('Total Score'), 'Elv_GradeLevel'] = df['testDetail.name'].str.extract('\w{8}:\s\w{5}\s\w{1,4}\s?(.-?\d?\s)\w{5,}\s?\w*')

df = df.groupby('student.studentNumber').apply(helper)

df['Elv_Domain'] = df['testDetail.name'].str.extract('\w{8}:\s\w{5}\s\w{1,4}\s?.-?\d?\s(\w{5,}\s?\w*)')

df['Elv_Level'] = df['testDetail.name'].str.extract('\w{8}:\s(\w{5}\s\w{1,4}\s?).-?\d?\s\w{5,}\s?\w*')

df=df.pivot_table(index=['student.studentNumber', 'testDetail.dateTaken', 'Elv_GradeLevel', 'Elv_Level'],
                  columns='Elv_Domain',
                 values=['testDetail.rawScore', 'testDetail.scaleScore', 'testDetail.resultLevel'],
                 aggfunc=np.max).reset_index(col_level=1)

df.columns=[' '.join(col).strip() for col in df.columns.tolist()]

print len(df)

df['testDetail.dateTaken'] = pd.to_datetime(df['testDetail.dateTaken'], format='%m/%d/%Y', errors='coerce')

df_nysitell_pre_2015 = df[(df['testDetail.dateTaken'] <= pd.to_datetime('01/01/2015'))]

print len(df_nysitell_pre_2015)

df_nysitell_post_2015 = df[(df['testDetail.dateTaken'] > pd.to_datetime('2015/01/01'))]

print len(df_nysitell_post_2015)

df_nysitell_pre_2015.to_csv(r"C:\Workspace\ny-islip\elp-test-score\nysitell_pre_2015.csv", index=False)
df_nysitell_post_2015.to_csv(r"C:\Workspace\ny-islip\elp-test-score\nysitell_post_2015.csv", index=False)
```
