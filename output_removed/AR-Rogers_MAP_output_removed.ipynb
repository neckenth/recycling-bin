```python
import pandas as pd
import numpy as np
import glob
import os

from pyep.settings import WORKSPACE_ROOT

def main():

    district = 'ar-rogers'
    district_db_name = 'ESLREPS-AR-ROGERS'

    subset = []

    for f in glob.glob('{}/{}/map/*.csv'.format(WORKSPACE_ROOT, district)):
        df = pd.read_csv(f, dtype=str)
        new_cols = [x.split('::')[1] for x in df.columns[1:]]
        new_cols.insert(0, df.columns[0])
        df.columns = new_cols
        df.loc[:, 'SourceFileName'] = os.path.split(f)[1]
        subset.append(df)

    df = pd.concat(subset)

    df.rename(columns = {
            'StudentID': 'StudentLocalID',
            'MeasurementScale': 'TestTypeName',
            'TestStartDate': 'TestAdministrationDate',
            'TestRITScore': 'Score1Value',
            'TestPercentile': 'Score2Value'
        }, inplace=True)

    test_subject_mapping = {
        'MAP Mathematics' : 'Math',
        'MAP Language Usage': 'ELA',
        'MAP Reading': 'Reading'
    }


    df.loc[:, 'DistrictDBName'] = district_db_name
    df.loc[:, 'StudentTestID'] = np.NaN
    df.loc[:, 'StudentGradeLevel'] = df.loc[:, 'GRADE'].map(
        lambda x: x[1] if len(x) == 2 and x[0] == '0' else x)
    df.loc[:, 'StudentGradeLevel'] = df.loc[:, 'StudentGradeLevel'].map(
        lambda x: 'K' if x == 'KF' else x)
    df.loc[:, 'TestGradeLevel'] = df.loc[:, 'StudentGradeLevel']
    df.loc[:, 'TestName'] = 'NWEA MAP'
    df.loc[:, 'TestID'] = '0'
    df.loc[:, 'TestTypeName'] = 'MAP ' + df.loc[:, 'TestTypeName']
    df.loc[:, 'TestTypeID'] = '0'
    df.loc[:, 'TestSubjectName'] = df.loc[:, 'TestTypeName'].map(test_subject_mapping)
    df.loc[:, 'TestSubjectID'] = '0'
    df.loc[:, 'Score1Label'] = 'RIT'
    df.loc[:, 'Score1Type'] = 'Scale'
    df.loc[:, 'Score2Label'] = 'Percentile'
    df.loc[:, 'Score2Type'] = 'Scale'

    headers = ['DistrictDBName',
                'StudentTestID',
                'StudentLocalID',
                'StudentGradeLevel',
                'TestAdministrationDate',
                'TestID',
                'TestName',
                'TestTypeID',
                'TestTypeName',
                'TestSubjectID',
                'TestSubjectName',
                'TestGradeLevel',
                'Score1Label',
                'Score1Type',
                'Score1Value',
                'Score2Label',
                'Score2Type',
                'Score2Value',
                'Score3Label',
                'Score3Type',
                'Score3Value',
                'Score4Label',
                'Score4Type',
                'Score4Value',
                'Score5Label',
                'Score5Type',
                'Score5Value',
                'Score6Label',
                'Score6Type',
                'Score6Value',
                'Score7Label',
                'Score7Type',
                'Score7Value',
                'Score8Label',
                'Score8Type',
                'Score8Value',
                'SourceFileName']

    df.loc[:, headers].to_csv('{}/{}/map/ELLSTDTST_AR-Rogers_MAP_Fall_17_18.csv'.format(WORKSPACE_ROOT, district), index=False)

main()
```
