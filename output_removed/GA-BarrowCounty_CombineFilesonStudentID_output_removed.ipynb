```python
import pandas as pd
from datetime import datetime as dt
import numpy as np


df1=pd.read_csv(r"C:\Workspace\ga-barrowcounty\student\Students.csv", sep=',', dtype=object)

#print len(df1)

df2=pd.read_csv(r"C:\Workspace\ga-barrowcounty\student\Student Dates and Services.csv", sep=',', dtype=object)

#print len(df2)

result=pd.merge(df1, df2, on='LocalStudentID', how='inner')

#print len(result)

#result.head()


def compare_dates(date):
    today = dt.today()
    diff = today - pd.to_datetime(date)
    if not pd.isnull(diff):
        return diff.days #astype('timedelta64[D]')
    else:
        return np.nan

result['elv_date_between'] = result['exitDate'].map(compare_dates)

#result.head()

def determine_status(student):
    condition_1 = (student.programStatus == 'Exited LEP')
    condition_2 = (student.elv_date_between < 365)
    condition_3 = (365 < student.elv_date_between < 730)
    condition_4 = (student.elv_date_between > 730)
    condition_5 = ()
    if condition_1 and condition_2:
        return 'M1'
    elif condition_1 and condition_3:
        return 'M2'
    elif condition_1 and condition_4:
        return 'Fully Exited'
    else:
        return 'Not Monitored'
    
result['elv_monitoring_status'] = result.apply(determine_status, axis=1)

result.head()
    
    

result.to_csv(r"C:\Workspace\ga-barrowcounty\student\Students_combined_derived.csv", sep=',', Index=False, dtype=object)

```
